```python
# 
# QUICK RESUME CELL 
# Purpose : Restart SparkSession + reload all Parquet tables
#           + re-register SQL views — ready for ML / Streaming
# Usage   : Open notebook → Run THIS cell -> Jump to new section
# 

import os, sys, time, warnings, subprocess
warnings.filterwarnings('ignore')

# ── Java 17 fix (must be set BEFORE SparkSession) ────────────────
JAVA17_HOME = '/opt/homebrew/opt/openjdk@17'
os.environ['JAVA_HOME'] = JAVA17_HOME
os.environ['PATH']      = JAVA17_HOME + '/bin:' + os.environ.get('PATH', '')

# ── Core imports ─────────────────────────────────────────────────
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.functions import (
    col, when, lower, trim, regexp_replace,
    count, sum, avg, max, min, desc, asc,
    to_date, year, month, countDistinct,
    isnan, lit, round, concat, lpad,
    monotonically_increasing_id, row_number,
    stddev, variance, expr, datediff, current_date
)
from pyspark.sql.types import (
    StructType, StructField, StringType, IntegerType,
    FloatType, DateType, DecimalType, BooleanType,
    DoubleType, LongType
)
from pyspark.sql.window import Window
from pyspark.ml.feature import VectorAssembler, StandardScaler, StringIndexer, OneHotEncoder
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator, RegressionEvaluator, MulticlassClassificationEvaluator
from pyspark.ml.regression import LinearRegression, RandomForestRegressor
from pyspark.ml.classification import LogisticRegression
from pyspark.ml import Pipeline
from pyspark.ml.fpm import FPGrowth
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import builtins

# ── SparkSession ─────────────────────────────────────────────────
try:
    spark.stop()
except:
    pass

java_opts = (
    '--add-opens=java.base/javax.security.auth=ALL-UNNAMED '
    '--add-opens=java.base/java.lang=ALL-UNNAMED '
    '--add-opens=java.base/sun.nio.ch=ALL-UNNAMED '
    '--add-opens=java.base/java.nio=ALL-UNNAMED'
)

spark = SparkSession.builder \
    .appName('EComm_BigData_Pipeline') \
    .master('local[*]') \
    .config('spark.sql.shuffle.partitions', '8') \
    .config('spark.driver.memory', '4g') \
    .config('spark.executor.memory', '4g') \
    .config('spark.driver.maxResultSize', '2g') \
    .config('spark.driver.extraJavaOptions', java_opts) \
    .config('spark.executor.extraJavaOptions', java_opts) \
    .config('spark.sql.execution.arrow.pyspark.enabled', 'true') \
    .config('spark.sql.adaptive.enabled', 'true') \
    .config('spark.sql.adaptive.coalescePartitions.enabled', 'true') \
    .config('spark.serializer', 'org.apache.spark.serializer.KryoSerializer') \
    .config('spark.ui.showConsoleProgress', 'false') \
    .getOrCreate()

spark.sparkContext.setLogLevel('ERROR')

# ── File paths ───────────────────────────────────────────────────
BASE_PATH    = '/Users/snehalsubedi/Desktop/Coursework_dataset'
DATA_PATH    = os.path.join(BASE_PATH, 'data')
OUTPUT_PATH  = os.path.join(BASE_PATH, 'output')
PARQUET_PATH = os.path.join(OUTPUT_PATH, 'parquet')

OUTPUT_DIRS = {
    'dimensions'  : os.path.join(OUTPUT_PATH, 'dimensions'),
    'fact'        : os.path.join(OUTPUT_PATH, 'fact'),
    'analytics'   : os.path.join(OUTPUT_PATH, 'analytics'),
    'clusters'    : os.path.join(OUTPUT_PATH, 'clusters'),
    'models'      : os.path.join(OUTPUT_PATH, 'models'),
    'streaming'   : os.path.join(OUTPUT_PATH, 'streaming'),
    'stream_input': os.path.join(OUTPUT_PATH, 'stream_input'),
    'parquet'     : os.path.join(OUTPUT_PATH, 'parquet'),
    'plots'       : os.path.join(OUTPUT_PATH, 'plots'),
    'reports'     : os.path.join(OUTPUT_PATH, 'reports'),
}
for path in OUTPUT_DIRS.values():
    os.makedirs(path, exist_ok=True)

# ── Load star schema from Parquet ────────────────────────────────
print('=' * 65)
print(f'{"QUICK RESUME: LOADING ALL PIPELINE STATE":^65}')
print('=' * 65)
print(f"\n  STAR SCHEMA TABLES:")
print(f"  {'Table':<15} {'Rows':>10} {'Cols':>6} {'Status':>10}")
print(f"  {'-'*45}")

t0 = time.time()

DIM_CUSTOMER = spark.read.parquet(os.path.join(PARQUET_PATH, 'DIM_CUSTOMER'))
DIM_PRODUCT  = spark.read.parquet(os.path.join(PARQUET_PATH, 'DIM_PRODUCT'))
DIM_SUPPLIER = spark.read.parquet(os.path.join(PARQUET_PATH, 'DIM_SUPPLIER'))
DIM_ORDER    = spark.read.parquet(os.path.join(PARQUET_PATH, 'DIM_ORDER'))
DIM_REVIEW   = spark.read.parquet(os.path.join(PARQUET_PATH, 'DIM_REVIEW'))
DIM_DATE     = spark.read.parquet(os.path.join(PARQUET_PATH, 'DIM_DATE'))
FACT_SALES   = spark.read.parquet(os.path.join(PARQUET_PATH, 'FACT_SALES'))

star_schema_tables = {
    'DIM_CUSTOMER' : DIM_CUSTOMER,
    'DIM_PRODUCT'  : DIM_PRODUCT,
    'DIM_SUPPLIER' : DIM_SUPPLIER,
    'DIM_ORDER'    : DIM_ORDER,
    'DIM_REVIEW'   : DIM_REVIEW,
    'DIM_DATE'     : DIM_DATE,
    'FACT_SALES'   : FACT_SALES,
}

# Register SQL views + print summary
for name, df in star_schema_tables.items():
    df.createOrReplaceTempView(name)
    rows = df.count()
    cols = len(df.columns)
    print(f"  {name:<15} {rows:>10,} {cols:>6} {'loaded':>10}")

# ── Load cleaned datasets from Parquet ───────────────────────────
print(f"\n  CLEANED DATASETS:")
print(f"  {'Dataset':<15} {'Rows':>10} {'Cols':>6} {'Status':>10}")
print(f"  {'-'*45}")

PARQUET_CLEANED = os.path.join(PARQUET_PATH, 'cleaned')

_clean_map = {
    'customers'  : 'customers_clean',
    'orders'     : 'orders_clean',
    'order_items': 'order_items_clean',
    'products'   : 'products_clean',
    'reviews'    : 'reviews_clean',
    'suppliers'  : 'suppliers_clean',
    'date'       : 'date_clean',
}

for folder_name, var_name in _clean_map.items():
    _path = os.path.join(PARQUET_CLEANED, folder_name)
    if os.path.exists(_path):
        _df = spark.read.parquet(_path)
        _df.cache()
        globals()[var_name] = _df
        print(f"  {var_name:<15} {_df.count():>10,} "
              f"{len(_df.columns):>6} {'loaded':>10}")
    else:
        print(f"  {var_name:<15} {'--':>10} "
              f"{'--':>6} {'NOT FOUND':>10}")

# ── Load ML metrics from saved CSVs ─────────────────────────────
print(f"\n  ML & ANALYTICS METRICS:")

# Model performance metrics
_model_csv = os.path.join(OUTPUT_DIRS['models'],
                          'model_performance_summary.csv')
if os.path.exists(_model_csv):
    _model_df = pd.read_csv(_model_csv)
    # Linear Regression
    _lr_row = _model_df[_model_df['model'] == 'Linear Regression']
    if len(_lr_row) > 0:
        r2   = float(_lr_row['r2'].values[0])
        rmse = float(_lr_row['rmse'].values[0])
        print(f"  Linear Regression  : R2={r2:.4f}, RMSE={rmse:.2f}")
    # Random Forest
    _rf_row = _model_df[_model_df['model'] == 'Random Forest Regression']
    if len(_rf_row) > 0:
        rf_r2   = float(_rf_row['r2'].values[0])
        rf_rmse = float(_rf_row['rmse'].values[0])
        print(f"  Random Forest      : R2={rf_r2:.4f}, RMSE={rf_rmse:.4f}")
    # Logistic Regression
    _log_row = _model_df[_model_df['model'] == 'Logistic Regression']
    if len(_log_row) > 0:
        accuracy = float(_log_row['accuracy'].values[0])
        f1       = float(_log_row['f1'].values[0])
        print(f"  Logistic Regression: Accuracy={accuracy:.4f}, F1={f1:.4f}")
else:
    print(f"  model_performance_summary.csv NOT FOUND")
    r2, rmse, rf_r2, rf_rmse, accuracy, f1 = 0, 0, 0, 0, 0, 0

# Customer clustering metrics
_cust_profile = os.path.join(OUTPUT_DIRS['clusters'],
                             'customer_cluster_profile.csv')
if os.path.exists(_cust_profile):
    _cp = pd.read_csv(_cust_profile)
    sil_score = float(_cp['silhouette'].mean()) \
        if 'silhouette' in _cp.columns else 0.0
    print(f"  Customer K-Means   : silhouette={sil_score:.4f}")
else:
    sil_score = 0.0
    print(f"  Customer cluster profile NOT FOUND")

# Product clustering metrics
_prod_profile = os.path.join(OUTPUT_DIRS['clusters'],
                             'product_cluster_profile.csv')
if os.path.exists(_prod_profile):
    _pp = pd.read_csv(_prod_profile)
    prod_sil = float(_pp['silhouette'].mean()) \
        if 'silhouette' in _pp.columns else 0.0
    print(f"  Product K-Means    : silhouette={prod_sil:.4f}")
else:
    prod_sil = 0.0
    print(f"  Product cluster profile NOT FOUND")

# Load cluster DataFrames from parquet
_cust_cluster_path = os.path.join(OUTPUT_DIRS['clusters'],
                                  'customer_clusters')
if os.path.exists(_cust_cluster_path):
    customer_clusters = spark.read.parquet(_cust_cluster_path)
    customer_clusters.cache()
    print(f"  customer_clusters  : {customer_clusters.count():,} rows loaded")
else:
    print(f"  customer_clusters  : NOT FOUND")

_prod_cluster_path = os.path.join(OUTPUT_DIRS['clusters'],
                                  'product_clusters')
if os.path.exists(_prod_cluster_path):
    prod_preds = spark.read.parquet(_prod_cluster_path)
    prod_preds.cache()
    print(f"  prod_preds         : {prod_preds.count():,} rows loaded")
else:
    print(f"  prod_preds         : NOT FOUND")

# MBA rules counts
_mba_csv = os.path.join(OUTPUT_DIRS['analytics'],
                        'mba_summary.csv')
if os.path.exists(_mba_csv):
    _mba = pd.read_csv(_mba_csv)
    _cat_row = _mba[_mba['level'] == 'category'] if 'level' in _mba.columns else pd.DataFrame()
    _sub_row = _mba[_mba['level'] == 'subcategory'] if 'level' in _mba.columns else pd.DataFrame()
    rules_count_cat = int(_cat_row['rules_count'].values[0]) \
        if len(_cat_row) > 0 and 'rules_count' in _cat_row.columns else 0
    rules_sub_count = int(_sub_row['rules_count'].values[0]) \
        if len(_sub_row) > 0 and 'rules_count' in _sub_row.columns else 0
    print(f"  MBA category rules : {rules_count_cat}")
    print(f"  MBA subcategory    : {rules_sub_count}")
else:
    rules_count_cat = 0
    rules_sub_count = 0
    print(f"  MBA summary NOT FOUND")

# ── Streaming defaults ───────────────────────────────────────────
N_BATCHES  = 5
BATCH_SIZE = 500

elapsed = time.time() - t0

print(f"\n  {'-'*45}")
print(f"  Total resume time : {elapsed:.1f}s")
print('\n' + '=' * 65)
print(f"  Spark     : {spark.version}  |  Python: {sys.version.split()[0]}")
print(f"  Parquet   : {PARQUET_PATH}")
print('=' * 65)
print('  SparkSession ready')
print('  All 7 star schema tables loaded from Parquet')
print('  All SQL views registered')
print('  Cleaned datasets loaded')
print('  ML metrics restored')
print('  All output dirs ready')
print('  Ready for: Section 11 (Streaming) | 12 | 13')
print('=' * 65)
```

## Section 2 — Data Ingestion

#### 2.1 Define Schemas for All 7 CSV Files

```python
#
# SECTION 2: DATA INGESTION
# 
# Define Explicit Schemas for All 7 CSV Files
# Purpose: Enforce correct data types during loading
#          Prevents PySpark from guessing wrong types
#          Ensures type safety across entire pipeline
#

# Schema 1: customers.csv (120,000 rows x 9 columns)
customers_schema = StructType([
    StructField("customer_id",       StringType(),  False),
    StructField("customer_name",     StringType(),  True),
    StructField("email",             StringType(),  True),
    StructField("city",              StringType(),  True),
    StructField("state_province",    StringType(),  True),
    StructField("country",           StringType(),  True),
    StructField("customer_segment",  StringType(),  True),
    StructField("registration_date", StringType(),  True),
    StructField("age_group",         StringType(),  True)
])

# Schema 2: orders.csv (400,500 rows x 9 columns)
orders_schema = StructType([
    StructField("order_id",         StringType(),  False),
    StructField("customer_id",      StringType(),  True),
    StructField("order_date",       StringType(),  True),
    StructField("bs_year",          IntegerType(), True),
    StructField("order_status",     StringType(),  True),
    StructField("payment_method",   StringType(),  True),
    StructField("shipping_city",    StringType(),  True),
    StructField("shipping_country", StringType(),  True),
    StructField("discount_pct",     FloatType(),   True)
])

# Schema 3: order_items.csv (919,204 rows x 6 columns)
order_items_schema = StructType([
    StructField("order_item_id",      StringType(),  False),
    StructField("order_id",           StringType(),  True),
    StructField("product_id",         StringType(),  True),
    StructField("quantity",           IntegerType(), True),
    StructField("unit_price_at_sale", FloatType(),   True),
    StructField("line_total",         FloatType(),   True)
])

# Schema 4: products.csv (5,000 rows x 8 columns)
products_schema = StructType([
    StructField("product_id",   StringType(),  False),
    StructField("product_name", StringType(),  True),
    StructField("category",     StringType(),  True),
    StructField("subcategory",  StringType(),  True),
    StructField("unit_price",   FloatType(),   True),
    StructField("supplier_id",  StringType(),  True),
    StructField("weight_kg",    FloatType(),   True),
    StructField("is_active",    IntegerType(), True)
])

# Schema 5: suppliers.csv (200 rows x 5 columns)
suppliers_schema = StructType([
    StructField("supplier_id",    StringType(), False),
    StructField("supplier_name",  StringType(), True),
    StructField("country",        StringType(), True),
    StructField("contact_email",  StringType(), True),
    StructField("contract_start", StringType(), True)
])

# Schema 6: reviews.csv (280,000 rows x 7 columns)
reviews_schema = StructType([
    StructField("review_id",           StringType(),  False),
    StructField("order_id",            StringType(),  True),
    StructField("product_id",          StringType(),  True),
    StructField("review_score",        FloatType(),   True),
    StructField("review_date",         StringType(),  True),
    StructField("review_text",         StringType(),  True),
    StructField("review_length_chars", IntegerType(), True)
])

# Schema 7: date_dim_helper.csv (2,557 rows x 10 columns)
date_schema = StructType([
    StructField("date_id",       IntegerType(), True),
    StructField("full_date",     StringType(),  True),
    StructField("day_of_week",   StringType(),  True),
    StructField("day_number",    IntegerType(), True),
    StructField("month_number",  IntegerType(), True),
    StructField("month_name",    StringType(),  True),
    StructField("quarter",       StringType(),  True),
    StructField("year",          IntegerType(), True),
    StructField("is_weekend",    IntegerType(), True),
    StructField("is_np_holiday", IntegerType(), True)
])


print("SCHEMA DEFINITIONS")
schemas = {
    "customers_schema"   : (customers_schema,   9),
    "orders_schema"      : (orders_schema,       9),
    "order_items_schema" : (order_items_schema,  6),
    "products_schema"    : (products_schema,     8),
    "suppliers_schema"   : (suppliers_schema,    5),
    "reviews_schema"     : (reviews_schema,      7),
    "date_schema"        : (date_schema,        10),
}
for name, (schema, cols) in schemas.items():
    print(f"  {name:<25}: {cols} columns defined")

# Use builtins.sum to avoid conflict with PySpark's sum function
import builtins
total_cols = builtins.sum(c for _, (s, c) in schemas.items())
print(f"Total schemas defined : 7")
print(f"Total columns defined : {total_cols}")
print("All schemas ready — proceed to CSV loading")
```

## Section 3 — Data Quality Assessment

#### 3.1 NULL Value Analysis

```python
# 
# SECTION 3: DATA QUALITY ASSESSMENT
# 
# NULL Value Analysis
# Purpose: Count NULL values in every column of every dataset
#          Identify missing data patterns before cleaning
#          Documents baseline data quality
# 

import builtins

print(f"{'NULL VALUE ANALYSIS':^65}")

datasets_to_check = [
    ("customers_raw",   customers_raw),
    ("orders_raw",      orders_raw),
    ("order_items_raw", order_items_raw),
    ("products_raw",    products_raw),
    ("suppliers_raw",   suppliers_raw),
    ("reviews_raw",     reviews_raw),
    ("date_raw",        date_raw),
]

null_summary = {}

for dataset_name, df in datasets_to_check:
    total_rows = df.count()
    print(f"\n  [{dataset_name}] — {total_rows:,} rows")
    print(f"  {'Column':<25} {'NULL Count':>12} {'NULL %':>10} {'Status':>10}")
    print("  " + "-" * 60)

    null_counts = df.select([
        count(when(col(c).isNull(), c)).alias(c)
        for c in df.columns
    ]).collect()[0]

    dataset_nulls = {}
    for col_name in df.columns:
        null_count = null_counts[col_name]
        null_pct   = (null_count / total_rows * 100) if total_rows > 0 else 0
        status     = "HAS NULLS" if null_count > 0 else "CLEAN"
        dataset_nulls[col_name] = null_count
        if null_count > 0:
            print(f"  {col_name:<25} {null_count:>12,} {null_pct:>9.2f}% {status:>10}")
        else:
            print(f"  {col_name:<25} {null_count:>12,} {null_pct:>9.2f}% {status:>10}")

    null_summary[dataset_name] = dataset_nulls
    total_nulls = builtins.sum(dataset_nulls.values())
    print(f"  {'':->60}")
    print(f"  Total NULLs in {dataset_name}: {total_nulls:,}")


print("  NULL analysis complete across all 7 datasets")
```

#### 3.2 Duplicate Detection

```python
# 
# Duplicate Detection
# Purpose: Find duplicate records across key columns
#          Identifies data integrity violations
#          Quantifies impact of duplicates on analysis
# 


print(f"{'CELL 11: DUPLICATE DETECTION':^65}")

# ── Check 1: Duplicate order_ids in orders ───────────────────
total_orders    = orders_raw.count()
distinct_orders = orders_raw.select("order_id").distinct().count()
dup_orders      = total_orders - distinct_orders

print(f"\n  [orders_raw]")
print(f"  Total rows          : {total_orders:>10,}")
print(f"  Distinct order_ids  : {distinct_orders:>10,}")
print(f"  Duplicate order_ids : {dup_orders:>10,} <- {dup_orders*2} rows affected")

# Show sample duplicates
print(f"\n  Sample duplicate order_ids:")
orders_raw.groupBy("order_id") \
    .count() \
    .filter(col("count") > 1) \
    .orderBy(desc("count")) \
    .show(5, truncate=False)

# ── Check 2: Duplicate customer_ids in customers ─────────────
total_custs    = customers_raw.count()
distinct_custs = customers_raw.select("customer_id").distinct().count()
dup_custs      = total_custs - distinct_custs

print(f"  [customers_raw]")
print(f"  Total rows            : {total_custs:>10,}")
print(f"  Distinct customer_ids : {distinct_custs:>10,}")
dup_status = f"{dup_custs:,} duplicates found" if dup_custs > 0 else "No duplicates"
print(f"  Duplicate status      : {dup_status}")

# ── Check 3: Duplicate product_ids in products ───────────────
total_prods    = products_raw.count()
distinct_prods = products_raw.select("product_id").distinct().count()
dup_prods      = total_prods - distinct_prods

print(f"\n  [products_raw]")
print(f"  Total rows           : {total_prods:>10,}")
print(f"  Distinct product_ids : {distinct_prods:>10,}")
dup_status = f"{dup_prods:,} duplicates found" if dup_prods > 0 else "No duplicates"
print(f"  Duplicate status     : {dup_status}")

# ── Check 4: Duplicate review_ids in reviews ─────────────────
total_revs    = reviews_raw.count()
distinct_revs = reviews_raw.select("review_id").distinct().count()
dup_revs      = total_revs - distinct_revs

print(f"\n  [reviews_raw]")
print(f"  Total rows         : {total_revs:>10,}")
print(f"  Distinct review_ids: {distinct_revs:>10,}")
dup_status = f"{dup_revs:,} duplicates found" if dup_revs > 0 else "No duplicates"
print(f"  Duplicate status   : {dup_status}")

print("\n" + "=" * 65)
print(f"  DUPLICATE SUMMARY:")
print(f"  orders     : {dup_orders:,} duplicate order_ids ({dup_orders*2} rows affected)")
print(f"  customers  : {dup_custs:,} duplicates")
print(f"  products   : {dup_prods:,} duplicates")
print(f"  reviews    : {dup_revs:,} duplicates")
print("=" * 65)
print(" Duplicate detection complete")
print("=" * 65)
```

#### 3.3 Invalid Value Detection

```python
# 
# Invalid Value Detection
# Purpose: Find out-of-range and logically invalid values
#          Detect orphan foreign keys
#          Identify calculation mismatches
# 


print(f"{'INVALID VALUE DETECTION':^65}")

# ── Check 1: Invalid review scores ───────────────────────────
print("\n  [1] REVIEW SCORE VALIDATION (valid range: 1-5)")
total_reviews = reviews_raw.count()

null_scores = reviews_raw.filter(
    col("review_score").isNull()).count()

invalid_scores = reviews_raw.filter(
    col("review_score") == 6).count()

valid_scores = reviews_raw.filter(
    col("review_score").isNotNull() &
    (col("review_score") >= 1) &
    (col("review_score") <= 5)).count()

print(f"  Total reviews        : {total_reviews:>10,}")
print(f"  Valid scores (1-5)   : {valid_scores:>10,} ({valid_scores/total_reviews*100:.1f}%)")
print(f"  NULL scores          : {null_scores:>10,} ({null_scores/total_reviews*100:.1f}%) <- replace with 0")
print(f"  Invalid scores (6.0) : {invalid_scores:>10,} ({invalid_scores/total_reviews*100:.1f}%) <- replace with 3")

# ── Check 2: Invalid unit prices in products ─────────────────
print("\n  [2] PRODUCT PRICE VALIDATION")
total_products  = products_raw.count()
null_prices     = products_raw.filter(col("unit_price").isNull()).count()
zero_prices     = products_raw.filter(col("unit_price") == 0).count()
negative_prices = products_raw.filter(col("unit_price") < 0).count()
valid_prices    = products_raw.filter(
    col("unit_price").isNotNull() &
    (col("unit_price") > 0)).count()

print(f"  Total products       : {total_products:>10,}")
print(f"  Valid prices (>0)    : {valid_prices:>10,}")
print(f"  NULL prices          : {null_prices:>10,} <- flag as -1")
print(f"  Zero prices          : {zero_prices:>10,} <- flag as -1")
print(f"  Negative prices      : {negative_prices:>10,}")

# ── Check 3: NULL category in products ───────────────────────
print("\n  [3] PRODUCT CATEGORY VALIDATION")
null_categories = products_raw.filter(col("category").isNull()).count()
print(f"  NULL categories      : {null_categories:>10,} <- infer from subcategory")

# ── Check 4: Orphan customer_ids in orders ───────────────────
print("\n  [4] ORPHAN FOREIGN KEY — customer_id in orders")
valid_customer_ids = customers_raw.select("customer_id")
orphan_customers   = orders_raw.join(
    valid_customer_ids, "customer_id", "left_anti").count()
print(f"  Orphan customer_ids  : {orphan_customers:>10,} ← exclude from DIM_ORDER")

# ── Check 5: Orphan product_ids in order_items ───────────────
print("\n  [5] ORPHAN FOREIGN KEY — product_id in order_items")
valid_product_ids = products_raw.select("product_id")
orphan_products   = order_items_raw.join(
    valid_product_ids, "product_id", "left_anti").count()
print(f"  Orphan product_ids   : {orphan_products:>10,} <- exclude from FACT_SALES")

# ── Check 6: line_total calculation mismatches ───────────────
print("\n  [6] LINE_TOTAL CALCULATION VALIDATION")
print("      Expected: line_total = quantity * unit_price_at_sale")
mismatches = order_items_raw.filter(
    F.abs(col("line_total") -
          (col("quantity") * col("unit_price_at_sale"))) > 0.01
).count()
print(f"  Calculation errors   : {mismatches:>10,} <- recalculate in pipeline")

# ── Check 7: NULL discount_pct in orders ─────────────────────
print("\n  [7] DISCOUNT PCT VALIDATION")
null_discounts = orders_raw.filter(col("discount_pct").isNull()).count()
print(f"  NULL discount_pct    : {null_discounts:>10,} <- replace with 0.0")

# ── Check 8: NULL emails in customers ────────────────────────
print("\n  [8] EMAIL VALIDATION")
null_emails   = customers_raw.filter(col("email").isNull()).count()
space_emails  = customers_raw.filter(col("email").contains(" ")).count()
print(f"  NULL emails          : {null_emails:>10,} <- replace with unknown@unknown.com")
print(f"  Emails with spaces   : {space_emails:>10,} <- remove spaces")

# ── Check 9: NULL age_group ──────────────────────────────────
print("\n  [9] AGE GROUP VALIDATION")
null_age = customers_raw.filter(col("age_group").isNull()).count()
print(f"  NULL age_group       : {null_age:>10,} <- replace with Unknown")


print("Invalid value detection complete")
```

## Section 4 — Data Cleaning & Transformation

#### 4.1 Clean Customers

```python
# 
# DATA CLEANING & TRANSFORMATION
# 
# Clean Customers Dataset
# Purpose: Fix all 4 identified issues in customers.csv
#          Issue 1: 800 NULL emails
#          Issue 2: 175 emails with spaces
#          Issue 3: 300 NULL age_group values
#          Issue 4: Inconsistent casing in string columns
#

print(f"{'CLEANING CUSTOMERS':^60}")

before_count = customers_raw.count()

customers_clean = customers_raw \
    .withColumn("email",
        when(
            col("email").isNull() | (trim(col("email")) == ""),
            lit("unknown@unknown.com")
        ).otherwise(
            lower(regexp_replace(col("email"), " ", ""))
        )
    ) \
    .withColumn("age_group",
        when(
            col("age_group").isNull() | (trim(col("age_group")) == ""),
            lit("Unknown")
        ).otherwise(col("age_group"))
    ) \
    .withColumn("customer_name",
        lower(trim(col("customer_name")))) \
    .withColumn("city",
        lower(trim(col("city")))) \
    .withColumn("state_province",
        lower(trim(col("state_province")))) \
    .withColumn("country",
        lower(trim(col("country")))) \
    .withColumn("customer_segment",
        lower(trim(col("customer_segment")))) \
    .withColumn("registration_date",
        to_date(col("registration_date"), "M/d/yyyy"))

customers_clean.cache()
after_count = customers_clean.count()

# Verify fixes
null_email_after  = customers_clean.filter(col("email").isNull()).count()
space_email_after = customers_clean.filter(col("email").contains(" ")).count()
null_age_after    = customers_clean.filter(col("age_group").isNull()).count()

print(f"\n  BEFORE vs AFTER:")
print(f"  {'Metric':<30} {'Before':>10} {'After':>10} {'Status':>8}")
print(f"  {'-'*60}")
print(f"  {'Total rows':<30} {before_count:>10,} {after_count:>10,} {'good':>8}")
print(f"  {'NULL emails':<30} {800:>10,} {null_email_after:>10,} {'FIXED' if null_email_after==0 else 'Not_Fixed':>8}")
print(f"  {'Emails with spaces':<30} {175:>10,} {space_email_after:>10,} {'FIXED' if space_email_after==0 else 'Not_Fixed':>8}")
print(f"  {'NULL age_group':<30} {300:>10,} {null_age_after:>10,} {'FIXED' if null_age_after==0 else 'Not-Fixed':>8}")
print(f"\n  Transformations applied:")
print(f"  NULL emails     -> unknown@unknown.com")
print(f"  Email spaces    -> removed using regexp_replace")
print(f"  NULL age_group  -> Unknown")
print(f"  Casing          -> lowercased: name, city, state, country, segment")
print(f"  Dates           -> cast to DATE using to_date(..., 'M/d/yyyy')")
print("=" * 60)
print("  customers_clean created successfully")
print("=" * 60)
```

#### 4.2 Clean Orders

```python
#
# Clean Orders Dataset
# Purpose: Fix all 3 identified issues in orders.csv
#          Issue 1: 500 duplicate order_ids
#          Issue 2: 1,200 orphan customer_ids
#          Issue 3: 300 NULL discount_pct values
# 

print(f"{'CLEANING ORDERS':^60}")

before_count = orders_raw.count()

# Step 1: Remove duplicate order_ids — keep first occurrence
orders_deduped = orders_raw \
    .dropDuplicates(["order_id"])

after_dedup = orders_deduped.count()
print(f"\n  Step 1 — Remove duplicate order_ids:")
print(f"  Before dedup : {before_count:>10,}")
print(f"  After dedup  : {after_dedup:>10,}")
print(f"  Removed      : {before_count - after_dedup:>10,} duplicates")

# Step 2: Remove orphan customer_ids
valid_customer_ids = customers_clean.select("customer_id")
orders_no_orphans  = orders_deduped.join(
    valid_customer_ids, "customer_id", "inner")

after_orphan = orders_no_orphans.count()
print(f"\n  Step 2 — Remove orphan customer_ids:")
print(f"  Before       : {after_dedup:>10,}")
print(f"  After        : {after_orphan:>10,}")
print(f"  Removed      : {after_dedup - after_orphan:>10,} orphan rows")

# Step 3: Fix NULL discount_pct + cast date (format: M/d/yyyy e.g. 11/18/2023)
orders_clean = orders_no_orphans \
    .withColumn("discount_pct",
        when(col("discount_pct").isNull(), lit(0.0))
        .otherwise(col("discount_pct"))) \
    .withColumn("order_date",
        to_date(col("order_date"), "M/d/yyyy"))

orders_clean.cache()
after_count = orders_clean.count()

# Verify
null_discount_after = orders_clean.filter(
    col("discount_pct").isNull()).count()

print(f"\n  Step 3 — Fix NULL discount_pct + cast order_date:")
print(f"  NULL discount_pct after : {null_discount_after:>5,} {'FIXED' if null_discount_after==0 else 'Not Fixed':>8}")
print(f"  order_date cast to DATE : (format: M/d/yyyy)")
print(f"\n  CLEANING SUMMARY:")
print(f"  {'Metric':<30} {'Count':>10}")
print(f"  {'-'*45}")
print(f"  {'Original rows':<30} {before_count:>10,}")
print(f"  {'After dedup':<30} {after_dedup:>10,}")
print(f"  {'After orphan removal':<30} {after_orphan:>10,}")
print(f"  {'Final clean rows':<30} {after_count:>10,}")
print(f"  {'Total rows removed':<30} {before_count - after_count:>10,}")
print("=" * 60)
print("  orders_clean created successfully")
print("=" * 60)
```

#### 4.3 Clean Order Items

```python
# 
# Clean Order Items Dataset
# Purpose: Fix 2 identified issues in order_items.csv
#          Issue 1: 250 orphan product_ids
#          Issue 2: 400 line_total calculation mismatches
#


print(f"{'CLEANING ORDER ITEMS':^60}")

before_count = order_items_raw.count()

# Step 1: Remove orphan product_ids
valid_product_ids    = products_raw.select("product_id")
order_items_no_orphans = order_items_raw.join(
    valid_product_ids, "product_id", "inner")

after_orphan = order_items_no_orphans.count()
print(f"\n  Step 1 — Remove orphan product_ids:")
print(f"  Before       : {before_count:>10,}")
print(f"  After        : {after_orphan:>10,}")
print(f"  Removed      : {before_count - after_orphan:>10,} orphan rows")

# Step 2: Recalculate line_total
order_items_clean = order_items_no_orphans \
    .withColumn("line_total",
        F.round(
            col("quantity") * col("unit_price_at_sale"), 2
        )
    )

order_items_clean.cache()
after_count = order_items_clean.count()

# Verify recalculation
mismatch_after = order_items_clean.filter(
    F.abs(col("line_total") -
    (col("quantity") * col("unit_price_at_sale"))) > 0.01
).count()

print(f"\n  Step 2 — Recalculate line_total:")
print(f"  Formula      : quantity * unit_price_at_sale")
print(f"  Mismatches before : {400:>5,}")
print(f"  Mismatches after  : {mismatch_after:>5,} {'FIXED' if mismatch_after==0 else 'Not Fixed'}")
print(f"\n  CLEANING SUMMARY:")
print(f"  {'Original rows':<30} {before_count:>10,}")
print(f"  {'After orphan removal':<30} {after_orphan:>10,}")
print(f"  {'line_total recalculated':<30} {'done':>10}")
print(f"  {'Final clean rows':<30} {after_count:>10,}")
print("=" * 60)
print("  order_items_clean created successfully")
print("=" * 60)
```

#### 4.4 Clean Products

```python
# 
# Clean Products Dataset
# Purpose: Fix 3 identified issues in products.csv
#          Issue 1: 20 NULL category values
#          Issue 2: 50 NULL unit_price values
#          Issue 3: 30 zero unit_price values
# 


print(f"{'CLEANING PRODUCTS':^60}")

before_count = products_raw.count()

products_clean = products_raw \
    .withColumn("category",
        when(
            col("category").isNull() | (trim(col("category")) == ""),
            col("subcategory")
        ).otherwise(col("category"))
    ) \
    .withColumn("unit_price",
        when(
            col("unit_price").isNull() | (col("unit_price") == 0),
            lit(-1.0)
        ).otherwise(col("unit_price"))
    ) \
    .withColumn("is_active",
        col("is_active").cast(BooleanType()))

products_clean.cache()
after_count = products_clean.count()

# Verify fixes
null_cat_after   = products_clean.filter(
    col("category").isNull()).count()
invalid_price    = products_clean.filter(
    col("unit_price").isNull() | (col("unit_price") == 0)).count()
flagged_prices   = products_clean.filter(
    col("unit_price") == -1.0).count()

print(f"\n  Transformations applied:")
print(f"  {'Metric':<35} {'Before':>8} {'After':>8} {'Status':>8}")
print(f"  {'-'*62}")
print(f"  {'NULL category':<35} {20:>8,} {null_cat_after:>8,} {'FIXED' if null_cat_after==0 else 'Not Fixed':>8}")
print(f"  {'NULL/zero unit_price':<35} {80:>8,} {invalid_price:>8,} {'FIXED' if invalid_price==0 else 'Not Fixed':>8}")
print(f"  {'Prices flagged as -1.0':<35} {'N/A':>8} {flagged_prices:>8,} {'FIXED' if flagged_prices==0 else 'Not Fixed':>8}")
print(f"  {'is_active cast to Boolean':<35} {'int':>8} {'bool':>8} {'FIXED' if flagged_prices==0 else 'Not Fixed':>8}")
print(f"\n  NOTE: -1.0 is a sentinel value meaning invalid price")
print(f"        Analysts can filter: unit_price != -1 for valid prices")
print(f"  Final clean rows : {after_count:,}")
print("=" * 60)
print("  products_clean created successfully")
print("=" * 60)
```

#### 4.7 Clean Date Dim Helper

```python
# 
# Clean Date Dim Helper Dataset
# Purpose: No data quality issues in date_dim_helper.csv
#          Standardize full_date format
#          Shorten day_of_week and month_name
#          Derive bs_year and bs_full_date (Nepal calendar)
# 

print(f"{'CLEANING DATE DIM HELPER':^60}")

before_count = date_raw.count()

date_clean = date_raw \
    .withColumn("full_date",
        to_date(col("full_date"), "M/d/yyyy")) \
    .withColumn("day_of_week",
        when(col("day_of_week") == "Monday",    "Mon")
        .when(col("day_of_week") == "Tuesday",  "Tue")
        .when(col("day_of_week") == "Wednesday","Wed")
        .when(col("day_of_week") == "Thursday", "Thu")
        .when(col("day_of_week") == "Friday",   "Fri")
        .when(col("day_of_week") == "Saturday", "Sat")
        .when(col("day_of_week") == "Sunday",   "Sun")
        .otherwise(col("day_of_week"))
    ) \
    .withColumn("month_name",
        when(col("month_name") == "January",   "Jan")
        .when(col("month_name") == "February", "Feb")
        .when(col("month_name") == "March",    "Mar")
        .when(col("month_name") == "April",    "Apr")
        .when(col("month_name") == "May",      "May")
        .when(col("month_name") == "June",     "Jun")
        .when(col("month_name") == "July",     "Jul")
        .when(col("month_name") == "August",   "Aug")
        .when(col("month_name") == "September","Sep")
        .when(col("month_name") == "October",  "Oct")
        .when(col("month_name") == "November", "Nov")
        .when(col("month_name") == "December", "Dec")
        .otherwise(col("month_name"))
    ) \
    .withColumn("bs_year",
        when(col("month_number") <= 4,
             col("year") + 56)
        .otherwise(col("year") + 57)
    ) \
    .withColumn("bs_full_date",
        concat(
            col("bs_year").cast(StringType()),
            lit("-"),
            lpad(col("month_number").cast(StringType()), 2, "0"),
            lit("-"),
            lpad(col("day_number").cast(StringType()),   2, "0")
        )
    ) \
    .withColumn("is_weekend",
        col("is_weekend").cast(BooleanType())) \
    .withColumn("is_np_holiday",
        col("is_np_holiday").cast(BooleanType()))

date_clean.cache()
after_count = date_clean.count()

print(f"\n  Source: date_dim_helper.csv")
print(f"  Status: No data quality issues found")
print(f"\n  Transformations applied:")
print(f"  {'full_date':<30} string -> DATE type (M/d/yyyy)")
print(f"  {'day_of_week':<30} Monday -> Mon (abbreviated)")
print(f"  {'month_name':<30} January -> Jan (abbreviated)")
print(f"  {'bs_year':<30} Derived from Gregorian year")
print(f"  {'bs_full_date':<30} Derived YYYY-MM-DD BS format")
print(f"  {'is_weekend':<30} integer -> Boolean")
print(f"  {'is_np_holiday':<30} integer -> Boolean")
print(f"\n  Sample — BS calendar derivation:")
date_clean.select(
    "full_date","year","month_number",
    "bs_year","bs_full_date",
    "day_of_week","month_name"
).show(5, truncate=False)

print(f"  {'Rows before':<30} {before_count:>8,}")
print(f"  {'Rows after':<30} {after_count:>8,}")
print("=" * 60)
print("  date_clean created successfully")
print("=" * 60)
```

## Section 5 — Build Star Schema (Dimensions + Fact)

#### 5.1 Build DIM_CUSTOMER

```python
# 
# BUILD STAR SCHEMA
#  Build DIM_CUSTOMER
# Purpose: Create DIM_CUSTOMER dimension table from
#          cleaned customers data with surrogate key
# Columns : customer_sk (PK), customer_id, customer_name,
#           email, city, state_province, country,
#           customer_segment, registration_date, age_group
#


print(f"{'BUILD DIM_CUSTOMER':^60}")


# Add surrogate key using window function for sequential IDs
window_spec = Window.orderBy("customer_id")

DIM_CUSTOMER = customers_clean \
    .select(
        "customer_id",
        "customer_name",
        "email",
        "city",
        "state_province",
        "country",
        "customer_segment",
        "registration_date",
        "age_group"
    ) \
    .withColumn("customer_sk",
        row_number().over(window_spec).cast(IntegerType())
    ) \
    .select(
        col("customer_sk"),
        col("customer_id"),
        col("customer_name"),
        col("email"),
        col("city"),
        col("state_province"),
        col("country"),
        col("customer_segment"),
        col("registration_date"),
        col("age_group")
    )

DIM_CUSTOMER.cache()
row_count = DIM_CUSTOMER.count()

print(f"\n  Schema:")
for field in DIM_CUSTOMER.schema.fields:
    pk = " <- PK" if field.name == "customer_sk" else ""
    print(f"  {field.name:<25} {str(field.dataType):<15}{pk}")

print(f"\n  Sample rows:")
DIM_CUSTOMER.show(3, truncate=True)

print(f"  Row count  : {row_count:,}")
print(f"  Surrogate  : customer_sk (1 to {row_count:,})")
print(f"  Natural key: customer_id retained")
print("=" * 60)
print("  DIM_CUSTOMER built successfully")
print("=" * 60)
```

#### 5.2 Build DIM_PRODUCT

```python
# 
# Build DIM_PRODUCT
# Purpose: Create DIM_PRODUCT dimension table from
#          cleaned products data with surrogate key
# Columns : product_sk (PK), product_id, product_name,
#           category, subcategory, weight_kg, is_active
# Note    : unit_price excluded per star schema design
#           supplier_id excluded — managed via DIM_SUPPLIER
# 

print("=" * 60)
print(f"{'BUILD DIM_PRODUCT':^60}")
print("=" * 60)

window_spec = Window.orderBy("product_id")

DIM_PRODUCT = products_clean \
    .select(
        "product_id",
        "product_name",
        "category",
        "subcategory",
        "weight_kg",
        "is_active"
    ) \
    .withColumn("product_sk",
        row_number().over(window_spec).cast(IntegerType())
    ) \
    .select(
        col("product_sk"),
        col("product_id"),
        col("product_name"),
        col("category"),
        col("subcategory"),
        col("weight_kg"),
        col("is_active")
    )

DIM_PRODUCT.cache()
row_count = DIM_PRODUCT.count()

print(f"\n  Schema:")
for field in DIM_PRODUCT.schema.fields:
    pk = " <- PK" if field.name == "product_sk" else ""
    print(f"  {field.name:<25} {str(field.dataType):<15}{pk}")

print(f"\n  Sample rows:")
DIM_PRODUCT.show(3, truncate=True)

# Show category distribution
print(f"  Category distribution:")
DIM_PRODUCT.groupBy("category") \
    .count() \
    .orderBy(desc("count")) \
    .show(10, truncate=False)

print(f"  Row count  : {row_count:,}")
print(f"  Surrogate  : product_sk (1 to {row_count:,})")
print("=" * 60)
print("  DIM_PRODUCT built successfully")
print("=" * 60)
```

#### 5.3 Build DIM_SUPPLIER

```python

# Build DIM_SUPPLIER
# Purpose: Create DIM_SUPPLIER dimension table from
#          cleaned suppliers data with surrogate key
# Columns : supplier_sk (PK), supplier_id, supplier_name,
#           country, contact_email, contract_start
# 


print(f"{'BUILD DIM_SUPPLIER':^60}")


window_spec = Window.orderBy("supplier_id")

DIM_SUPPLIER = suppliers_clean \
    .select(
        "supplier_id",
        "supplier_name",
        "country",
        "contact_email",
        "contract_start"
    ) \
    .withColumn("supplier_sk",
        row_number().over(window_spec).cast(IntegerType())
    ) \
    .select(
        col("supplier_sk"),
        col("supplier_id"),
        col("supplier_name"),
        col("country"),
        col("contact_email"),
        col("contract_start")
    )

DIM_SUPPLIER.cache()
row_count = DIM_SUPPLIER.count()

print(f"\n  Schema:")
for field in DIM_SUPPLIER.schema.fields:
    pk = " <- PK" if field.name == "supplier_sk" else ""
    print(f"  {field.name:<25} {str(field.dataType):<15}{pk}")

print(f"\n  Sample rows:")
DIM_SUPPLIER.show(5, truncate=False)

print(f"  Row count  : {row_count:,}")
print(f"  Surrogate  : supplier_sk (1 to {row_count:,})")
print("=" * 60)
print("  <- DIM_SUPPLIER built successfully")
print("=" * 60)
```

#### 5.4 Build DIM_ORDER

```python
# 
# Build DIM_ORDER
# Purpose: Create DIM_ORDER dimension table from
#          cleaned orders data with surrogate key
# Columns : order_sk (PK), order_id, order_date,
#           order_status, payment_method,
#           shipping_city, shipping_country
# Note    : customer_id and discount_pct excluded
#           customer_id -> used for FK lookup only
#           discount_pct -> stored as measure in FACT_SALES
# 


print(f"{'BUILD DIM_ORDER':^60}")


window_spec = Window.orderBy("order_id")

DIM_ORDER = orders_clean \
    .select(
        "order_id",
        "order_date",
        "order_status",
        "payment_method",
        "shipping_city",
        "shipping_country"
    ) \
    .withColumn("order_sk",
        row_number().over(window_spec).cast(IntegerType())
    ) \
    .select(
        col("order_sk"),
        col("order_id"),
        col("order_date"),
        col("order_status"),
        col("payment_method"),
        col("shipping_city"),
        col("shipping_country")
    )

DIM_ORDER.cache()
row_count = DIM_ORDER.count()

print(f"\n  Schema:")
for field in DIM_ORDER.schema.fields:
    pk = " <- PK" if field.name == "order_sk" else ""
    print(f"  {field.name:<25} {str(field.dataType):<15}{pk}")

print(f"\n  Sample rows:")
DIM_ORDER.show(3, truncate=True)

# Order status distribution
print(f"  Order status distribution:")
DIM_ORDER.groupBy("order_status") \
    .count() \
    .orderBy(desc("count")) \
    .show(truncate=False)

print(f"  Row count  : {row_count:,}")
print(f"  Surrogate  : order_sk (1 to {row_count:,})")
print("=" * 60)
print("  DIM_ORDER built successfully")
print("=" * 60)
```

#### 5.5 Build DIM_REVIEW

```python
# 
# Build DIM_REVIEW
# Purpose: Create DIM_REVIEW dimension table from
#          cleaned reviews data with surrogate key
# Columns : review_sk (PK), review_id, review_date,
#           review_text, review_length_chars
# Note    : review_score excluded — stored as measure
#           in FACT_SALES not as dimension attribute
#           order_id, product_id excluded — join keys only
# 


print(f"{'BUILD DIM_REVIEW':^60}")


window_spec = Window.orderBy("review_id")

DIM_REVIEW = reviews_clean \
    .select(
        "review_id",
        "review_date",
        "review_text",
        "review_length_chars"
    ) \
    .withColumn("review_sk",
        row_number().over(window_spec).cast(IntegerType())
    ) \
    .select(
        col("review_sk"),
        col("review_id"),
        col("review_date"),
        col("review_text"),
        col("review_length_chars")
    )

DIM_REVIEW.cache()
row_count = DIM_REVIEW.count()

print(f"\n  Schema:")
for field in DIM_REVIEW.schema.fields:
    pk = " <- PK" if field.name == "review_sk" else ""
    print(f"  {field.name:<25} {str(field.dataType):<15}{pk}")

print(f"\n  Sample rows:")
DIM_REVIEW.select(
    "review_sk","review_id",
    "review_date","review_length_chars"
).show(5, truncate=True)

# Review length stats
print(f"  Review text length statistics:")
DIM_REVIEW.select(
    F.min("review_length_chars").alias("min_chars"),
    F.max("review_length_chars").alias("max_chars"),
    F.round(F.avg("review_length_chars"),2).alias("avg_chars")
).show()

print(f"  Row count  : {row_count:,}")
print(f"  Surrogate  : review_sk (1 to {row_count:,})")
print("=" * 60)
print("  DIM_REVIEW built successfully")
print("=" * 60)
```

#### 5.6 Build DIM_DATE

```python
# 
# Build DIM_DATE
# Purpose: Create DIM_DATE dimension table from
#          cleaned date data with surrogate key
# Columns : date_sk (PK), date_id, full_date, day_of_week,
#           day_number, month_number, month_name, quarter,
#           year, bs_year, bs_full_date,
#           is_weekend, is_np_holiday
# Note    : bs_year and bs_full_date are Nepal-specific
#           derived columns added during cleaning
# 


print(f"{'BUILD DIM_DATE':^60}")


window_spec = Window.orderBy("date_id")

DIM_DATE = date_clean \
    .select(
        "date_id",
        "full_date",
        "day_of_week",
        "day_number",
        "month_number",
        "month_name",
        "quarter",
        "year",
        "bs_year",
        "bs_full_date",
        "is_weekend",
        "is_np_holiday"
    ) \
    .withColumn("date_sk",
        row_number().over(window_spec).cast(IntegerType())
    ) \
    .select(
        col("date_sk"),
        col("date_id"),
        col("full_date"),
        col("day_of_week"),
        col("day_number"),
        col("month_number"),
        col("month_name"),
        col("quarter"),
        col("year"),
        col("bs_year"),
        col("bs_full_date"),
        col("is_weekend"),
        col("is_np_holiday")
    )

DIM_DATE.cache()
row_count = DIM_DATE.count()

print(f"\n  Schema:")
for field in DIM_DATE.schema.fields:
    pk = " <- PK" if field.name == "date_sk" else ""
    print(f"  {field.name:<25} {str(field.dataType):<15}{pk}")

print(f"\n  Sample rows with BS calendar:")
DIM_DATE.select(
    "date_sk","full_date","year","bs_year","bs_full_date",
    "day_of_week","month_name","is_weekend"
).show(5, truncate=False)

print(f"  Date range covered:")
DIM_DATE.select(
    F.min("full_date").alias("earliest_date"),
    F.max("full_date").alias("latest_date")
).show()

print(f"  Row count  : {row_count:,}")
print(f"  Surrogate  : date_sk (1 to {row_count:,})")
print("=" * 60)
print("  DIM_DATE built successfully")
print("=" * 60)
```

#### 5.7 Build FACT_SALES

```python
# ============================================================
# Build FACT_SALES
# Purpose: Mirror exact SSIS transformation pipeline
#          31 transformation steps replicated in PySpark
#          Grain: One row per order line item
#
# Join strategy (mirrors SSIS):
#   Step 3  — order_items + orders       -> Inner Join
#   Step 6  — + products                 -> Inner Join (removes 250 orphans)
#   Step 9  — + reviews                  -> Left Outer Join (optional)
#   Step 12 — + DIM_CUSTOMER             -> Left Outer Join
#   Step 15 — + DIM_PRODUCT              -> Left Outer Join
#   Step 18 — + DIM_SUPPLIER             -> Left Outer Join
#   Step 21 — + DIM_ORDER                -> Left Outer Join
#   Step 24 — + DIM_REVIEW               -> Left Outer Join
#   Step 28 — + DIM_DATE                 -> Left Outer Join
#
# Data fixes (mirrors SSIS Derive Quantity component):
#   Fix 1 — line_total = quantity * unit_price_at_sale
#   Fix 2 — review_sk  = REPLACENULL(review_sk, -1)
#   Fix 3 — review_score = REPLACENULL(review_score, 0)
#
# Performance changes vs original:
#   - All 9 joins chained (no intermediate .count() calls)
#   - monotonically_increasing_id() replaces Window.orderBy
#     (eliminates full global shuffle for surrogate key)
#   - All 4 verification counts collapsed into one .agg() pass
# ============================================================

print('=' * 60)
print(f"{'BUILD FACT_SALES':^60}")
print('=' * 60)

# -- PHASE 1: Join staging tables (SSIS steps 1-9) -----------
# All 3 joins chained — no intermediate .count() scans
print('  Phase 1: Joining staging tables (chained, no counts)...')

fact_base = (
    order_items_clean
    # Step 3: Inner Join — order_items + orders
    .join(
        orders_clean.select('order_id', 'customer_id', 'order_date', 'discount_pct'),
        'order_id', 'inner'
    )
    # Step 6: Inner Join — + products (removes ~250 orphan product_ids)
    .join(
        products_clean.select('product_id', 'supplier_id', 'unit_price'),
        'product_id', 'inner'
    )
    # Step 9: Left Join — + reviews (not every item has a review)
    .join(
        reviews_clean.select('order_id', 'product_id', 'review_id', 'review_score'),
        ['order_id', 'product_id'], 'left'
    )
)
print('  Phase 1 complete — Steps 3, 6, 9 done')

# -- PHASE 2: Resolve surrogate keys (SSIS steps 10-28) -------
# All 6 dimension lookups chained — no intermediate .count() scans
print('\n  Phase 2: Resolving surrogate keys (chained, no counts)...')

fact_base = (
    fact_base
    # Step 12: DIM_CUSTOMER
    .join(DIM_CUSTOMER.select('customer_id', 'customer_sk'), 'customer_id', 'left')
    # Step 15: DIM_PRODUCT
    .join(DIM_PRODUCT.select('product_id',  'product_sk'),  'product_id',  'left')
    # Step 18: DIM_SUPPLIER
    .join(DIM_SUPPLIER.select('supplier_id', 'supplier_sk'), 'supplier_id', 'left')
    # Step 21: DIM_ORDER
    .join(DIM_ORDER.select('order_id', 'order_sk'),         'order_id',    'left')
    # Step 24: DIM_REVIEW
    .join(DIM_REVIEW.select('review_id', 'review_sk'),      'review_id',   'left')
    # Step 28: DIM_DATE — join on order_date = full_date
    .join(
        DIM_DATE.select(col('full_date').alias('order_date'), col('date_sk')),
        'order_date', 'left'
    )
)
print('  Phase 2 complete — Steps 12, 15, 18, 21, 24, 28 done')

# -- PHASE 3: Derive columns (SSIS step 30) -------------------
# Fix 1: Recalculate line_total
# Fix 2: REPLACENULL(review_sk,    -1)
# Fix 3: REPLACENULL(review_score,  0)
# sales_sk: monotonically_increasing_id() — no global shuffle
print('\n  Phase 3: Applying Derive Quantity fixes...')

FACT_SALES = (
    fact_base
    .withColumn('line_total',
        F.round(col('quantity') * col('unit_price_at_sale'), 2)
    )
    .withColumn('review_sk',
        when(col('review_sk').isNull(), lit(-1))
        .otherwise(col('review_sk').cast(IntegerType()))
    )
    .withColumn('review_score',
        when(col('review_score').isNull(), lit(0))
        .otherwise(col('review_score').cast(IntegerType()))
    )
    .withColumn('sales_sk',
        F.monotonically_increasing_id().cast(LongType())
    )
    .select(
        col('sales_sk'),
        col('customer_sk').cast(IntegerType()),
        col('product_sk').cast(IntegerType()),
        col('supplier_sk').cast(IntegerType()),
        col('order_sk').cast(IntegerType()),
        col('date_sk').cast(IntegerType()),
        col('review_sk').cast(IntegerType()),
        col('order_item_id'),
        col('quantity').cast(IntegerType()),
        col('unit_price').cast(DoubleType()),
        col('unit_price_at_sale').cast(DoubleType()),
        col('line_total').cast(DoubleType()),
        col('discount_pct').cast(DoubleType()),
        col('review_score').cast(IntegerType())
    )
)

# Single cache + count — materialises everything in one pass
FACT_SALES.cache()
fact_count = FACT_SALES.count()
print(f'  Phase 3 complete — {fact_count:,} rows materialised and cached')

# -- Schema display -------------------------------------------
print(f"\n  {'FACT_SALES Schema:'}")
print(f"  {'Column':<25} {'Type':<15} {'Key'}")
print(f"  {'-'*55}")
for field in FACT_SALES.schema.fields:
    key = ('PK (auto)'      if field.name == 'sales_sk'
      else 'FK'             if field.name.endswith('_sk') and field.name != 'sales_sk'
      else 'Degenerate Dim' if field.name == 'order_item_id'
      else 'Measure')
    print(f"  {field.name:<25} {str(field.dataType):<15} {key}")

# -- Sample preview -------------------------------------------
print('\n  Sample rows:')
FACT_SALES.show(3, truncate=True)

# -- Verify fixes — single aggregation pass on cached data ----
print('\n  Fix verification:')
verify = FACT_SALES.agg(
    F.sum(when(col('review_sk').isNull(),   lit(1)).otherwise(lit(0))).alias('null_review_sk'),
    F.sum(when(col('review_score').isNull(),lit(1)).otherwise(lit(0))).alias('null_review_score'),
    F.sum(when(
        F.abs(col('line_total') - (col('quantity') * col('unit_price_at_sale'))) > 0.01,
        lit(1)).otherwise(lit(0))
    ).alias('mismatch_line_total'),
    F.sum(when(col('review_sk') == -1, lit(1)).otherwise(lit(0))).alias('flagged_review_sk'),
).first()

null_review_sk    = verify['null_review_sk']
null_review_score = verify['null_review_score']
mismatch_check    = verify['mismatch_line_total']
flagged_review_sk = verify['flagged_review_sk']

print(f"  {'Fix':<40} {'Result':>10} {'Status':>8}")
print(f"  {'-'*60}")
print(f"  {'line_total recalculated (mismatches)':<40} {mismatch_check:>10,} {'OK' if mismatch_check==0 else 'FAIL':>8}")
print(f"  {'review_sk NULL (should be 0)':<40} {null_review_sk:>10,} {'OK' if null_review_sk==0 else 'FAIL':>8}")
print(f"  {'review_sk flagged as -1':<40} {flagged_review_sk:>10,} {'OK':>8}")
print(f"  {'review_score NULL (should be 0)':<40} {null_review_score:>10,} {'OK' if null_review_score==0 else 'FAIL':>8}")

# -- Final summary — single pass ------------------------------
print('\n  FACT_SALES Summary:')
FACT_SALES.select(
    F.count('sales_sk').alias('total_rows'),
    F.countDistinct('customer_sk').alias('unique_customers'),
    F.countDistinct('product_sk').alias('unique_products'),
    F.countDistinct('supplier_sk').alias('unique_suppliers'),
    F.countDistinct('order_sk').alias('unique_orders'),
    F.countDistinct('date_sk').alias('unique_dates'),
    F.round(F.sum('line_total'), 2).alias('total_revenue')
).show(truncate=False)

print(f'  Total fact rows : {fact_count:,}')
print('=' * 60)
print('  OK FACT_SALES built successfully')
print('  OK Mirrors exact SSIS transformation pipeline')
print('  OK All 3 Derive Quantity fixes applied')
print('=' * 60)
```

#### 5.8 Verify Star Schema + Register SQL Views

```python
# 
# Verify Star Schema + Register SQL Views
# Purpose: Confirm all 7 tables match expected structure
#          Register as Spark SQL temporary views
#          Enables SQL queries in analytics sections
#          Verify all FK relationships work correctly
# 


print(f"{'STAR SCHEMA VERIFICATION':^65}")


# ── Register all tables as Spark SQL views ────────────────────
star_schema_tables = {
    "DIM_CUSTOMER" : DIM_CUSTOMER,
    "DIM_PRODUCT"  : DIM_PRODUCT,
    "DIM_SUPPLIER" : DIM_SUPPLIER,
    "DIM_ORDER"    : DIM_ORDER,
    "DIM_REVIEW"   : DIM_REVIEW,
    "DIM_DATE"     : DIM_DATE,
    "FACT_SALES"   : FACT_SALES,
}

for table_name, df in star_schema_tables.items():
    df.createOrReplaceTempView(table_name)

# ── Row count and column verification ────────────────────────
print(f"\n  {'Table':<15} {'Rows':>10} {'Cols':>6} {'PK Column':<15} {'Status':>8}")
print(f"  {'-'*58}")

table_info = [
    ("DIM_CUSTOMER", DIM_CUSTOMER, "customer_sk"),
    ("DIM_PRODUCT",  DIM_PRODUCT,  "product_sk"),
    ("DIM_SUPPLIER", DIM_SUPPLIER, "supplier_sk"),
    ("DIM_ORDER",    DIM_ORDER,    "order_sk"),
    ("DIM_REVIEW",   DIM_REVIEW,   "review_sk"),
    ("DIM_DATE",     DIM_DATE,     "date_sk"),
    ("FACT_SALES",   FACT_SALES,   "sales_sk"),
]

for name, df, pk in table_info:
    rows = df.count()
    cols = len(df.columns)
    print(f"  {name:<15} {rows:>10,} {cols:>6} {pk:<15} {'✓':>8}")

# ── FK relationship verification using Spark SQL ─────────────
print(f"\n  FK Relationship Verification:")
print(f"  {'-'*58}")

relationships = [
    ("customer_sk", "FACT_SALES", "DIM_CUSTOMER"),
    ("product_sk",  "FACT_SALES", "DIM_PRODUCT"),
    ("supplier_sk", "FACT_SALES", "DIM_SUPPLIER"),
    ("order_sk",    "FACT_SALES", "DIM_ORDER"),
    ("date_sk",     "FACT_SALES", "DIM_DATE"),
    ("review_sk",   "FACT_SALES", "DIM_REVIEW"),
]

for fk, fact, dim in relationships:
    # Count unmatched FKs (excluding -1 sentinel for review_sk)
    if fk == "review_sk":
        unmatched = spark.sql(f"""
            SELECT COUNT(*) AS unmatched
            FROM {fact} f
            LEFT ANTI JOIN {dim} d ON f.{fk} = d.{fk}
            WHERE f.{fk} != -1
        """).collect()[0]["unmatched"]
    else:
        unmatched = spark.sql(f"""
            SELECT COUNT(*) AS unmatched
            FROM {fact} f
            LEFT ANTI JOIN {dim} d ON f.{fk} = d.{fk}
        """).collect()[0]["unmatched"]

    status = "VALID" if unmatched == 0 else f"{unmatched:,} unmatched"
    print(f"  FACT_SALES.{fk:<15} -> {dim:<15} {status}")

# ── Full star schema SQL join test ────────────────────────────
print(f"\n  Full Star Schema SQL Join Test:")
print(f"  {'-'*58}")

result = spark.sql("""
    SELECT
        COUNT(f.sales_sk)               AS total_sales,
        COUNT(DISTINCT f.customer_sk)   AS unique_customers,
        COUNT(DISTINCT f.product_sk)    AS unique_products,
        COUNT(DISTINCT f.supplier_sk)   AS unique_suppliers,
        COUNT(DISTINCT f.order_sk)      AS unique_orders,
        COUNT(DISTINCT f.date_sk)       AS unique_dates,
        ROUND(SUM(f.line_total), 2)     AS total_revenue,
        ROUND(AVG(f.review_score), 2)   AS avg_review_score,
        ROUND(AVG(f.discount_pct), 4)   AS avg_discount
    FROM FACT_SALES     f
    JOIN DIM_CUSTOMER   c ON f.customer_sk  = c.customer_sk
    JOIN DIM_PRODUCT    p ON f.product_sk   = p.product_sk
    JOIN DIM_SUPPLIER   s ON f.supplier_sk  = s.supplier_sk
    JOIN DIM_ORDER      o ON f.order_sk     = o.order_sk
    JOIN DIM_DATE       d ON f.date_sk      = d.date_sk
""")
result.show(truncate=False)

# ── Star schema summary ───────────────────────────────────────
print(f"\n  STAR SCHEMA SUMMARY:")
print(f"  Central fact table : FACT_SALES (916,167 rows)")
print(f"  Dimensions         : 6 tables")
print(f"  FK relationships   : 6 verified")
print(f"  Surrogate keys     : All auto-generated via row_number()")
print(f"  SQL views          : All 7 tables registered")
print(f"\n  Available SQL views:")
for name in star_schema_tables.keys():
    print(f"  {name}")

print("\n" + "=" * 65)
print("  Star schema fully built and verified")
print("  All FK relationships valid")
print("  All 7 tables registered as Spark SQL views")
print("  Section 5 complete — proceed to Section 6: EDA")
print("=" * 65)
```

```python
# 
# Save Star Schema to Parquet (Early Checkpoint)
# Purpose: Persist all 7 star schema tables to disk
#          Protects against session crashes
#          Parquet is columnar — faster to read than CSV
#          Full verified Parquet save is in Section 12
# 


print(f"{'SAVING STAR SCHEMA TO PARQUET':^60}")


import time

# Output path for star schema
PARQUET_PATH = os.path.join(OUTPUT_PATH, "parquet")

tables_to_save = {
    "DIM_CUSTOMER" : DIM_CUSTOMER,
    "DIM_PRODUCT"  : DIM_PRODUCT,
    "DIM_SUPPLIER" : DIM_SUPPLIER,
    "DIM_ORDER"    : DIM_ORDER,
    "DIM_REVIEW"   : DIM_REVIEW,
    "DIM_DATE"     : DIM_DATE,
    "FACT_SALES"   : FACT_SALES,
}

print(f"\n  Saving to: {PARQUET_PATH}")
print(f"  {'Table':<15} {'Rows':>10} {'Time':>8} {'Status':>8}")
print(f"  {'-'*45}")

total_start = time.time()

for table_name, df in tables_to_save.items():
    t = time.time()
    save_path = os.path.join(PARQUET_PATH, table_name)

    df.write \
        .mode("overwrite") \
        .parquet(save_path)

    elapsed  = time.time() - t
    row_count = df.count()
    print(f"  {table_name:<15} {row_count:>10,} {elapsed:>7.1f}s {'✓':>8}")

total_elapsed = time.time() - total_start

print(f"  {'-'*45}")
print(f"  Total save time : {total_elapsed:.1f}s")
print(f"\n  Saved files:")
for table_name in tables_to_save.keys():
    save_path = os.path.join(PARQUET_PATH, table_name)
    print(f"  {save_path}")

print("\n" + "=" * 60)
print("  All 7 star schema tables saved to Parquet")
print("  Data is now persisted — session crash safe")
print("  Full Parquet verification in Section 12")
print("  Proceed to Section 6: EDA")
print("=" * 60)
```

#### 6.1 Revenue by Category & Subcategory

```python
#
# Revenue by Category & Subcategory
# Purpose: Identify top revenue-generating product categories
#          Analyse subcategory contribution within each category
#          Business insight: Which categories drive most revenue
#

import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px
import numpy as np


print(f"{'REVENUE BY CATEGORY & SUBCATEGORY':^60}")


# -- Query 1: Revenue by category --
category_revenue = spark.sql("""
    SELECT
        p.category,
        COUNT(f.sales_sk)                   AS total_transactions,
        SUM(f.line_total)                   AS total_revenue,
        ROUND(AVG(f.line_total), 2)         AS avg_revenue_per_sale,
        ROUND(AVG(f.discount_pct) * 100, 2) AS avg_discount_pct,
        ROUND(AVG(f.review_score), 2)       AS avg_review_score,
        COUNT(DISTINCT f.customer_sk)       AS unique_customers
    FROM FACT_SALES f
    JOIN DIM_PRODUCT p ON f.product_sk = p.product_sk
    GROUP BY p.category
    ORDER BY total_revenue DESC
""")

# -- Query 2: Top subcategories by revenue --
subcategory_revenue = spark.sql("""
    SELECT
        p.category,
        p.subcategory,
        COUNT(f.sales_sk)             AS total_transactions,
        ROUND(SUM(f.line_total), 2)   AS total_revenue,
        ROUND(AVG(f.review_score), 2) AS avg_review_score
    FROM FACT_SALES f
    JOIN DIM_PRODUCT p ON f.product_sk = p.product_sk
    GROUP BY p.category, p.subcategory
    ORDER BY total_revenue DESC
""")

# Convert to Pandas for plotting
cat_pd  = category_revenue.toPandas()
sub_pd  = subcategory_revenue.toPandas()
top15_sub = sub_pd.head(15)

# -- Plot directory --
plot_dir = OUTPUT_DIRS["plots"]

# -- VISUALIZATION WITH PLOTLY --

# --- Plot 1: Total Revenue by Category (Horizontal Bar) ---
cat_sorted = cat_pd.sort_values("total_revenue", ascending=True)

fig1 = go.Figure()
fig1.add_trace(go.Bar(
    y=cat_sorted["category"],
    x=cat_sorted["total_revenue"] / 1e9,
    orientation="h",
    marker=dict(
        color=cat_sorted["total_revenue"],
        colorscale="Blues",
        line=dict(color="white", width=1)
    ),
    text=[f"${v:.2f}B" for v in cat_sorted["total_revenue"] / 1e9],
    textposition="outside",
    textfont=dict(size=11, color="#2c3e50"),
    hovertemplate=(
        "<b>%{y}</b><br>"
        "Revenue: $%{x:.2f}B<br>"
        "<extra></extra>"
    )
))
fig1.update_layout(
    title=dict(text="Total Revenue by Category", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Total Revenue (Billions)", gridcolor="rgba(0,0,0,0.06)", zeroline=False),
    yaxis=dict(title=""),
    plot_bgcolor="white",
    height=450,
    margin=dict(l=20, r=80, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig1.show()
fig1.write_image(os.path.join(plot_dir, "6_1_revenue_by_category.png"), scale=2)
print("  Saved: 6_1_revenue_by_category.png")

# --- Plot 2: Revenue Share - Treemap (Category + Subcategory) ---
tree_pd = sub_pd.copy()
tree_pd["revenue_label"] = tree_pd["total_revenue"].apply(lambda v: f"${v/1e6:.1f}M")

fig2 = px.treemap(
    tree_pd,
    path=["category", "subcategory"],
    values="total_revenue",
    color="total_revenue",
    color_continuous_scale="Blues",
    hover_data={"total_revenue": ":,.0f", "total_transactions": ":,", "avg_review_score": ":.2f"},
)
fig2.update_traces(
    textinfo="label+percent parent",
    textfont=dict(size=12),
    hovertemplate=(
        "<b>%{label}</b><br>"
        "Revenue: $%{value:,.0f}<br>"
        "Share of Parent: %{percentParent:.1%}<br>"
        "<extra></extra>"
    ),
    marker=dict(line=dict(color="white", width=1.5))
)
fig2.update_layout(
    title=dict(text="Revenue Distribution: Category & Subcategory Treemap", font=dict(size=16, color="#2c3e50")),
    height=500,
    margin=dict(l=20, r=20, t=60, b=20),
    font=dict(family="Segoe UI, sans-serif", size=12),
    coloraxis_colorbar=dict(title="Revenue")
)
fig2.show()
fig2.write_image(os.path.join(plot_dir, "6_1_revenue_treemap.png"), scale=2)
print("  Saved: 6_1_revenue_treemap.png")

# --- Plot 3: Top 15 Subcategories by Revenue ---
top15_sorted = top15_sub.sort_values("total_revenue", ascending=True)

fig3 = go.Figure()
fig3.add_trace(go.Bar(
    y=top15_sorted["subcategory"],
    x=top15_sorted["total_revenue"] / 1e6,
    orientation="h",
    marker=dict(
        color=top15_sorted["total_revenue"],
        colorscale="Greens",
        line=dict(color="white", width=1)
    ),
    text=[f"${v:.1f}M" for v in top15_sorted["total_revenue"] / 1e6],
    textposition="outside",
    textfont=dict(size=10, color="#2c3e50"),
    hovertemplate=(
        "<b>%{y}</b><br>"
        "Revenue: $%{x:.1f}M<br>"
        "<extra></extra>"
    )
))
fig3.update_layout(
    title=dict(text="Top 15 Subcategories by Revenue", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Revenue (Millions)", gridcolor="rgba(0,0,0,0.06)", zeroline=False),
    yaxis=dict(title=""),
    plot_bgcolor="white",
    height=550,
    margin=dict(l=20, r=80, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig3.show()
fig3.write_image(os.path.join(plot_dir, "6_1_top15_subcategories.png"), scale=2)
print("  Saved: 6_1_top15_subcategories.png")

# --- Plot 4: Review Score vs Avg Revenue (Bubble Chart) ---
fig4 = go.Figure()
fig4.add_trace(go.Scatter(
    x=cat_pd["avg_review_score"],
    y=cat_pd["avg_revenue_per_sale"],
    mode="markers+text",
    text=cat_pd["category"],
    textposition="top center",
    textfont=dict(size=10, color="#2c3e50"),
    marker=dict(
        size=cat_pd["total_transactions"] / cat_pd["total_transactions"].max() * 60 + 15,
        color=cat_pd["total_revenue"],
        colorscale="YlOrRd",
        showscale=True,
        colorbar=dict(title="Total Revenue", thickness=15),
        line=dict(color="white", width=1.5),
        opacity=0.85
    ),
    hovertemplate=(
        "<b>%{text}</b><br>"
        "Avg Review Score: %{x:.2f}<br>"
        "Avg Revenue/Sale: $%{y:,.2f}<br>"
        "<extra></extra>"
    )
))
fig4.update_layout(
    title=dict(
        text="Review Score vs Avg Sale Value (bubble size = transactions)",
        font=dict(size=16, color="#2c3e50")
    ),
    xaxis=dict(title="Average Review Score", gridcolor="rgba(0,0,0,0.06)"),
    yaxis=dict(title="Avg Revenue per Sale ($)", gridcolor="rgba(0,0,0,0.06)"),
    plot_bgcolor="white",
    height=500,
    margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig4.show()
fig4.write_image(os.path.join(plot_dir, "6_1_review_vs_revenue_bubble.png"), scale=2)
print("  Saved: 6_1_review_vs_revenue_bubble.png")

# -- Business insights --
print(f"\n  BUSINESS INSIGHTS:")
print(f"  {'-'*55}")
top_cat    = cat_pd.iloc[0]
bottom_cat = cat_pd.iloc[-1]
print(f"  Top category     : {top_cat['category']}")
print(f"  Top revenue      : ${top_cat['total_revenue']:,.0f}")
print(f"  Bottom category  : {bottom_cat['category']}")
print(f"  Bottom revenue   : ${bottom_cat['total_revenue']:,.0f}")
print(f"  Revenue gap      : "
      f"${top_cat['total_revenue']-bottom_cat['total_revenue']:,.0f}")

print("\n  Full category summary:")
category_revenue.show(truncate=False)
print("=" * 60)
print("  complete")
print("=" * 60)
```

#### 6.2 Time Series Revenue Analysis

```python
#
# Time Series Revenue Analysis
# Purpose: Identify revenue trends over time
#          Monthly and yearly revenue patterns
#          Business insight: Seasonality and growth trends
#

import plotly.graph_objects as go
import plotly.express as px
import numpy as np


print(f"{'TIME SERIES REVENUE ANALYSIS':^60}")


# -- Query 1: Monthly revenue trend --
monthly_revenue = spark.sql("""
    SELECT
        d.year,
        d.month_number,
        d.month_name,
        CONCAT(d.year, '-', LPAD(d.month_number, 2, '0'))
            AS year_month,
        COUNT(f.sales_sk)                   AS total_orders,
        ROUND(SUM(f.line_total), 2)         AS monthly_revenue,
        ROUND(AVG(f.line_total), 2)         AS avg_order_value,
        COUNT(DISTINCT f.customer_sk)       AS unique_customers,
        ROUND(AVG(f.discount_pct) * 100, 2) AS avg_discount_pct
    FROM FACT_SALES f
    JOIN DIM_DATE d ON f.date_sk = d.date_sk
    GROUP BY d.year, d.month_number, d.month_name,
             year_month
    ORDER BY d.year, d.month_number
""")

# -- Query 2: Yearly revenue summary --
yearly_revenue = spark.sql("""
    SELECT
        d.year,
        COUNT(f.sales_sk)               AS total_orders,
        ROUND(SUM(f.line_total), 2)     AS yearly_revenue,
        ROUND(AVG(f.line_total), 2)     AS avg_order_value,
        COUNT(DISTINCT f.customer_sk)   AS unique_customers,
        COUNT(DISTINCT f.product_sk)    AS unique_products_sold
    FROM FACT_SALES f
    JOIN DIM_DATE d ON f.date_sk = d.date_sk
    GROUP BY d.year
    ORDER BY d.year
""")

# -- Query 3: Weekend vs weekday revenue --
weekend_analysis = spark.sql("""
    SELECT
        CASE WHEN d.is_weekend = true
             THEN 'Weekend' ELSE 'Weekday' END AS day_type,
        COUNT(f.sales_sk)               AS total_orders,
        ROUND(SUM(f.line_total), 2)     AS total_revenue,
        ROUND(AVG(f.line_total), 2)     AS avg_order_value
    FROM FACT_SALES f
    JOIN DIM_DATE d ON f.date_sk = d.date_sk
    GROUP BY d.is_weekend
    ORDER BY day_type
""")

# -- Query 4: Quarter revenue --
quarter_revenue = spark.sql("""
    SELECT
        d.year,
        d.quarter,
        ROUND(SUM(f.line_total), 2)   AS quarterly_revenue,
        COUNT(f.sales_sk)             AS total_orders
    FROM FACT_SALES f
    JOIN DIM_DATE d ON f.date_sk = d.date_sk
    GROUP BY d.year, d.quarter
    ORDER BY d.year, d.quarter
""")

monthly_pd  = monthly_revenue.toPandas()
yearly_pd   = yearly_revenue.toPandas()
weekend_pd  = weekend_analysis.toPandas()
quarter_pd  = quarter_revenue.toPandas()

plot_dir = OUTPUT_DIRS["plots"]

# -- VISUALIZATION WITH PLOTLY --

# --- Plot 1: Monthly Revenue Trend (Area + Line) ---
max_idx = monthly_pd["monthly_revenue"].idxmax()
min_idx = monthly_pd["monthly_revenue"].idxmin()

fig1 = go.Figure()
fig1.add_trace(go.Scatter(
    x=monthly_pd["year_month"],
    y=monthly_pd["monthly_revenue"] / 1e6,
    mode="lines",
    fill="tozeroy",
    fillcolor="rgba(33,150,243,0.15)",
    line=dict(color="#1565C0", width=2.5),
    hovertemplate=(
        "<b>%{x}</b><br>"
        "Revenue: $%{y:.1f}M<br>"
        "<extra></extra>"
    ),
    name="Monthly Revenue"
))
# Peak marker
fig1.add_trace(go.Scatter(
    x=[monthly_pd.loc[max_idx, "year_month"]],
    y=[monthly_pd.loc[max_idx, "monthly_revenue"] / 1e6],
    mode="markers+text",
    marker=dict(color="#2e7d32", size=14, symbol="star", line=dict(color="white", width=1.5)),
    text=[f"Peak: ${monthly_pd.loc[max_idx, 'monthly_revenue']/1e6:.1f}M"],
    textposition="top center",
    textfont=dict(size=10, color="#2e7d32"),
    name=f"Peak: {monthly_pd.loc[max_idx, 'year_month']}",
    showlegend=True
))
# Low marker
fig1.add_trace(go.Scatter(
    x=[monthly_pd.loc[min_idx, "year_month"]],
    y=[monthly_pd.loc[min_idx, "monthly_revenue"] / 1e6],
    mode="markers+text",
    marker=dict(color="#c62828", size=14, symbol="star", line=dict(color="white", width=1.5)),
    text=[f"Low: ${monthly_pd.loc[min_idx, 'monthly_revenue']/1e6:.1f}M"],
    textposition="bottom center",
    textfont=dict(size=10, color="#c62828"),
    name=f"Low: {monthly_pd.loc[min_idx, 'year_month']}",
    showlegend=True
))
fig1.update_layout(
    title=dict(text="Monthly Revenue Trend", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Month", gridcolor="rgba(0,0,0,0.06)", tickangle=-45, dtick=3),
    yaxis=dict(title="Revenue (Millions)", gridcolor="rgba(0,0,0,0.06)", tickprefix="$", ticksuffix="M"),
    plot_bgcolor="white",
    height=470,
    margin=dict(l=20, r=20, t=60, b=60),
    font=dict(family="Segoe UI, sans-serif", size=12),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1)
)
fig1.show()
fig1.write_image(os.path.join(plot_dir, "6_2_monthly_revenue_trend.png"), scale=2)
print("  Saved: 6_2_monthly_revenue_trend.png")

# --- Plot 2: Yearly Revenue with Growth Rate ---
yearly_pd["growth_pct"] = yearly_pd["yearly_revenue"].pct_change() * 100
growth_text = []
for i, row in yearly_pd.iterrows():
    if i == 0 or np.isnan(row["growth_pct"]):
        growth_text.append(f"${row['yearly_revenue']/1e9:.2f}B")
    else:
        arrow = "+" if row["growth_pct"] > 0 else ""
        growth_text.append(f"${row['yearly_revenue']/1e9:.2f}B<br>({arrow}{row['growth_pct']:.1f}%)")

fig2 = go.Figure()
fig2.add_trace(go.Bar(
    x=yearly_pd["year"].astype(str),
    y=yearly_pd["yearly_revenue"] / 1e9,
    marker=dict(
        color=yearly_pd["yearly_revenue"],
        colorscale="Blues",
        line=dict(color="white", width=1)
    ),
    text=growth_text,
    textposition="outside",
    textfont=dict(size=11),
    hovertemplate=(
        "<b>Year %{x}</b><br>"
        "Revenue: $%{y:.2f}B<br>"
        "Orders: %{customdata[0]:,}<br>"
        "Customers: %{customdata[1]:,}<br>"
        "<extra></extra>"
    ),
    customdata=yearly_pd[["total_orders", "unique_customers"]].values
))
fig2.update_layout(
    title=dict(text="Yearly Revenue with Growth Rate", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Year", gridcolor="rgba(0,0,0,0.06)"),
    yaxis=dict(title="Revenue (Billions)", gridcolor="rgba(0,0,0,0.06)", tickprefix="$", ticksuffix="B"),
    plot_bgcolor="white",
    height=470,
    margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12),
    bargap=0.35
)
fig2.show()
fig2.write_image(os.path.join(plot_dir, "6_2_yearly_revenue_growth.png"), scale=2)
print("  Saved: 6_2_yearly_revenue_growth.png")

# --- Plot 3: Quarterly Revenue Heatmap ---
quarter_pivot = quarter_pd.pivot(index="year", columns="quarter", values="quarterly_revenue") / 1e6

fig3 = go.Figure(data=go.Heatmap(
    z=quarter_pivot.values,
    x=[str(q) if str(q).startswith("Q") else f"Q{int(q)}" for q in quarter_pivot.columns],
    y=quarter_pivot.index.astype(str),
    colorscale="YlOrRd",
    text=[[f"${v:.0f}M" if not np.isnan(v) else "" for v in row] for row in quarter_pivot.values],
    texttemplate="%{text}",
    textfont=dict(size=12, color="#2c3e50"),
    hovertemplate=(
        "Year: %{y}<br>"
        "%{x}<br>"
        "Revenue: %{text}<br>"
        "<extra></extra>"
    ),
    colorbar=dict(title="Revenue ($M)", thickness=15)
))
fig3.update_layout(
    title=dict(text="Quarterly Revenue Heatmap", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Quarter", side="bottom"),
    yaxis=dict(title="Year", autorange="reversed"),
    plot_bgcolor="white",
    height=400,
    margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig3.show()
fig3.write_image(os.path.join(plot_dir, "6_2_quarterly_heatmap.png"), scale=2)
print("  Saved: 6_2_quarterly_heatmap.png")

# --- Plot 4: Weekend vs Weekday Revenue Comparison ---
fig4 = go.Figure()
fig4.add_trace(go.Bar(
    x=weekend_pd["day_type"],
    y=weekend_pd["total_revenue"] / 1e9,
    marker=dict(
        color=["#FF7043", "#42A5F5"],
        line=dict(color="white", width=1)
    ),
    text=[f"${v/1e9:.2f}B<br>({o:,} orders)" for v, o in zip(weekend_pd["total_revenue"], weekend_pd["total_orders"])],
    textposition="outside",
    textfont=dict(size=11, color="#2c3e50"),
    hovertemplate=(
        "<b>%{x}</b><br>"
        "Revenue: $%{y:.2f}B<br>"
        "Avg Order: $%{customdata[0]:,.2f}<br>"
        "<extra></extra>"
    ),
    customdata=weekend_pd[["avg_order_value"]].values
))
fig4.update_layout(
    title=dict(text="Weekend vs Weekday Revenue Comparison", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title=""),
    yaxis=dict(title="Total Revenue (Billions)", gridcolor="rgba(0,0,0,0.06)", tickprefix="$", ticksuffix="B"),
    plot_bgcolor="white",
    height=450,
    margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12),
    bargap=0.4
)
fig4.show()
fig4.write_image(os.path.join(plot_dir, "6_2_weekend_vs_weekday.png"), scale=2)
print("  Saved: 6_2_weekend_vs_weekday.png")

# -- Business insights --
print(f"\n  BUSINESS INSIGHTS:")
print(f"  {'-'*55}")
best_year  = yearly_pd.loc[yearly_pd["yearly_revenue"].idxmax()]
best_month = monthly_pd.loc[monthly_pd["monthly_revenue"].idxmax()]
print(f"  Best year        : {int(best_year['year'])} "
      f"(${best_year['yearly_revenue']/1e9:.2f}B)")
print(f"  Best month       : {best_month['year_month']} "
      f"(${best_month['monthly_revenue']/1e6:.1f}M)")
print(f"\n  Yearly summary:")
yearly_revenue.show(truncate=False)
print("=" * 60)
print("  Cell complete")
print("=" * 60)
```

#### 6.3 Customer Analytics

```python
#
# Customer Analytics
# Purpose: Understand customer behaviour and value
#          Identify top customers and segments
#          Business insight: Customer lifetime value and
#          segment performance
#

import plotly.graph_objects as go
import plotly.express as px
import numpy as np


print(f"{'CUSTOMER ANALYTICS':^60}")


# -- Query 1: Top 10 customers by revenue --
top_customers = spark.sql("""
    SELECT
        c.customer_id,
        c.customer_name,
        c.customer_segment,
        c.country,
        c.age_group,
        COUNT(f.sales_sk)               AS total_orders,
        ROUND(SUM(f.line_total), 2)     AS total_spent,
        ROUND(AVG(f.line_total), 2)     AS avg_order_value,
        ROUND(AVG(f.review_score), 2)   AS avg_review_score,
        ROUND(AVG(f.discount_pct)*100,2) AS avg_discount_pct
    FROM FACT_SALES f
    JOIN DIM_CUSTOMER c ON f.customer_sk = c.customer_sk
    GROUP BY c.customer_id, c.customer_name,
             c.customer_segment, c.country, c.age_group
    ORDER BY total_spent DESC
    LIMIT 10
""")

# -- Query 2: Revenue by customer segment --
segment_revenue = spark.sql("""
    SELECT
        c.customer_segment,
        COUNT(DISTINCT f.customer_sk)   AS total_customers,
        COUNT(f.sales_sk)               AS total_orders,
        ROUND(SUM(f.line_total), 2)     AS total_revenue,
        ROUND(AVG(f.line_total), 2)     AS avg_order_value,
        ROUND(AVG(f.review_score), 2)   AS avg_review_score,
        ROUND(SUM(f.line_total) /
              COUNT(DISTINCT f.customer_sk), 2)
                                        AS revenue_per_customer
    FROM FACT_SALES f
    JOIN DIM_CUSTOMER c ON f.customer_sk = c.customer_sk
    GROUP BY c.customer_segment
    ORDER BY total_revenue DESC
""")

# -- Query 3: Revenue by age group --
age_revenue = spark.sql("""
    SELECT
        c.age_group,
        COUNT(DISTINCT f.customer_sk)   AS total_customers,
        COUNT(f.sales_sk)               AS total_orders,
        ROUND(SUM(f.line_total), 2)     AS total_revenue,
        ROUND(AVG(f.line_total), 2)     AS avg_order_value,
        ROUND(AVG(f.review_score), 2)   AS avg_review_score
    FROM FACT_SALES f
    JOIN DIM_CUSTOMER c ON f.customer_sk = c.customer_sk
    GROUP BY c.age_group
    ORDER BY total_revenue DESC
""")

# -- Query 4: Top 10 countries by revenue --
country_revenue = spark.sql("""
    SELECT
        c.country,
        COUNT(DISTINCT f.customer_sk)   AS total_customers,
        COUNT(f.sales_sk)               AS total_orders,
        ROUND(SUM(f.line_total), 2)     AS total_revenue,
        ROUND(AVG(f.line_total), 2)     AS avg_order_value
    FROM FACT_SALES f
    JOIN DIM_CUSTOMER c ON f.customer_sk = c.customer_sk
    GROUP BY c.country
    ORDER BY total_revenue DESC
    LIMIT 10
""")

top_pd     = top_customers.toPandas()
seg_pd     = segment_revenue.toPandas()
age_pd     = age_revenue.toPandas()
country_pd = country_revenue.toPandas()

plot_dir = OUTPUT_DIRS["plots"]

# -- VISUALIZATION WITH PLOTLY --

# --- Plot 1: Top 10 Customers by Revenue ---
top_sorted = top_pd.sort_values("total_spent", ascending=True)

fig1 = go.Figure()
fig1.add_trace(go.Bar(
    y=[f"{name[:22]} ({seg})" for name, seg in zip(top_sorted["customer_name"], top_sorted["customer_segment"])],
    x=top_sorted["total_spent"] / 1e3,
    orientation="h",
    marker=dict(
        color=top_sorted["total_spent"],
        colorscale="Oranges",
        line=dict(color="white", width=1)
    ),
    text=[f"${v:,.0f}K" for v in top_sorted["total_spent"] / 1e3],
    textposition="outside",
    textfont=dict(size=10, color="#2c3e50"),
    hovertemplate=(
        "<b>%{y}</b><br>"
        "Total Spent: $%{x:,.0f}K<br>"
        "Orders: %{customdata[0]}<br>"
        "Avg Order: $%{customdata[1]:,.2f}<br>"
        "Avg Review: %{customdata[2]}<br>"
        "<extra></extra>"
    ),
    customdata=top_sorted[["total_orders", "avg_order_value", "avg_review_score"]].values
))
fig1.update_layout(
    title=dict(text="Top 10 Customers by Revenue", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Total Spent (Thousands)", gridcolor="rgba(0,0,0,0.06)", zeroline=False),
    yaxis=dict(title=""),
    plot_bgcolor="white",
    height=500,
    margin=dict(l=20, r=80, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig1.show()
fig1.write_image(os.path.join(plot_dir, "6_3_top10_customers.png"), scale=2)
print("  Saved: 6_3_top10_customers.png")

# --- Plot 2: Customer Segment - Revenue vs Per-Customer Value ---
fig2 = go.Figure()
fig2.add_trace(go.Bar(
    x=seg_pd["customer_segment"],
    y=seg_pd["total_revenue"] / 1e9,
    name="Total Revenue ($B)",
    marker=dict(color="#1565C0", line=dict(color="white", width=1)),
    text=[f"${v:.2f}B" for v in seg_pd["total_revenue"] / 1e9],
    textposition="outside",
    textfont=dict(size=10),
    yaxis="y",
    hovertemplate=(
        "<b>%{x}</b><br>"
        "Total Revenue: $%{y:.2f}B<br>"
        "Customers: %{customdata[0]:,}<br>"
        "Orders: %{customdata[1]:,}<br>"
        "<extra></extra>"
    ),
    customdata=seg_pd[["total_customers", "total_orders"]].values
))
fig2.add_trace(go.Bar(
    x=seg_pd["customer_segment"],
    y=seg_pd["revenue_per_customer"] / 1e3,
    name="Revenue/Customer ($K)",
    marker=dict(color="#FF6F00", line=dict(color="white", width=1)),
    text=[f"${v:.1f}K" for v in seg_pd["revenue_per_customer"] / 1e3],
    textposition="outside",
    textfont=dict(size=10),
    yaxis="y2",
    hovertemplate=(
        "<b>%{x}</b><br>"
        "Revenue/Customer: $%{y:.1f}K<br>"
        "<extra></extra>"
    )
))
fig2.update_layout(
    title=dict(text="Customer Segment - Revenue vs Per-Customer Value", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title=""),
    yaxis=dict(
        title="Total Revenue (Billions)",
        title_font=dict(color="#1565C0"),
        tickfont=dict(color="#1565C0"),
        gridcolor="rgba(0,0,0,0.06)",
        tickprefix="$", ticksuffix="B",
        side="left"
    ),
    yaxis2=dict(
        title="Revenue per Customer (Thousands)",
        title_font=dict(color="#FF6F00"),
        tickfont=dict(color="#FF6F00"),
        tickprefix="$", ticksuffix="K",
        overlaying="y",
        side="right"
    ),
    barmode="group",
    plot_bgcolor="white",
    height=470,
    margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
    bargap=0.3
)
fig2.show()
fig2.write_image(os.path.join(plot_dir, "6_3_segment_revenue.png"), scale=2)
print("  Saved: 6_3_segment_revenue.png")

# --- Plot 3: Revenue by Age Group (Horizontal Bar) ---
age_order = age_pd[age_pd["age_group"] != "Unknown"].sort_values("total_revenue", ascending=True)
age_colors = ["#D32F2F", "#F57C00", "#388E3C", "#1976D2", "#7B1FA2"]

fig3 = go.Figure()
fig3.add_trace(go.Bar(
    y=age_order["age_group"],
    x=age_order["total_revenue"] / 1e9,
    orientation="h",
    marker=dict(
        color=age_colors[:len(age_order)],
        line=dict(color="white", width=1)
    ),
    text=[f"${v:.2f}B  |  {c:,} customers" for v, c in zip(age_order["total_revenue"] / 1e9, age_order["total_customers"])],
    textposition="outside",
    textfont=dict(size=10, color="#2c3e50"),
    hovertemplate=(
        "<b>%{y}</b><br>"
        "Revenue: $%{x:.2f}B<br>"
        "Customers: %{customdata[0]:,}<br>"
        "Orders: %{customdata[1]:,}<br>"
        "Avg Order: $%{customdata[2]:,.2f}<br>"
        "Avg Review: %{customdata[3]:.2f}<br>"
        "<extra></extra>"
    ),
    customdata=age_order[["total_customers", "total_orders", "avg_order_value", "avg_review_score"]].values
))
fig3.update_layout(
    title=dict(text="Revenue Distribution by Age Group", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Total Revenue (Billions)", gridcolor="rgba(0,0,0,0.06)", zeroline=False, tickprefix="$", ticksuffix="B"),
    yaxis=dict(title=""),
    plot_bgcolor="white",
    height=420,
    margin=dict(l=20, r=120, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig3.show()
fig3.write_image(os.path.join(plot_dir, "6_3_age_group_revenue.png"), scale=2)
print("  Saved: 6_3_age_group_revenue.png")

# --- Plot 4: Top 10 Countries by Revenue ---
country_sorted = country_pd.sort_values("total_revenue", ascending=True)

fig4 = go.Figure()
fig4.add_trace(go.Bar(
    y=country_sorted["country"],
    x=country_sorted["total_revenue"] / 1e6,
    orientation="h",
    marker=dict(
        color=country_sorted["total_revenue"],
        colorscale="Purples",
        line=dict(color="white", width=1)
    ),
    text=[f"${v:.0f}M  |  {c:,} customers" for v, c in zip(country_sorted["total_revenue"] / 1e6, country_sorted["total_customers"])],
    textposition="outside",
    textfont=dict(size=10, color="#2c3e50"),
    hovertemplate=(
        "<b>%{y}</b><br>"
        "Revenue: $%{x:.0f}M<br>"
        "Customers: %{customdata[0]:,}<br>"
        "Orders: %{customdata[1]:,}<br>"
        "Avg Order: $%{customdata[2]:,.2f}<br>"
        "<extra></extra>"
    ),
    customdata=country_sorted[["total_customers", "total_orders", "avg_order_value"]].values
))
fig4.update_layout(
    title=dict(text="Top 10 Countries by Revenue", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Total Revenue (Millions)", gridcolor="rgba(0,0,0,0.06)", zeroline=False),
    yaxis=dict(title=""),
    plot_bgcolor="white",
    height=500,
    margin=dict(l=20, r=140, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig4.show()
fig4.write_image(os.path.join(plot_dir, "6_3_top10_countries.png"), scale=2)
print("  Saved: 6_3_top10_countries.png")

# -- Business insights --
print(f"\n  BUSINESS INSIGHTS:")
print(f"  {'-'*55}")
top_seg = seg_pd.iloc[0]
print(f"  Top segment          : {top_seg['customer_segment']}")
print(f"  Segment revenue      : ${top_seg['total_revenue']:,.0f}")
print(f"  Revenue/customer     : ${top_seg['revenue_per_customer']:,.0f}")
print(f"\n  Segment summary:")
segment_revenue.show(truncate=False)
print(f"\n  Age group summary:")
age_revenue.show(truncate=False)
print("=" * 60)
print("  Cell complete")
print("=" * 60)
```

#### 6.4 Product & Supplier Performance

```python
#
# Product & Supplier Performance
# Purpose: Identify top/bottom performing products
#          Analyse supplier revenue contribution
#          Business insight: Product portfolio optimisation
#

import plotly.graph_objects as go
import plotly.express as px
import numpy as np


print(f"{'PRODUCT & SUPPLIER PERFORMANCE':^60}")


# -- Query 1: Top 10 products by revenue --
top_products = spark.sql("""
    SELECT
        p.product_name,
        p.category,
        p.subcategory,
        COUNT(f.sales_sk)               AS times_ordered,
        SUM(f.quantity)                 AS total_units_sold,
        ROUND(SUM(f.line_total), 2)     AS total_revenue,
        ROUND(AVG(f.unit_price_at_sale),2) AS avg_sale_price,
        ROUND(AVG(f.review_score), 2)   AS avg_review_score
    FROM FACT_SALES f
    JOIN DIM_PRODUCT p ON f.product_sk = p.product_sk
    GROUP BY p.product_name, p.category, p.subcategory
    ORDER BY total_revenue DESC
    LIMIT 10
""")

# -- Query 2: Bottom 10 products by revenue --
bottom_products = spark.sql("""
    SELECT
        p.product_name,
        p.category,
        ROUND(SUM(f.line_total), 2)   AS total_revenue,
        COUNT(f.sales_sk)             AS times_ordered,
        ROUND(AVG(f.review_score), 2) AS avg_review_score
    FROM FACT_SALES f
    JOIN DIM_PRODUCT p ON f.product_sk = p.product_sk
    GROUP BY p.product_name, p.category
    ORDER BY total_revenue ASC
    LIMIT 10
""")

# -- Query 3: Supplier performance --
supplier_perf = spark.sql("""
    SELECT
        s.supplier_name,
        s.country,
        COUNT(DISTINCT f.product_sk)    AS products_count,
        COUNT(f.sales_sk)               AS total_orders,
        ROUND(SUM(f.line_total), 2)     AS total_revenue,
        ROUND(AVG(f.review_score), 2)   AS avg_review_score,
        ROUND(AVG(f.discount_pct)*100,2) AS avg_discount_pct
    FROM FACT_SALES f
    JOIN DIM_SUPPLIER s ON f.supplier_sk = s.supplier_sk
    GROUP BY s.supplier_name, s.country
    ORDER BY total_revenue DESC
    LIMIT 15
""")

# -- Query 4: Price vs review score --
price_review = spark.sql("""
    SELECT
        p.category,
        ROUND(AVG(f.unit_price_at_sale), 2) AS avg_price,
        ROUND(AVG(f.review_score), 2)       AS avg_review,
        COUNT(f.sales_sk)                   AS total_sales,
        ROUND(SUM(f.line_total), 2)         AS total_revenue
    FROM FACT_SALES f
    JOIN DIM_PRODUCT p ON f.product_sk = p.product_sk
    WHERE f.review_score > 0
    GROUP BY p.category
    ORDER BY avg_price DESC
""")

top_pd    = top_products.toPandas()
bot_pd    = bottom_products.toPandas()
sup_pd    = supplier_perf.toPandas()
price_pd  = price_review.toPandas()

plot_dir = OUTPUT_DIRS["plots"]

# -- VISUALIZATION WITH PLOTLY --

# --- Plot 1: Top 10 Products by Revenue ---
top_sorted = top_pd.sort_values("total_revenue", ascending=True)

fig1 = go.Figure()
fig1.add_trace(go.Bar(
    y=[n[:28] for n in top_sorted["product_name"]],
    x=top_sorted["total_revenue"] / 1e6,
    orientation="h",
    marker=dict(
        color=top_sorted["total_revenue"],
        colorscale="Greens",
        line=dict(color="white", width=1)
    ),
    text=[f"${v:.1f}M" for v in top_sorted["total_revenue"] / 1e6],
    textposition="outside",
    textfont=dict(size=10, color="#2c3e50"),
    hovertemplate=(
        "<b>%{y}</b><br>"
        "Revenue: $%{x:.1f}M<br>"
        "Category: %{customdata[0]}<br>"
        "Units Sold: %{customdata[1]:,}<br>"
        "Avg Price: $%{customdata[2]:,.2f}<br>"
        "Avg Review: %{customdata[3]}<br>"
        "<extra></extra>"
    ),
    customdata=top_sorted[["category", "total_units_sold", "avg_sale_price", "avg_review_score"]].values
))
fig1.update_layout(
    title=dict(text="Top 10 Products by Revenue", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Revenue (Millions)", gridcolor="rgba(0,0,0,0.06)", zeroline=False),
    yaxis=dict(title=""),
    plot_bgcolor="white",
    height=500,
    margin=dict(l=20, r=80, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig1.show()
fig1.write_image(os.path.join(plot_dir, "6_4_top10_products.png"), scale=2)
print("  Saved: 6_4_top10_products.png")

# --- Plot 2: Bottom 10 Products (Underperformers) ---
bot_sorted = bot_pd.sort_values("total_revenue", ascending=True)

fig2 = go.Figure()
fig2.add_trace(go.Bar(
    y=[n[:28] for n in bot_sorted["product_name"]],
    x=bot_sorted["total_revenue"],
    orientation="h",
    marker=dict(
        color=bot_sorted["total_revenue"],
        colorscale="Reds_r",
        line=dict(color="white", width=1)
    ),
    text=[f"${v:,.0f}" for v in bot_sorted["total_revenue"]],
    textposition="outside",
    textfont=dict(size=10, color="#2c3e50"),
    hovertemplate=(
        "<b>%{y}</b><br>"
        "Revenue: $%{x:,.0f}<br>"
        "Category: %{customdata[0]}<br>"
        "Times Ordered: %{customdata[1]}<br>"
        "Avg Review: %{customdata[2]}<br>"
        "<extra></extra>"
    ),
    customdata=bot_sorted[["category", "times_ordered", "avg_review_score"]].values
))
fig2.update_layout(
    title=dict(text="Bottom 10 Products - Underperformers", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Total Revenue ($)", gridcolor="rgba(0,0,0,0.06)", zeroline=False),
    yaxis=dict(title=""),
    plot_bgcolor="white",
    height=500,
    margin=dict(l=20, r=80, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig2.show()
fig2.write_image(os.path.join(plot_dir, "6_4_bottom10_products.png"), scale=2)
print("  Saved: 6_4_bottom10_products.png")

# --- Plot 3: Top 15 Suppliers by Revenue ---
sup_sorted = sup_pd.sort_values("total_revenue", ascending=True)

fig3 = go.Figure()
fig3.add_trace(go.Bar(
    y=[f"{n[:22]} ({c})" for n, c in zip(sup_sorted["supplier_name"], sup_sorted["country"])],
    x=sup_sorted["total_revenue"] / 1e6,
    orientation="h",
    marker=dict(
        color=sup_sorted["total_revenue"],
        colorscale="Blues",
        line=dict(color="white", width=1)
    ),
    text=[f"${v:.0f}M" for v in sup_sorted["total_revenue"] / 1e6],
    textposition="outside",
    textfont=dict(size=9, color="#2c3e50"),
    hovertemplate=(
        "<b>%{y}</b><br>"
        "Revenue: $%{x:.0f}M<br>"
        "Products: %{customdata[0]}<br>"
        "Orders: %{customdata[1]:,}<br>"
        "Avg Review: %{customdata[2]}<br>"
        "Avg Discount: %{customdata[3]}%%<br>"
        "<extra></extra>"
    ),
    customdata=sup_sorted[["products_count", "total_orders", "avg_review_score", "avg_discount_pct"]].values
))
fig3.update_layout(
    title=dict(text="Top 15 Suppliers by Revenue", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Revenue (Millions)", gridcolor="rgba(0,0,0,0.06)", zeroline=False),
    yaxis=dict(title=""),
    plot_bgcolor="white",
    height=550,
    margin=dict(l=20, r=80, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig3.show()
fig3.write_image(os.path.join(plot_dir, "6_4_top15_suppliers.png"), scale=2)
print("  Saved: 6_4_top15_suppliers.png")

# --- Plot 4: Price vs Review Score by Category (Bubble) ---
avg_review_line = price_pd["avg_review"].mean()

fig4 = go.Figure()
fig4.add_trace(go.Scatter(
    x=price_pd["avg_price"],
    y=price_pd["avg_review"],
    mode="markers+text",
    text=price_pd["category"],
    textposition="top center",
    textfont=dict(size=10, color="#2c3e50"),
    marker=dict(
        size=price_pd["total_sales"] / price_pd["total_sales"].max() * 60 + 15,
        color=price_pd["total_revenue"],
        colorscale="RdYlGn",
        showscale=True,
        colorbar=dict(title="Total Revenue", thickness=15),
        line=dict(color="white", width=1.5),
        opacity=0.85
    ),
    hovertemplate=(
        "<b>%{text}</b><br>"
        "Avg Price: $%{x:,.2f}<br>"
        "Avg Review: %{y:.2f}<br>"
        "Total Sales: %{customdata[0]:,}<br>"
        "Revenue: $%{customdata[1]:,.0f}<br>"
        "<extra></extra>"
    ),
    customdata=price_pd[["total_sales", "total_revenue"]].values
))
# Average review line
fig4.add_hline(
    y=avg_review_line,
    line_dash="dash",
    line_color="#1565C0",
    opacity=0.5,
    annotation_text=f"Avg Review: {avg_review_line:.2f}",
    annotation_position="top left",
    annotation_font=dict(size=10, color="#1565C0")
)
fig4.update_layout(
    title=dict(
        text="Price vs Review Score by Category (bubble size = sales count)",
        font=dict(size=16, color="#2c3e50")
    ),
    xaxis=dict(title="Average Sale Price ($)", gridcolor="rgba(0,0,0,0.06)"),
    yaxis=dict(title="Average Review Score", gridcolor="rgba(0,0,0,0.06)"),
    plot_bgcolor="white",
    height=500,
    margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig4.show()
fig4.write_image(os.path.join(plot_dir, "6_4_price_vs_review.png"), scale=2)
print("  Saved: 6_4_price_vs_review.png")

# -- Business insights --
print(f"\n  BUSINESS INSIGHTS:")
print(f"  {'-'*55}")
top_p  = top_pd.iloc[0]
top_s  = sup_pd.iloc[0]
print(f"  Best product     : {top_p['product_name'][:40]}")
print(f"  Product revenue  : ${top_p['total_revenue']:,.0f}")
print(f"  Best supplier    : {top_s['supplier_name']}")
print(f"  Supplier revenue : ${top_s['total_revenue']:,.0f}")
print(f"\n  Top 10 products:")
top_products.show(truncate=True)
print("=" * 60)
print("  Cell complete")
print("=" * 60)
```

#### 6.5 Order Analytics & Payment Analysis

```python
#
# Order Analytics & Payment Analysis
# Purpose: Analyse order status patterns
#          Understand payment method preferences
#          Discount impact on revenue
#          Business insight: Operations & payment optimisation
#

import plotly.graph_objects as go
import plotly.express as px
import numpy as np
import pandas as pd


print(f"{'ORDER ANALYTICS & PAYMENT ANALYSIS':^60}")


# -- Query 1: Order status distribution --
order_status = spark.sql("""
    SELECT
        o.order_status,
        COUNT(DISTINCT f.order_sk)      AS total_orders,
        ROUND(SUM(f.line_total), 2)     AS total_revenue,
        ROUND(AVG(f.line_total), 2)     AS avg_order_value,
        ROUND(AVG(f.discount_pct)*100,2) AS avg_discount_pct
    FROM FACT_SALES f
    JOIN DIM_ORDER o ON f.order_sk = o.order_sk
    GROUP BY o.order_status
    ORDER BY total_orders DESC
""")

# -- Query 2: Payment method analysis --
payment_method = spark.sql("""
    SELECT
        o.payment_method,
        COUNT(DISTINCT f.order_sk)      AS total_orders,
        ROUND(SUM(f.line_total), 2)     AS total_revenue,
        ROUND(AVG(f.line_total), 2)     AS avg_order_value,
        ROUND(AVG(f.discount_pct)*100,2) AS avg_discount_pct
    FROM FACT_SALES f
    JOIN DIM_ORDER o ON f.order_sk = o.order_sk
    GROUP BY o.payment_method
    ORDER BY total_revenue DESC
""")

# -- Query 3: Discount impact analysis --
discount_impact = spark.sql("""
    SELECT
        CASE
            WHEN f.discount_pct = 0         THEN '0% (No Discount)'
            WHEN f.discount_pct <= 0.10     THEN '1-10%'
            WHEN f.discount_pct <= 0.20     THEN '11-20%'
            WHEN f.discount_pct <= 0.30     THEN '21-30%'
            ELSE '30%+'
        END AS discount_range,
        COUNT(f.sales_sk)               AS total_sales,
        ROUND(SUM(f.line_total), 2)     AS total_revenue,
        ROUND(AVG(f.line_total), 2)     AS avg_order_value,
        ROUND(AVG(f.review_score), 2)   AS avg_review_score
    FROM FACT_SALES f
    GROUP BY discount_range
    ORDER BY total_revenue DESC
""")

# -- Query 4: Shipping country performance --
shipping_country = spark.sql("""
    SELECT
        o.shipping_country,
        COUNT(DISTINCT f.order_sk)      AS total_orders,
        ROUND(SUM(f.line_total), 2)     AS total_revenue,
        ROUND(AVG(f.line_total), 2)     AS avg_order_value
    FROM FACT_SALES f
    JOIN DIM_ORDER o ON f.order_sk = o.order_sk
    GROUP BY o.shipping_country
    ORDER BY total_revenue DESC
    LIMIT 10
""")

status_pd   = order_status.toPandas()
payment_pd  = payment_method.toPandas()
discount_pd = discount_impact.toPandas()
shipping_pd = shipping_country.toPandas()

plot_dir = OUTPUT_DIRS["plots"]

# -- VISUALIZATION WITH PLOTLY --

# --- Plot 1: Order Status Distribution (Horizontal Stacked Bar) ---
status_colors = {
    "Delivered" : "#4CAF50",
    "Shipped"   : "#2196F3",
    "Processing": "#FF9800",
    "Cancelled" : "#F44336",
    "Returned"  : "#9C27B0",
}
status_pd["color"] = status_pd["order_status"].map(status_colors).fillna("#607D8B")
status_pd["pct"] = (status_pd["total_orders"] / status_pd["total_orders"].sum() * 100).round(1)

fig1 = go.Figure()
fig1.add_trace(go.Bar(
    x=status_pd["order_status"],
    y=status_pd["total_orders"],
    marker=dict(
        color=status_pd["color"].tolist(),
        line=dict(color="white", width=1)
    ),
    text=[f"{o:,} orders<br>{p}%" for o, p in zip(status_pd["total_orders"], status_pd["pct"])],
    textposition="outside",
    textfont=dict(size=10, color="#2c3e50"),
    hovertemplate=(
        "<b>%{x}</b><br>"
        "Orders: %{y:,}<br>"
        "Revenue: $%{customdata[0]:,.0f}<br>"
        "Avg Order: $%{customdata[1]:,.2f}<br>"
        "Avg Discount: %{customdata[2]}%<br>"
        "<extra></extra>"
    ),
    customdata=status_pd[["total_revenue", "avg_order_value", "avg_discount_pct"]].values
))
fig1.update_layout(
    title=dict(text="Order Status Distribution", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title=""),
    yaxis=dict(title="Number of Orders", gridcolor="rgba(0,0,0,0.06)"),
    plot_bgcolor="white",
    height=450,
    margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12),
    bargap=0.3
)
fig1.show()
fig1.write_image(os.path.join(plot_dir, "6_5_order_status.png"), scale=2)
print("  Saved: 6_5_order_status.png")

# --- Plot 2: Payment Method - Revenue & Avg Order Value ---
fig2 = go.Figure()
fig2.add_trace(go.Bar(
    x=payment_pd["payment_method"],
    y=payment_pd["total_revenue"] / 1e9,
    name="Total Revenue ($B)",
    marker=dict(color="#1565C0", line=dict(color="white", width=1)),
    text=[f"${v:.2f}B" for v in payment_pd["total_revenue"] / 1e9],
    textposition="outside",
    textfont=dict(size=10),
    hovertemplate=(
        "<b>%{x}</b><br>"
        "Revenue: $%{y:.2f}B<br>"
        "Orders: %{customdata[0]:,}<br>"
        "<extra></extra>"
    ),
    customdata=payment_pd[["total_orders"]].values,
    yaxis="y"
))
fig2.add_trace(go.Bar(
    x=payment_pd["payment_method"],
    y=payment_pd["avg_order_value"] / 1e3,
    name="Avg Order Value ($K)",
    marker=dict(color="#E65100", line=dict(color="white", width=1)),
    text=[f"${v:.1f}K" for v in payment_pd["avg_order_value"] / 1e3],
    textposition="outside",
    textfont=dict(size=10),
    hovertemplate=(
        "<b>%{x}</b><br>"
        "Avg Order: $%{y:.1f}K<br>"
        "Avg Discount: %{customdata[0]}%<br>"
        "<extra></extra>"
    ),
    customdata=payment_pd[["avg_discount_pct"]].values,
    yaxis="y2"
))
fig2.update_layout(
    title=dict(text="Payment Method - Revenue & Avg Order Value", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title=""),
    yaxis=dict(
        title="Total Revenue (Billions)",
        title_font=dict(color="#1565C0"),
        tickfont=dict(color="#1565C0"),
        gridcolor="rgba(0,0,0,0.06)",
        tickprefix="$", ticksuffix="B",
        side="left"
    ),
    yaxis2=dict(
        title="Avg Order Value (Thousands)",
        title_font=dict(color="#E65100"),
        tickfont=dict(color="#E65100"),
        tickprefix="$", ticksuffix="K",
        overlaying="y",
        side="right"
    ),
    barmode="group",
    plot_bgcolor="white",
    height=470,
    margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
    bargap=0.3
)
fig2.show()
fig2.write_image(os.path.join(plot_dir, "6_5_payment_method.png"), scale=2)
print("  Saved: 6_5_payment_method.png")

# --- Plot 3: Discount Range Impact on Revenue & Reviews ---
discount_order = ["0% (No Discount)", "1-10%", "11-20%", "21-30%", "30%+"]
discount_pd["discount_range"] = pd.Categorical(
    discount_pd["discount_range"],
    categories=discount_order, ordered=True
)
discount_pd = discount_pd.sort_values("discount_range")

bar_colors = ["#1B5E20", "#43A047", "#FDD835", "#FB8C00", "#E53935"]

fig3 = go.Figure()
fig3.add_trace(go.Bar(
    x=discount_pd["discount_range"].astype(str),
    y=discount_pd["total_revenue"] / 1e9,
    name="Total Revenue ($B)",
    marker=dict(
        color=bar_colors[:len(discount_pd)],
        line=dict(color="white", width=1)
    ),
    text=[f"${v:.1f}B" for v in discount_pd["total_revenue"] / 1e9],
    textposition="outside",
    textfont=dict(size=10, color="#2c3e50"),
    hovertemplate=(
        "<b>%{x}</b><br>"
        "Revenue: $%{y:.2f}B<br>"
        "Sales: %{customdata[0]:,}<br>"
        "Avg Order: $%{customdata[1]:,.2f}<br>"
        "<extra></extra>"
    ),
    customdata=discount_pd[["total_sales", "avg_order_value"]].values,
    yaxis="y"
))
fig3.add_trace(go.Scatter(
    x=discount_pd["discount_range"].astype(str),
    y=discount_pd["avg_review_score"],
    name="Avg Review Score",
    mode="lines+markers",
    line=dict(color="#7B1FA2", width=2.5),
    marker=dict(size=10, symbol="diamond", color="#7B1FA2", line=dict(color="white", width=1.5)),
    hovertemplate=(
        "<b>%{x}</b><br>"
        "Avg Review: %{y:.2f}<br>"
        "<extra></extra>"
    ),
    yaxis="y2"
))
fig3.update_layout(
    title=dict(text="Discount Range Impact on Revenue & Reviews", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Discount Range"),
    yaxis=dict(
        title="Total Revenue (Billions)",
        gridcolor="rgba(0,0,0,0.06)",
        tickprefix="$", ticksuffix="B",
        side="left"
    ),
    yaxis2=dict(
        title="Avg Review Score",
        title_font=dict(color="#7B1FA2"),
        tickfont=dict(color="#7B1FA2"),
        overlaying="y",
        side="right",
        range=[0, 5]
    ),
    plot_bgcolor="white",
    height=470,
    margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
    bargap=0.35
)
fig3.show()
fig3.write_image(os.path.join(plot_dir, "6_5_discount_impact.png"), scale=2)
print("  Saved: 6_5_discount_impact.png")

# --- Plot 4: Top 10 Shipping Countries by Revenue ---
ship_sorted = shipping_pd.sort_values("total_revenue", ascending=True)

fig4 = go.Figure()
fig4.add_trace(go.Bar(
    y=ship_sorted["shipping_country"],
    x=ship_sorted["total_revenue"] / 1e6,
    orientation="h",
    marker=dict(
        color=ship_sorted["total_revenue"],
        colorscale="Purples",
        line=dict(color="white", width=1)
    ),
    text=[f"${v:.0f}M  |  {o:,} orders" for v, o in zip(ship_sorted["total_revenue"] / 1e6, ship_sorted["total_orders"])],
    textposition="outside",
    textfont=dict(size=10, color="#2c3e50"),
    hovertemplate=(
        "<b>%{y}</b><br>"
        "Revenue: $%{x:.0f}M<br>"
        "Orders: %{customdata[0]:,}<br>"
        "Avg Order: $%{customdata[1]:,.2f}<br>"
        "<extra></extra>"
    ),
    customdata=ship_sorted[["total_orders", "avg_order_value"]].values
))
fig4.update_layout(
    title=dict(text="Top 10 Shipping Countries by Revenue", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Revenue (Millions)", gridcolor="rgba(0,0,0,0.06)", zeroline=False),
    yaxis=dict(title=""),
    plot_bgcolor="white",
    height=500,
    margin=dict(l=20, r=140, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig4.show()
fig4.write_image(os.path.join(plot_dir, "6_5_shipping_countries.png"), scale=2)
print("  Saved: 6_5_shipping_countries.png")

# -- Business insights --
print(f"\n  BUSINESS INSIGHTS:")
print(f"  {'-'*55}")
top_pay    = payment_pd.iloc[0]
cancelled  = status_pd[status_pd["order_status"] == "Cancelled"]
print(f"  Top payment method : {top_pay['payment_method']}")
print(f"  Payment revenue    : ${top_pay['total_revenue']:,.0f}")
if not cancelled.empty:
    print(f"  Cancelled orders   : {int(cancelled.iloc[0]['total_orders']):,}")
print(f"\n  Payment method summary:")
payment_method.show(truncate=False)
print("=" * 60)
print("  Cell complete")
print("=" * 60)
```

#### 6.6 Review Score & Satisfaction Analysis

```python
#
# Review Score & Satisfaction Analysis
# Purpose: Understand customer satisfaction patterns
#          Correlate satisfaction with revenue and segments
#          Business insight: Product quality improvements
#

import plotly.graph_objects as go
import plotly.express as px
import numpy as np


print(f"{'REVIEW & SATISFACTION ANALYSIS':^60}")


# -- Query 1: Review score distribution --
score_dist = spark.sql("""
    SELECT
        f.review_score,
        COUNT(*)                        AS count,
        ROUND(COUNT(*) * 100.0 /
              SUM(COUNT(*)) OVER(), 2)  AS percentage,
        ROUND(AVG(f.line_total), 2)     AS avg_revenue
    FROM FACT_SALES f
    WHERE f.review_score > 0
    GROUP BY f.review_score
    ORDER BY f.review_score
""")

# -- Query 2: Review score by category --
score_by_cat = spark.sql("""
    SELECT
        p.category,
        ROUND(AVG(CASE WHEN f.review_score > 0
              THEN f.review_score END), 2) AS avg_score,
        COUNT(CASE WHEN f.review_score > 0
              THEN 1 END)                  AS reviewed_count,
        COUNT(CASE WHEN f.review_score = 0
              THEN 1 END)                  AS unreviewed_count,
        ROUND(SUM(f.line_total), 2)        AS total_revenue
    FROM FACT_SALES f
    JOIN DIM_PRODUCT p ON f.product_sk = p.product_sk
    GROUP BY p.category
    ORDER BY avg_score DESC
""")

# -- Query 3: Score by customer segment --
score_by_seg = spark.sql("""
    SELECT
        c.customer_segment,
        ROUND(AVG(CASE WHEN f.review_score > 0
              THEN f.review_score END), 3) AS avg_score,
        COUNT(CASE WHEN f.review_score = 5
              THEN 1 END)                  AS five_star_count,
        COUNT(CASE WHEN f.review_score = 1
              THEN 1 END)                  AS one_star_count,
        COUNT(*)                           AS total_reviews
    FROM FACT_SALES f
    JOIN DIM_CUSTOMER c ON f.customer_sk = c.customer_sk
    WHERE f.review_score > 0
    GROUP BY c.customer_segment
    ORDER BY avg_score DESC
""")

# -- Query 4: Monthly avg review trend --
monthly_score = spark.sql("""
    SELECT
        d.year,
        d.month_number,
        CONCAT(d.year, '-',
               LPAD(d.month_number, 2, '0')) AS year_month,
        ROUND(AVG(CASE WHEN f.review_score > 0
              THEN f.review_score END), 3)   AS avg_score,
        COUNT(CASE WHEN f.review_score > 0
              THEN 1 END)                    AS review_count
    FROM FACT_SALES f
    JOIN DIM_DATE d ON f.date_sk = d.date_sk
    GROUP BY d.year, d.month_number, year_month
    ORDER BY d.year, d.month_number
""")

score_pd    = score_dist.toPandas()
cat_pd      = score_by_cat.toPandas()
seg_pd      = score_by_seg.toPandas()
monthly_pd  = monthly_score.toPandas()

plot_dir = OUTPUT_DIRS["plots"]

# -- VISUALIZATION WITH PLOTLY --

# --- Plot 1: Review Score Distribution (Funnel Chart) ---
score_labels = {1: "1 - Very Poor", 2: "2 - Poor", 3: "3 - Neutral", 4: "4 - Good", 5: "5 - Excellent"}
score_colors = {1: "#E53935", 2: "#FB8C00", 3: "#FDD835", 4: "#7CB342", 5: "#1E88E5"}

fig1 = go.Figure()
fig1.add_trace(go.Funnel(
    y=[score_labels.get(int(s), str(s)) for s in score_pd["review_score"]],
    x=score_pd["count"],
    textinfo="value+percent total",
    textposition="inside",
    textfont=dict(size=12, color="white"),
    marker=dict(
        color=[score_colors.get(int(s), "#90A4AE") for s in score_pd["review_score"]],
        line=dict(color="white", width=2)
    ),
    hovertemplate=(
        "<b>%{y}</b><br>"
        "Count: %{x:,}<br>"
        "Avg Revenue: $%{customdata[0]:,.2f}<br>"
        "<extra></extra>"
    ),
    customdata=score_pd[["avg_revenue"]].values,
    connector=dict(line=dict(color="rgba(0,0,0,0.1)", width=1))
))
fig1.update_layout(
    title=dict(text="Review Score Distribution (Scores 1-5)", font=dict(size=16, color="#2c3e50")),
    plot_bgcolor="white",
    height=450,
    margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12),
    funnelmode="stack"
)
fig1.show()
fig1.write_image(os.path.join(plot_dir, "6_6_score_distribution.png"), scale=2)
print("  Saved: 6_6_score_distribution.png")

# --- Plot 2: Avg Review Score by Category (Lollipop / Dot Plot) ---
cat_sorted = cat_pd.sort_values("avg_score", ascending=True)
overall_avg = cat_sorted["avg_score"].mean()

fig2 = go.Figure()
# Stem lines
for i, row in cat_sorted.iterrows():
    fig2.add_trace(go.Scatter(
        x=[0, row["avg_score"]],
        y=[row["category"], row["category"]],
        mode="lines",
        line=dict(color="rgba(0,0,0,0.15)", width=6),
        showlegend=False,
        hoverinfo="skip"
    ))
# Dots
fig2.add_trace(go.Scatter(
    x=cat_sorted["avg_score"],
    y=cat_sorted["category"],
    mode="markers+text",
    marker=dict(
        size=16,
        color=cat_sorted["avg_score"],
        colorscale="RdYlGn",
        cmin=1, cmax=5,
        line=dict(color="white", width=2),
        showscale=True,
        colorbar=dict(title="Score", thickness=12, len=0.6)
    ),
    text=[f"{v:.2f}" for v in cat_sorted["avg_score"]],
    textposition="middle right",
    textfont=dict(size=11, color="#2c3e50"),
    hovertemplate=(
        "<b>%{y}</b><br>"
        "Avg Score: %{x:.2f}<br>"
        "Reviewed: %{customdata[0]:,}<br>"
        "Unreviewed: %{customdata[1]:,}<br>"
        "Revenue: $%{customdata[2]:,.0f}<br>"
        "<extra></extra>"
    ),
    customdata=cat_sorted[["reviewed_count", "unreviewed_count", "total_revenue"]].values,
    name="Avg Score"
))
# Reference lines
fig2.add_vline(x=3.0, line_dash="dash", line_color="#E53935", opacity=0.5,
               annotation_text="Neutral (3.0)", annotation_position="top left",
               annotation_font=dict(size=9, color="#E53935"))
fig2.add_vline(x=overall_avg, line_dash="dash", line_color="#1565C0", opacity=0.5,
               annotation_text=f"Avg ({overall_avg:.2f})", annotation_position="top right",
               annotation_font=dict(size=9, color="#1565C0"))
fig2.update_layout(
    title=dict(text="Average Review Score by Category", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Average Review Score", gridcolor="rgba(0,0,0,0.06)", range=[0, 5.5]),
    yaxis=dict(title=""),
    plot_bgcolor="white",
    height=500,
    margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12),
    showlegend=False
)
fig2.show()
fig2.write_image(os.path.join(plot_dir, "6_6_score_by_category.png"), scale=2)
print("  Saved: 6_6_score_by_category.png")

# --- Plot 3: 5-Star vs 1-Star Reviews by Segment (Diverging Bar) ---
seg_pd["five_star_pct"] = (seg_pd["five_star_count"] / seg_pd["total_reviews"] * 100).round(1)
seg_pd["one_star_pct"] = (seg_pd["one_star_count"] / seg_pd["total_reviews"] * 100).round(1)

fig3 = go.Figure()
fig3.add_trace(go.Bar(
    y=seg_pd["customer_segment"],
    x=seg_pd["five_star_count"],
    orientation="h",
    name="5-Star Reviews",
    marker=dict(color="#1E88E5", line=dict(color="white", width=1)),
    text=[f"{c:,} ({p}%)" for c, p in zip(seg_pd["five_star_count"], seg_pd["five_star_pct"])],
    textposition="outside",
    textfont=dict(size=10),
    hovertemplate=(
        "<b>%{y}</b><br>"
        "5-Star: %{x:,}<br>"
        "Avg Score: %{customdata[0]:.3f}<br>"
        "<extra></extra>"
    ),
    customdata=seg_pd[["avg_score"]].values
))
fig3.add_trace(go.Bar(
    y=seg_pd["customer_segment"],
    x=-seg_pd["one_star_count"],
    orientation="h",
    name="1-Star Reviews",
    marker=dict(color="#E53935", line=dict(color="white", width=1)),
    text=[f"{c:,} ({p}%)" for c, p in zip(seg_pd["one_star_count"], seg_pd["one_star_pct"])],
    textposition="outside",
    textfont=dict(size=10),
    hovertemplate=(
        "<b>%{y}</b><br>"
        "1-Star: %{customdata[0]:,}<br>"
        "<extra></extra>"
    ),
    customdata=seg_pd[["one_star_count"]].values
))
fig3.update_layout(
    title=dict(text="5-Star vs 1-Star Reviews by Customer Segment", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(
        title="Review Count (1-Star <-- | --> 5-Star)",
        gridcolor="rgba(0,0,0,0.06)",
        zeroline=True,
        zerolinecolor="rgba(0,0,0,0.3)",
        zerolinewidth=1.5
    ),
    yaxis=dict(title=""),
    barmode="overlay",
    plot_bgcolor="white",
    height=420,
    margin=dict(l=20, r=80, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1)
)
fig3.show()
fig3.write_image(os.path.join(plot_dir, "6_6_segment_star_comparison.png"), scale=2)
print("  Saved: 6_6_segment_star_comparison.png")

# --- Plot 4: Monthly Average Review Score Trend (Area + Markers) ---
avg_overall = monthly_pd["avg_score"].mean()

fig4 = go.Figure()
fig4.add_trace(go.Scatter(
    x=monthly_pd["year_month"],
    y=monthly_pd["avg_score"],
    mode="lines+markers",
    fill="tozeroy",
    fillcolor="rgba(123,31,250,0.1)",
    line=dict(color="#7B1FA2", width=2.5),
    marker=dict(size=5, color="#7B1FA2", line=dict(color="white", width=1)),
    hovertemplate=(
        "<b>%{x}</b><br>"
        "Avg Score: %{y:.3f}<br>"
        "Reviews: %{customdata[0]:,}<br>"
        "<extra></extra>"
    ),
    customdata=monthly_pd[["review_count"]].values,
    name="Monthly Avg Score"
))
fig4.add_hline(y=3.0, line_dash="dash", line_color="#E53935", opacity=0.5,
               annotation_text="Neutral (3.0)", annotation_position="top left",
               annotation_font=dict(size=9, color="#E53935"))
fig4.add_hline(y=avg_overall, line_dash="dash", line_color="#1565C0", opacity=0.5,
               annotation_text=f"Overall Avg ({avg_overall:.2f})", annotation_position="top right",
               annotation_font=dict(size=9, color="#1565C0"))
fig4.update_layout(
    title=dict(text="Monthly Average Review Score Trend", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Month", gridcolor="rgba(0,0,0,0.06)", tickangle=-45, dtick=3),
    yaxis=dict(title="Average Review Score", gridcolor="rgba(0,0,0,0.06)", range=[0, 5.5]),
    plot_bgcolor="white",
    height=470,
    margin=dict(l=20, r=20, t=60, b=60),
    font=dict(family="Segoe UI, sans-serif", size=12),
    showlegend=False
)
fig4.show()
fig4.write_image(os.path.join(plot_dir, "6_6_monthly_review_trend.png"), scale=2)
print("  Saved: 6_6_monthly_review_trend.png")

# -- Business insights --
print(f"\n  BUSINESS INSIGHTS:")
print(f"  {'-'*55}")
five_star = score_pd[score_pd["review_score"] == 5]
one_star  = score_pd[score_pd["review_score"] == 1]
if not five_star.empty:
    print(f"  5-star reviews  : "
          f"{int(five_star['count'].values[0]):,} "
          f"({float(five_star['percentage'].values[0]):.1f}%)")
if not one_star.empty:
    print(f"  1-star reviews  : "
          f"{int(one_star['count'].values[0]):,} "
          f"({float(one_star['percentage'].values[0]):.1f}%)")
print(f"  Overall avg score: {monthly_pd['avg_score'].mean():.3f}")
print(f"\n  Score by category:")
score_by_cat.show(truncate=False)
print("=" * 60)
print("  Cell complete")
print("=" * 60)
```

## Section 7 — Customer Clustering (K-Means)

#### 7.1 RFM Feature Engineering

```python
# 
# RFM Feature Engineering
# Purpose: Build Recency, Frequency, Monetary features
#          per customer from FACT_SALES + DIM_DATE
#          RFM is the gold standard for customer segmentation
#
# RFM Definition:
#   Recency  — How recently did the customer purchase?
#              (days since last order — lower = better)
#   Frequency — How often does the customer purchase?
#              (total number of orders placed)
#   Monetary  — How much has the customer spent?
#              (total revenue from customer)
# =


print(f"{'RFM FEATURE ENGINEERING':^60}")


# ── Step 1: Find reference date (max order date) ─────────────
max_date = spark.sql("""
    SELECT MAX(o.order_date) AS max_date
    FROM DIM_ORDER o
""").collect()[0]["max_date"]

print(f"\n  Reference date (max order date): {max_date}")

# ── Step 2: Compute RFM per customer ─────────────────────────
rfm_query = spark.sql(f"""
    SELECT
        f.customer_sk,
        c.customer_id,
        c.customer_name,
        c.customer_segment,
        c.country,
        c.age_group,

        -- RECENCY: days since last purchase
        DATEDIFF(DATE '{max_date}',
                 MAX(o.order_date))         AS recency_days,

        -- FREQUENCY: number of unique orders
        COUNT(DISTINCT f.order_sk)          AS frequency,

        -- MONETARY: total amount spent
        ROUND(SUM(f.line_total), 2)         AS monetary,

        -- Additional features
        ROUND(AVG(f.line_total), 2)         AS avg_order_value,
        ROUND(AVG(f.discount_pct) * 100, 2) AS avg_discount_pct,
        ROUND(AVG(CASE WHEN f.review_score > 0
              THEN f.review_score END), 2)  AS avg_review_score,
        SUM(f.quantity)                     AS total_items_bought,
        COUNT(DISTINCT f.product_sk)        AS unique_products_bought
    FROM FACT_SALES     f
    JOIN DIM_CUSTOMER   c ON f.customer_sk = c.customer_sk
    JOIN DIM_ORDER      o ON f.order_sk    = o.order_sk
    GROUP BY
        f.customer_sk, c.customer_id,
        c.customer_name, c.customer_segment,
        c.country, c.age_group
""")

rfm_df = rfm_query.cache()
rfm_count = rfm_df.count()

print(f"\n  RFM features computed for: {rfm_count:,} customers")
print(f"\n  RFM Statistics:")
rfm_df.select(
    F.round(F.avg("recency_days"),  2).alias("avg_recency_days"),
    F.round(F.avg("frequency"),     2).alias("avg_frequency"),
    F.round(F.avg("monetary"),      2).alias("avg_monetary"),
    F.round(F.min("recency_days"),  2).alias("min_recency"),
    F.round(F.max("recency_days"),  2).alias("max_recency"),
    F.round(F.min("frequency"),     2).alias("min_frequency"),
    F.round(F.max("frequency"),     2).alias("max_frequency"),
    F.round(F.min("monetary"),      2).alias("min_monetary"),
    F.round(F.max("monetary"),      2).alias("max_monetary"),
).show(truncate=False)

print(f"\n  Sample RFM rows:")
rfm_df.select(
    "customer_id", "customer_segment",
    "recency_days", "frequency",
    "monetary", "avg_order_value"
).show(5, truncate=False)

print("=" * 60)
print("  RFM features engineered successfully")
print("  Ready for K-Means clustering")
print("=" * 60)
```

#### 7.2 Feature Assembly & Scaling

```python
# 
# Feature Assembly & Scaling
# Purpose: Prepare RFM features for K-Means algorithm
#          VectorAssembler combines features into vector
#          StandardScaler normalizes scale differences
#          (monetary ranges millions, frequency ranges 1-10)
# 


print(f"{'FEATURE ASSEMBLY & SCALING':^60}")


# ── Step 1: Drop NULLs in feature columns ────────────────────
rfm_clean = rfm_df.dropna(subset=[
    "recency_days", "frequency",
    "monetary", "avg_order_value",
    "avg_discount_pct", "avg_review_score"
]).fillna({"avg_review_score": 3.0})

print(f"  Rows before NULL drop : {rfm_df.count():,}")
print(f"  Rows after NULL drop  : {rfm_clean.count():,}")

# ── Step 2: Define feature columns ───────────────────────────
# Using core RFM + additional behavioral features
feature_cols = [
    "recency_days",       # How recent
    "frequency",          # How often
    "monetary",           # How much total
    "avg_order_value",    # Average spend per order
    "avg_discount_pct",   # Discount sensitivity
    "total_items_bought", # Purchase volume
]

print(f"\n  Features selected for clustering:")
for i, f_col in enumerate(feature_cols, 1):
    print(f"  {i}. {f_col}")

# ── Step 3: Assemble feature vector ──────────────────────────
assembler = VectorAssembler(
    inputCols=feature_cols,
    outputCol="features_raw",
    handleInvalid="skip"
)

assembled_df = assembler.transform(rfm_clean)

# ── Step 4: Standardize features (Z-score scaling) ───────────
# Critical: Without scaling, monetary (millions) dominates
# recency (days) and frequency (small integers)
scaler = StandardScaler(
    inputCol="features_raw",
    outputCol="features",
    withStd=True,
    withMean=True
)

scaler_model  = scaler.fit(assembled_df)
scaled_df     = scaler_model.transform(assembled_df)
scaled_df     = scaled_df.cache()
scaled_count  = scaled_df.count()

print(f"\n  Assembly & scaling complete:")
print(f"  Total customers scaled : {scaled_count:,}")
print(f"  Feature vector size    : {len(feature_cols)}")
print(f"  Scaling method         : StandardScaler (Z-score)")
print(f"  withMean               : True (center at 0)")
print(f"  withStd                : True (unit variance)")
print(f"\n  Sample scaled features (first 3 rows):")
scaled_df.select(
    "customer_id", "features_raw", "features"
).show(3, truncate=80)

print("=" * 60)
print("  Features assembled and scaled")
print("  Ready for elbow method")
print("=" * 60)
```

#### 7.3 Elbow Method (Find Optimal K)

```python
#
# Elbow Method -- Find Optimal K
# Purpose: Determine the best number of clusters
#          Run K-Means for K=2 to K=8
#          Plot inertia (WSSSE) -- look for the elbow point
#          Silhouette score confirms optimal K
#


print(f"{'ELBOW METHOD -- FINDING OPTIMAL K':^60}")


import builtins
import plotly.graph_objects as go
from plotly.subplots import make_subplots

k_range    = range(2, 9)
inertia    = []
silhouette = []

print(f"\n  Running K-Means for K = 2 to 8...")
print(f"  {'K':<5} {'Inertia (WSSSE)':>20} {'Silhouette':>14}")
print(f"  {'-'*42}")

evaluator = ClusteringEvaluator(
    featuresCol="features",
    metricName="silhouette",
    distanceMeasure="squaredEuclidean"
)

for k in k_range:
    kmeans = KMeans(
        featuresCol="features",
        k=k,
        seed=42,
        maxIter=20,
        initMode="k-means||"
    )
    model      = kmeans.fit(scaled_df)
    wssse      = model.summary.trainingCost
    preds      = model.transform(scaled_df)
    sil_score  = evaluator.evaluate(preds)
    inertia.append(wssse)
    silhouette.append(sil_score)
    print(f"  K={k:<3} {wssse:>22,.2f} {sil_score:>14.4f}")

# -- Compute elbow point --
# Rate of change in inertia
deltas    = [inertia[i] - inertia[i+1]
             for i in range(len(inertia)-1)]
accel     = [deltas[i] - deltas[i+1]
             for i in range(len(deltas)-1)]
elbow_k   = list(k_range)[accel.index(
    builtins.max(accel)) + 1]
best_sil_k = list(k_range)[
    silhouette.index(builtins.max(silhouette))]

print(f"\n  Elbow point detected : K = {elbow_k}")
print(f"  Best silhouette score: K = {best_sil_k} "
      f"(score={builtins.max(silhouette):.4f})")

# -- VISUALIZATION WITH PLOTLY --
plot_dir = OUTPUT_DIRS["plots"]
k_list = list(k_range)

# --- Plot 1: Inertia (Elbow) Curve ---
fig1 = go.Figure()
fig1.add_trace(go.Scatter(
    x=k_list,
    y=inertia,
    mode="lines+markers",
    line=dict(color="#1565C0", width=2.5),
    marker=dict(size=10, color="white", line=dict(color="#1565C0", width=2.5)),
    fill="tozeroy",
    fillcolor="rgba(21,101,192,0.08)",
    name="Inertia (WSSSE)",
    hovertemplate="K=%{x}<br>Inertia: %{y:,.0f}<extra></extra>"
))
# Elbow point marker
fig1.add_trace(go.Scatter(
    x=[elbow_k],
    y=[inertia[elbow_k - 2]],
    mode="markers+text",
    marker=dict(color="#E53935", size=18, symbol="star", line=dict(color="white", width=2)),
    text=[f"Elbow K={elbow_k}"],
    textposition="top right",
    textfont=dict(size=12, color="#E53935"),
    name=f"Elbow at K={elbow_k}",
    showlegend=True
))
fig1.add_vline(x=elbow_k, line_dash="dash", line_color="#E53935", opacity=0.5)
fig1.update_layout(
    title=dict(text="Inertia vs K (Elbow Method)", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Number of Clusters (K)", dtick=1, gridcolor="rgba(0,0,0,0.06)"),
    yaxis=dict(title="Inertia (WSSSE)", gridcolor="rgba(0,0,0,0.06)"),
    plot_bgcolor="white",
    height=450,
    margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1)
)
fig1.show()
fig1.write_image(os.path.join(plot_dir, "7_3_elbow_inertia.png"), scale=2)
print("  Saved: 7_3_elbow_inertia.png")

# --- Plot 2: Silhouette Score ---
bar_colors = ["#4CAF50" if k == best_sil_k else "#90CAF9" for k in k_range]

fig2 = go.Figure()
fig2.add_trace(go.Bar(
    x=k_list,
    y=silhouette,
    marker=dict(color=bar_colors, line=dict(color="white", width=1)),
    text=[f"{v:.3f}" for v in silhouette],
    textposition="outside",
    textfont=dict(size=11, color="#2c3e50"),
    hovertemplate="K=%{x}<br>Silhouette: %{y:.4f}<extra></extra>"
))
fig2.add_vline(x=best_sil_k, line_dash="dash", line_color="#2e7d32", opacity=0.5,
               annotation_text=f"Best K={best_sil_k}", annotation_position="top right",
               annotation_font=dict(size=11, color="#2e7d32"))
fig2.update_layout(
    title=dict(text="Silhouette Score vs K (higher = better clusters)", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Number of Clusters (K)", dtick=1, gridcolor="rgba(0,0,0,0.06)"),
    yaxis=dict(title="Silhouette Score", gridcolor="rgba(0,0,0,0.06)"),
    plot_bgcolor="white",
    height=450,
    margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12),
    bargap=0.35
)
fig2.show()
fig2.write_image(os.path.join(plot_dir, "7_3_silhouette_scores.png"), scale=2)
print("  Saved: 7_3_silhouette_scores.png")

print(f"\n  CONCLUSION:")
print(f"  Optimal K selected : 4")
print(f"  Reason             : Elbow point K={elbow_k},")
print(f"                       4 segments is standard")
print(f"                       for customer analytics")
print("=" * 60)
print("  Cell complete -- using K=4 for final model")
print("=" * 60)
```

#### 7.4 Train Final K-Means Model

```python
# 
# Train Final K-Means Model (K=4) - FIXED VERSION
# Purpose: Train the final customer segmentation model
#          Fixed: predictionCol="cluster" added to both
#          KMeans and ClusteringEvaluator
#


print(f"{'TRAINING FINAL K-MEANS MODEL (K=4)':^60}")


OPTIMAL_K = 4

# -- Train model -----------------------------------------------
print(f"\n  Training K-Means with K={OPTIMAL_K}...")
t_start = time.time()

kmeans_final = KMeans(
    featuresCol="features",
    predictionCol="cluster",      # <- FIXED
    k=OPTIMAL_K,
    seed=42,
    maxIter=50,
    initMode="k-means||",
    tol=1e-4
)

kmeans_model  = kmeans_final.fit(scaled_df)
t_elapsed     = time.time() - t_start

# -- Assign clusters -------------------------------------------
customer_clusters = kmeans_model.transform(scaled_df)
customer_clusters = customer_clusters.cache()
cluster_count     = customer_clusters.count()

# -- Model evaluation ------------------------------------------
evaluator = ClusteringEvaluator(
    featuresCol="features",
    predictionCol="cluster",      # <- FIXED
    metricName="silhouette",
    distanceMeasure="squaredEuclidean"
)
sil_score = evaluator.evaluate(customer_clusters)
wssse     = kmeans_model.summary.trainingCost

# -- Cluster sizes ---------------------------------------------
cluster_sizes = customer_clusters \
    .groupBy("cluster") \
    .count() \
    .orderBy("cluster") \
    .collect()

print(f"\n  Training complete in {t_elapsed:.1f}s")
print(f"  Total customers clustered : {cluster_count:,}")
print(f"  Silhouette Score          : {sil_score:.4f}")
print(f"  WSSSE (Inertia)           : {wssse:,.2f}")
print(f"\n  Cluster sizes:")
print(f"  {'Cluster':<10} {'Count':>10} {'%':>8}")
print(f"  {'-'*30}")
for row in cluster_sizes:
    pct = row["count"] / cluster_count * 100
    print(f"  Cluster {row['cluster']:<3} "
          f"{row['count']:>10,} {pct:>7.1f}%")

# -- Cluster centroids -----------------------------------------
print(f"\n  Cluster centroids (scaled space):")
centroids = kmeans_model.clusterCenters()
for i, centroid in enumerate(centroids):
    print(f"  Cluster {i}: "
          f"[recency={centroid[0]:.2f}, "
          f"freq={centroid[1]:.2f}, "
          f"monetary={centroid[2]:.2f}, "
          f"aov={centroid[3]:.2f}]")

print("=" * 60)
print(f"    K-Means model trained (K={OPTIMAL_K})")
print(f"    {cluster_count:,} customers assigned to clusters")
print("=" * 60)
```

## Section 8 — Product Clustering (K-Means)

#### 8.1 Product Feature Engineering

```python
# 
# Product Feature Engineering
# Purpose: Build product performance features from
#          FACT_SALES joined with DIM_PRODUCT
#          Identify product performance groups:
#          Bestsellers, Mid-range, Slow movers, Premium
#
# Features engineered:
#   sales_volume      -- how many times ordered
#   total_revenue     -- total money generated
#   avg_sale_price    -- average price sold at
#   avg_discount_pct  -- how often discounted
#   avg_review_score  -- customer satisfaction
#   total_units_sold  -- quantity moved
#   unique_customers  -- customer reach
# 


print(f"{'PRODUCT FEATURE ENGINEERING':^60}")


# -- Compute product performance features ---------------------
product_features = spark.sql("""
    SELECT
        f.product_sk,
        p.product_id,
        p.product_name,
        p.category,
        p.subcategory,
        p.weight_kg,

        -- Sales volume
        COUNT(f.sales_sk)                       AS sales_volume,

        -- Revenue
        ROUND(SUM(f.line_total), 2)             AS total_revenue,
        ROUND(AVG(f.line_total), 2)             AS avg_revenue_per_sale,

        -- Pricing
        ROUND(AVG(f.unit_price_at_sale), 2)     AS avg_sale_price,
        ROUND(AVG(f.unit_price), 2)             AS avg_list_price,

        -- Discount behaviour
        ROUND(AVG(f.discount_pct) * 100, 2)    AS avg_discount_pct,

        -- Review score (exclude 0 = unreviewed)
        ROUND(AVG(CASE WHEN f.review_score > 0
              THEN f.review_score END), 2)      AS avg_review_score,
        COUNT(CASE WHEN f.review_score > 0
              THEN 1 END)                       AS review_count,

        -- Volume metrics
        SUM(f.quantity)                         AS total_units_sold,
        COUNT(DISTINCT f.customer_sk)           AS unique_customers,
        COUNT(DISTINCT f.order_sk)              AS unique_orders

    FROM FACT_SALES   f
    JOIN DIM_PRODUCT  p ON f.product_sk = p.product_sk
    GROUP BY
        f.product_sk, p.product_id, p.product_name,
        p.category, p.subcategory, p.weight_kg
    ORDER BY total_revenue DESC
""")

product_features_df = product_features.cache()
prod_count = product_features_df.count()

print(f"\n  Product features computed: {prod_count:,} products")
print(f"\n  Feature Statistics:")
product_features_df.select(
    F.round(F.avg("sales_volume"),      2).alias("avg_sales_volume"),
    F.round(F.avg("total_revenue"),     2).alias("avg_total_revenue"),
    F.round(F.avg("avg_sale_price"),    2).alias("avg_sale_price"),
    F.round(F.avg("avg_review_score"),  2).alias("avg_review_score"),
    F.round(F.avg("avg_discount_pct"),  2).alias("avg_discount_pct"),
    F.round(F.avg("total_units_sold"),  2).alias("avg_units_sold"),
    F.round(F.avg("unique_customers"),  2).alias("avg_unique_customers"),
).show(truncate=False)

print(f"\n  Top 5 products by revenue:")
product_features_df.select(
    "product_name", "category",
    "total_revenue", "sales_volume",
    "avg_review_score"
).show(5, truncate=40)

print(f"\n  Bottom 5 products by revenue:")
product_features_df.orderBy("total_revenue") \
    .select(
        "product_name", "category",
        "total_revenue", "sales_volume",
        "avg_review_score"
    ).show(5, truncate=40)

print("=" * 60)
print("    Product features engineered successfully")
print("=" * 60)
```

#### 8.2 Feature Assembly, Scaling & Elbow Method

```python
# 
# Feature Assembly, Scaling & Elbow Method
# Purpose: Prepare product features for K-Means
#          Find optimal K using elbow + silhouette
#          Combined into one cell for efficiency
# 


print(f"{'ASSEMBLY, SCALING & ELBOW METHOD':^60}")


import builtins

# -- Step 1: Drop NULLs ---------------------------------------
prod_feature_cols = [
    "sales_volume",
    "total_revenue",
    "avg_sale_price",
    "avg_discount_pct",
    "avg_review_score",
    "total_units_sold",
    "unique_customers"
]

prod_clean = product_features_df \
    .dropna(subset=prod_feature_cols) \
    .fillna({"avg_review_score": 3.0})

print(f"  Products before NULL drop : {product_features_df.count():,}")
print(f"  Products after NULL drop  : {prod_clean.count():,}")
print(f"\n  Feature columns:")
for i, fc in enumerate(prod_feature_cols, 1):
    print(f"  {i}. {fc}")

# -- Step 2: Assemble + Scale ----------------------------------
prod_assembler = VectorAssembler(
    inputCols=prod_feature_cols,
    outputCol="features_raw",
    handleInvalid="skip"
)
prod_assembled = prod_assembler.transform(prod_clean)

prod_scaler = StandardScaler(
    inputCol="features_raw",
    outputCol="features",
    withStd=True,
    withMean=True
)
prod_scaler_model = prod_scaler.fit(prod_assembled)
prod_scaled       = prod_scaler_model \
    .transform(prod_assembled).cache()

print(f"\n  Products scaled: {prod_scaled.count():,}")

# -- Step 3: Elbow Method for Products ------------------------
k_range     = range(2, 8)
inertia_p   = []
silhouette_p= []

prod_evaluator = ClusteringEvaluator(
    featuresCol="features",
    predictionCol="cluster",
    metricName="silhouette",
    distanceMeasure="squaredEuclidean"
)

print(f"\n  Running K-Means for K = 2 to 7...")
print(f"  {'K':<5} {'Inertia':>18} {'Silhouette':>14}")
print(f"  {'-'*40}")

for k in k_range:
    km = KMeans(
        featuresCol="features",
        predictionCol="cluster",
        k=k, seed=42,
        maxIter=20,
        initMode="k-means||"
    )
    km_model     = km.fit(prod_scaled)
    wssse        = km_model.summary.trainingCost
    preds        = km_model.transform(prod_scaled)
    sil          = prod_evaluator.evaluate(preds)
    inertia_p.append(wssse)
    silhouette_p.append(sil)
    print(f"  K={k:<3} {wssse:>20,.2f} {sil:>14.4f}")

# -- Find optimal K --------------------------------------------
best_sil_k_p = list(k_range)[
    silhouette_p.index(builtins.max(silhouette_p))]

# Compute elbow
deltas_p = [inertia_p[i] - inertia_p[i+1]
            for i in range(len(inertia_p)-1)]
if len(deltas_p) > 1:
    accel_p  = [deltas_p[i] - deltas_p[i+1]
                for i in range(len(deltas_p)-1)]
    elbow_k_p = list(k_range)[
        accel_p.index(builtins.max(accel_p)) + 1]
else:
    elbow_k_p = best_sil_k_p

print(f"\n  Elbow point detected  : K = {elbow_k_p}")
print(f"  Best silhouette score : K = {best_sil_k_p} "
      f"(score={builtins.max(silhouette_p):.4f})")

# -- VISUALIZATION (Plotly) ------------------------------------
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import os

fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=(
        "Product Inertia vs K",
        "Product Silhouette Score vs K"
    ),
    horizontal_spacing=0.12
)

# --- Inertia (left) ---
fig.add_trace(go.Scatter(
    x=list(k_range), y=inertia_p,
    mode="lines+markers",
    line=dict(color="#388E3C", width=2.5),
    marker=dict(size=9, color="white",
                line=dict(color="#388E3C", width=2)),
    name="Inertia",
    showlegend=False
), row=1, col=1)

fig.add_vline(x=elbow_k_p, line_dash="dash",
              line_color="red", line_width=2,
              annotation_text=f"Elbow K={elbow_k_p}",
              annotation_position="top right",
              row=1, col=1)

fig.add_trace(go.Scatter(
    x=[elbow_k_p], y=[inertia_p[elbow_k_p - 2]],
    mode="markers",
    marker=dict(size=16, color="red", symbol="star"),
    name=f"Elbow K={elbow_k_p}",
    showlegend=False
), row=1, col=1)

# --- Silhouette (right) ---
colors_sil = ["#4CAF50" if k == best_sil_k_p
              else "#A5D6A7" for k in k_range]

fig.add_trace(go.Bar(
    x=list(k_range), y=silhouette_p,
    marker_color=colors_sil,
    marker_line=dict(color="white", width=0.8),
    text=[f"{v:.3f}" for v in silhouette_p],
    textposition="outside",
    textfont=dict(size=10, color="#2c3e50"),
    name="Silhouette",
    showlegend=False
), row=1, col=2)

fig.add_vline(x=best_sil_k_p, line_dash="dash",
              line_color="green", line_width=2,
              annotation_text=f"Best K={best_sil_k_p}",
              annotation_position="top right",
              row=1, col=2)

fig.update_xaxes(title_text="Number of Clusters (K)",
                 dtick=1, row=1, col=1)
fig.update_yaxes(title_text="Inertia (WSSSE)",
                 gridcolor="rgba(0,0,0,0.06)", row=1, col=1)
fig.update_xaxes(title_text="Number of Clusters (K)",
                 dtick=1, row=1, col=2)
fig.update_yaxes(title_text="Silhouette Score",
                 gridcolor="rgba(0,0,0,0.06)", row=1, col=2)

fig.update_layout(
    title=dict(text="Product K-Means -- Elbow Method",
               font=dict(size=16, color="#2c3e50")),
    plot_bgcolor="white",
    height=450, width=950,
    margin=dict(l=20, r=20, t=80, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)

plot_dir = OUTPUT_DIRS["plots"]
plot_path = os.path.join(plot_dir, "8_2_product_elbow.png")
fig.write_image(plot_path, scale=2)
fig.show()
print(f"    Plot saved: {plot_path}")
print(f"\n  Using K=3 for product clustering")
print("=" * 60)
print("    Cell complete")
print("=" * 60)
```

#### 8.3 Train Product K-Means & Analyze Clusters

```python
# 
# Train Product K-Means & Analyze Clusters
# Purpose: Train final product K-Means model with K=4
#          Profile each cluster using percentile-based
#          labeling (fixes the all-same-label issue)
#          Business labels: Bestseller, Mid-Range,
#                           Slow Mover, Premium Niche
# 


print(f"{'TRAIN & ANALYZE PRODUCT CLUSTERS':^60}")


PROD_K = 3

# -- Train model -----------------------------------------------
print(f"\n  Training product K-Means (K={PROD_K})...")
t_start = time.time()

prod_kmeans = KMeans(
    featuresCol="features",
    predictionCol="cluster",
    k=PROD_K, seed=42,
    maxIter=50,
    initMode="k-means||",
    tol=1e-4
)

prod_model   = prod_kmeans.fit(prod_scaled)
t_elapsed    = time.time() - t_start

prod_preds   = prod_model.transform(prod_scaled).cache()
prod_sil     = prod_evaluator.evaluate(prod_preds)
prod_wssse   = prod_model.summary.trainingCost

print(f"  Training complete in   : {t_elapsed:.1f}s")
print(f"  Silhouette Score       : {prod_sil:.4f}")
print(f"  WSSSE                  : {prod_wssse:,.2f}")
print(f"  Products clustered     : {prod_preds.count():,}")

# -- Compute cluster profiles ----------------------------------
prod_cluster_profile = prod_preds \
    .groupBy("cluster") \
    .agg(
        F.count("product_sk").alias("product_count"),
        F.round(F.avg("sales_volume"),      1).alias("avg_sales_volume"),
        F.round(F.avg("total_revenue"),     2).alias("avg_total_revenue"),
        F.round(F.avg("avg_sale_price"),    2).alias("avg_sale_price"),
        F.round(F.avg("avg_review_score"),  2).alias("avg_review_score"),
        F.round(F.avg("avg_discount_pct"),  2).alias("avg_discount_pct"),
        F.round(F.avg("total_units_sold"),  1).alias("avg_units_sold"),
        F.round(F.avg("unique_customers"),  1).alias("avg_unique_customers"),
        F.round(F.sum("total_revenue"),     2).alias("cluster_total_revenue")
    ) \
    .orderBy("cluster")

prod_profile_pd = prod_cluster_profile.toPandas()

# -- Percentile-based labeling ---------------------------------
# Rank clusters by avg_total_revenue + avg_sales_volume
# to assign meaningful business labels
prod_profile_pd["revenue_rank"] = prod_profile_pd[
    "avg_total_revenue"].rank(ascending=False)
prod_profile_pd["volume_rank"]  = prod_profile_pd[
    "avg_sales_volume"].rank(ascending=False)
prod_profile_pd["price_rank"]   = prod_profile_pd[
    "avg_sale_price"].rank(ascending=False)
prod_profile_pd["combined_rank"] = (
    prod_profile_pd["revenue_rank"] +
    prod_profile_pd["volume_rank"]
) / 2

def assign_product_label(row):
    rev_r   = row["revenue_rank"]
    vol_r   = row["volume_rank"]
    price_r = row["price_rank"]
    n       = PROD_K

    if rev_r == 1 and vol_r <= 2:
        return "Bestseller"
    elif price_r == 1 and rev_r <= 2:
        return "Premium Niche"
    elif rev_r >= n and vol_r >= n:
        return "Slow Mover"
    else:
        return "Mid-Range Performer"

prod_profile_pd["product_label"] = prod_profile_pd.apply(
    assign_product_label, axis=1)

# -- Print profile ---------------------------------------------
print(f"\n  PRODUCT CLUSTER PROFILE:")
print(f"  {'='*72}")
print(f"  {'Cluster':<10} {'Label':<22} {'Count':>7} "
      f"{'Avg Revenue':>14} {'Avg Price':>12} "
      f"{'Review':>8} {'Discount':>9}")
print(f"  {'-'*72}")

for _, row in prod_profile_pd.iterrows():
    print(f"  Cluster {int(row['cluster']):<3} "
          f"{row['product_label']:<22} "
          f"{int(row['product_count']):>7,} "
          f"${row['avg_total_revenue']:>13,.0f} "
          f"${row['avg_sale_price']:>11,.0f} "
          f"{row['avg_review_score']:>8.2f} "
          f"{row['avg_discount_pct']:>8.1f}%")
print(f"  {'='*72}")

# -- VISUALIZATION WITH PLOTLY --
import plotly.graph_objects as go
import plotly.express as px
plot_dir = OUTPUT_DIRS["plots"]

prod_colors = {
    "Bestseller"         : "#1E88E5",
    "Premium Niche"      : "#8E24AA",
    "Mid-Range Performer": "#43A047",
    "Slow Mover"         : "#E53935"
}
bar_colors_prod = [prod_colors.get(row, "#90A4AE") for row in prod_profile_pd["product_label"]]
labels_prod = prod_profile_pd["product_label"].tolist()
total_prods = prod_profile_pd["product_count"].sum()

# Plot 1: Product count per segment
fig1 = go.Figure()
fig1.add_trace(go.Bar(
    x=labels_prod, y=prod_profile_pd["product_count"],
    marker=dict(color=bar_colors_prod, line=dict(color="white", width=1)),
    text=[f"{v:,} ({v/total_prods*100:.1f}%)" for v in prod_profile_pd["product_count"]],
    textposition="outside", textfont=dict(size=10, color="#2c3e50"),
    hovertemplate="<b>%{x}</b><br>Products: %{y:,}<extra></extra>"
))
fig1.update_layout(
    title=dict(text="Products per Segment", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="", tickangle=-10), yaxis=dict(title="Product Count", gridcolor="rgba(0,0,0,0.06)"),
    plot_bgcolor="white", height=430, margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12), bargap=0.3
)
fig1.show()
fig1.write_image(os.path.join(plot_dir, "8_3_products_per_segment.png"), scale=2)
print("  Saved: 8_3_products_per_segment.png")

# Plot 2: Avg revenue per product segment
fig2 = go.Figure()
fig2.add_trace(go.Bar(
    x=labels_prod, y=prod_profile_pd["avg_total_revenue"] / 1e3,
    marker=dict(color=bar_colors_prod, line=dict(color="white", width=1)),
    text=[f"${v:,.0f}K" for v in prod_profile_pd["avg_total_revenue"] / 1e3],
    textposition="outside", textfont=dict(size=10, color="#2c3e50"),
    hovertemplate="<b>%{x}</b><br>Avg Revenue: $%{y:,.0f}K<extra></extra>"
))
fig2.update_layout(
    title=dict(text="Avg Revenue per Product Segment", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="", tickangle=-10), yaxis=dict(title="Avg Revenue (Thousands)", gridcolor="rgba(0,0,0,0.06)", tickprefix="$", ticksuffix="K"),
    plot_bgcolor="white", height=430, margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12), bargap=0.3
)
fig2.show()
fig2.write_image(os.path.join(plot_dir, "8_3_avg_revenue_product.png"), scale=2)
print("  Saved: 8_3_avg_revenue_product.png")

# Plot 3: Price vs Sales Volume bubble
fig3 = go.Figure()
for _, row in prod_profile_pd.iterrows():
    color = prod_colors.get(row["product_label"], "#90A4AE")
    fig3.add_trace(go.Scatter(
        x=[row["avg_sale_price"]], y=[row["avg_sales_volume"]],
        mode="markers+text",
        marker=dict(size=row["cluster_total_revenue"] / prod_profile_pd["cluster_total_revenue"].max() * 60 + 20,
                    color=color, opacity=0.85, line=dict(color="white", width=2)),
        text=[f"{row['product_label']} ({int(row['product_count'])})"],
        textposition="top center", textfont=dict(size=9, color="#2c3e50"),
        name=row["product_label"],
        hovertemplate=f"<b>{row['product_label']}</b><br>Avg Price: ${row['avg_sale_price']:,.0f}<br>Sales Volume: {row['avg_sales_volume']:.0f}<br>Products: {int(row['product_count'])}<extra></extra>"
    ))
fig3.update_layout(
    title=dict(text="Price vs Sales Volume (bubble = total revenue)", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Avg Sale Price ($)", gridcolor="rgba(0,0,0,0.06)"),
    yaxis=dict(title="Avg Sales Volume", gridcolor="rgba(0,0,0,0.06)"),
    plot_bgcolor="white", height=480, margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12), showlegend=False
)
fig3.show()
fig3.write_image(os.path.join(plot_dir, "8_3_price_vs_volume.png"), scale=2)
print("  Saved: 8_3_price_vs_volume.png")

# Plot 4: Review Score vs Discount
fig4 = go.Figure()
fig4.add_trace(go.Bar(
    x=labels_prod, y=prod_profile_pd["avg_review_score"],
    name="Avg Review Score",
    marker=dict(color=bar_colors_prod, line=dict(color="white", width=1)),
    text=[f"{v:.2f}" for v in prod_profile_pd["avg_review_score"]],
    textposition="outside", textfont=dict(size=10),
    hovertemplate="<b>%{x}</b><br>Review: %{y:.2f}<extra></extra>",
    yaxis="y"
))
fig4.add_trace(go.Scatter(
    x=labels_prod, y=prod_profile_pd["avg_discount_pct"],
    name="Avg Discount %", mode="lines+markers",
    line=dict(color="#E53935", width=2.5),
    marker=dict(size=10, symbol="diamond", color="#E53935", line=dict(color="white", width=1.5)),
    hovertemplate="<b>%{x}</b><br>Discount: %{y:.1f}%<extra></extra>",
    yaxis="y2"
))
fig4.update_layout(
    title=dict(text="Review Score vs Discount by Segment", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="", tickangle=-10),
    yaxis=dict(title="Avg Review Score", gridcolor="rgba(0,0,0,0.06)", range=[0, 6]),
    yaxis2=dict(title="Avg Discount %", title_font=dict(color="#E53935"), tickfont=dict(color="#E53935"),
                overlaying="y", side="right", range=[0, 30]),
    plot_bgcolor="white", height=430, margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1), bargap=0.3
)
fig4.show()
fig4.write_image(os.path.join(plot_dir, "8_3_review_vs_discount.png"), scale=2)
print("  Saved: 8_3_review_vs_discount.png")

# Plot 5: Revenue share (horizontal stacked bar)
total_rev_p = prod_profile_pd["cluster_total_revenue"].sum()
fig5 = go.Figure()
for i, row in prod_profile_pd.iterrows():
    color = prod_colors.get(row["product_label"], "#90A4AE")
    pct = row["cluster_total_revenue"] / total_rev_p * 100
    fig5.add_trace(go.Bar(
        y=["Revenue Share"], x=[pct], orientation="h",
        name=f"{row['product_label']} ({pct:.1f}%)",
        marker=dict(color=color, line=dict(color="white", width=1)),
        text=f"{pct:.1f}%", textposition="inside", textfont=dict(size=12, color="white"),
        hovertemplate=f"<b>{row['product_label']}</b><br>${row['cluster_total_revenue']/1e9:.2f}B ({pct:.1f}%)<extra></extra>"
    ))
fig5.update_layout(
    title=dict(text="Revenue Share by Product Segment", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Revenue Share (%)", gridcolor="rgba(0,0,0,0.06)", range=[0, 105]),
    yaxis=dict(title=""), barmode="stack",
    plot_bgcolor="white", height=250, margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12),
    legend=dict(orientation="h", yanchor="bottom", y=-0.4, xanchor="center", x=0.5)
)
fig5.show()
fig5.write_image(os.path.join(plot_dir, "8_3_product_revenue_share.png"), scale=2)
print("  Saved: 8_3_product_revenue_share.png")

# Plot 6: Category distribution per cluster
if "category" in prod_preds.columns:
    cat_cluster = prod_preds.groupBy("cluster", "category").count().toPandas()
else:
    cat_cluster = prod_preds.join(
        product_features_df.select("product_sk", "category").dropDuplicates(["product_sk"]), "product_sk"
    ).groupBy("cluster", "category").count().toPandas()

fig6 = go.Figure()
categories_list = sorted(cat_cluster["category"].unique())
import plotly.express as px
cat_colors_list = px.colors.qualitative.Set3
for i, cat in enumerate(categories_list):
    cat_data = cat_cluster[cat_cluster["category"] == cat].sort_values("cluster")
    fig6.add_trace(go.Bar(
        x=[f"C{int(c)}" for c in cat_data["cluster"]],
        y=cat_data["count"],
        name=cat[:18],
        marker=dict(color=cat_colors_list[i % len(cat_colors_list)], line=dict(color="white", width=0.5)),
        hovertemplate=f"<b>{cat}</b><br>Cluster %{{x}}<br>Count: %{{y}}<extra></extra>"
    ))
fig6.update_layout(
    title=dict(text="Category Distribution per Cluster", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Cluster"), yaxis=dict(title="Product Count", gridcolor="rgba(0,0,0,0.06)"),
    barmode="stack", plot_bgcolor="white", height=450,
    margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12),
    legend=dict(font=dict(size=9))
)
fig6.show()
fig6.write_image(os.path.join(plot_dir, "8_3_category_distribution.png"), scale=2)
print("  Saved: 8_3_category_distribution.png")

old_plot = os.path.join(OUTPUT_DIRS["plots"], "cell45_product_clusters.png")
if os.path.exists(old_plot):
    os.remove(old_plot)
    print(f"  Old plot deleted: {old_plot}")

# -- Business recommendations ----------------------------------
print(f"\n  BUSINESS RECOMMENDATIONS:")
print(f"  {'='*55}")
for _, row in prod_profile_pd.iterrows():
    label = row["product_label"]
    if label == "Bestseller":
        action = "Increase stock, protect margins, feature prominently"
    elif label == "Premium Niche":
        action = "Maintain exclusivity, target high-value customers"
    elif label == "Slow Mover":
        action = "Apply promotions, consider discontinuation"
    else:
        action = "Optimize pricing, bundle with bestsellers"
    print(f"  {label:<22} -> {action}")
print("=" * 60)
print("    Cell complete")
print("=" * 60)
```

## Section 9 — Predictive Models (Sales Focused)

#### 9.1 Feature Engineering for All 3 Models

```python
# 
# Feature Engineering for All 3 Models
# Purpose: Build features from FACT_SALES star schema
#          Prepare one enriched DataFrame used by all 3 models
#
# Model 1 -- Linear Regression:
#   Predict line_total (sales revenue)
#   Features: quantity, unit_price, discount_pct,
#             category, customer_segment, month, year
#
# Model 2 -- Random Forest Regression:
#   Predict review_score (customer satisfaction)
#   Features: category, customer_segment, discount_pct,
#             quantity, unit_price, order_status, age_group
#
# Model 3 -- Logistic Regression (Classification):
#   Predict order outcome: Delivered(1) vs Cancelled(0)
#   Features: discount_pct, quantity, unit_price,
#             shipping_country, payment_method, category
# 


print(f"{'FEATURE ENGINEERING FOR ML MODELS':^60}")


# -- Build enriched dataset from star schema -------------------
ml_base = spark.sql("""
    SELECT
        f.sales_sk,
        f.line_total,
        f.quantity,
        f.unit_price,
        f.unit_price_at_sale,
        f.discount_pct,
        f.review_score,

        p.category,
        p.subcategory,
        p.weight_kg,

        c.customer_segment,
        c.age_group,
        c.country         AS customer_country,

        o.order_status,
        o.payment_method,
        o.shipping_country,

        d.month_number,
        d.year,
        d.quarter,
        d.is_weekend

    FROM FACT_SALES   f
    JOIN DIM_PRODUCT  p ON f.product_sk   = p.product_sk
    JOIN DIM_CUSTOMER c ON f.customer_sk  = c.customer_sk
    JOIN DIM_ORDER    o ON f.order_sk     = o.order_sk
    JOIN DIM_DATE     d ON f.date_sk      = d.date_sk
""")

# -- Drop NULLs and filter valid rows -------------------------
ml_base = ml_base \
    .filter(col("line_total")   > 0) \
    .filter(col("unit_price")   > 0) \
    .filter(col("quantity")     > 0) \
    .filter(col("category").isNotNull()) \
    .filter(col("customer_segment").isNotNull()) \
    .filter(col("order_status").isNotNull()) \
    .filter(col("payment_method").isNotNull()) \
    .filter(col("shipping_country").isNotNull()) \
    .dropna()

ml_base = ml_base.cache()
ml_count = ml_base.count()

print(f"\n  Enriched ML dataset: {ml_count:,} rows")
print(f"\n  Feature overview:")
print(f"  {'Column':<22} {'Type':<15} {'Usage'}")
print(f"  {'-'*60}")
feature_info = [
    ("line_total",        "DoubleType", "Target -- Model 1"),
    ("review_score",      "IntegerType","Target -- Model 2"),
    ("order_status",      "StringType", "Target -- Model 3"),
    ("quantity",          "IntegerType","Feature -- all models"),
    ("unit_price",        "DoubleType", "Feature -- all models"),
    ("discount_pct",      "DoubleType", "Feature -- all models"),
    ("category",          "StringType", "Feature -- all models"),
    ("customer_segment",  "StringType", "Feature -- all models"),
    ("month_number",      "IntegerType","Feature -- Model 1"),
    ("year",              "IntegerType","Feature -- Model 1"),
    ("age_group",         "StringType", "Feature -- Model 2"),
    ("order_status",      "StringType", "Feature -- Model 2"),
    ("payment_method",    "StringType", "Feature -- Model 3"),
    ("shipping_country",  "StringType", "Feature -- Model 3"),
    ("is_weekend",        "BooleanType","Feature -- Model 3"),
]
for col_name, dtype, usage in feature_info:
    print(f"  {col_name:<22} {dtype:<15} {usage}")

# -- Show class distributions ----------------------------------
print(f"\n  Order status distribution (Model 3 target):")
ml_base.groupBy("order_status") \
    .count() \
    .orderBy(desc("count")) \
    .show(truncate=False)

print(f"\n  Review score distribution (Model 2 target):")
ml_base.filter(col("review_score") > 0) \
    .groupBy("review_score") \
    .count() \
    .orderBy("review_score") \
    .show()

print("=" * 60)
print(f"    ML base dataset ready: {ml_count:,} rows")
print("=" * 60)
```

#### 9.2 Model 1: Linear Regression (Predict Revenue)

```python
# 
# Model 1 -- Linear Regression
# Purpose: Predict line_total (transaction revenue)
#          Business value: Forecast future sales revenue
#          given product, discount and customer attributes
#
# Features: quantity, unit_price, discount_pct,
#           category (encoded), customer_segment (encoded),
#           month_number, year, is_weekend
# Target  : line_total (continuous -- regression)
# 


print(f"{'MODEL 1 -- LINEAR REGRESSION':^60}")


# -- Step 1: Encode categorical features ----------------------
print("\n  Step 1: Encoding categorical features...")

# StringIndexer for category
cat_indexer = StringIndexer(
    inputCol="category",
    outputCol="category_idx",
    handleInvalid="keep"
)
# StringIndexer for customer_segment
seg_indexer = StringIndexer(
    inputCol="customer_segment",
    outputCol="segment_idx",
    handleInvalid="keep"
)
# Boolean to integer for is_weekend
ml_lr = ml_base \
    .withColumn("is_weekend_int",
        col("is_weekend").cast(IntegerType()))

# -- Step 2: Build Pipeline ------------------------------------
lr_feature_cols = [
    "quantity",
    "unit_price",
    "discount_pct",
    "category_idx",
    "segment_idx",
    "month_number",
    "year",
    "is_weekend_int"
]

lr_assembler = VectorAssembler(
    inputCols=lr_feature_cols,
    outputCol="features",
    handleInvalid="skip"
)

lr_scaler = StandardScaler(
    inputCol="features",
    outputCol="scaled_features",
    withStd=True,
    withMean=True
)

lr_model_def = LinearRegression(
    featuresCol="scaled_features",
    labelCol="line_total",
    predictionCol="lr_prediction",
    maxIter=100,
    regParam=0.1,
    elasticNetParam=0.0
)

lr_pipeline = Pipeline(stages=[
    cat_indexer,
    seg_indexer,
    lr_assembler,
    lr_scaler,
    lr_model_def
])

# -- Step 3: Train/test split ----------------------------------
print("  Step 2: Splitting train/test (80/20)...")
lr_train, lr_test = ml_lr.randomSplit(
    [0.8, 0.2], seed=42)

print(f"  Training rows : {lr_train.count():,}")
print(f"  Test rows     : {lr_test.count():,}")

# -- Step 4: Train model ---------------------------------------
print("\n  Step 3: Training Linear Regression model...")
t_start  = time.time()
lr_fitted = lr_pipeline.fit(lr_train)
t_elapsed = time.time() - t_start
print(f"  Training time : {t_elapsed:.1f}s")

# -- Step 5: Predict and evaluate -----------------------------
print("\n  Step 4: Evaluating on test set...")
lr_predictions = lr_fitted.transform(lr_test)

lr_evaluator_rmse = RegressionEvaluator(
    labelCol="line_total",
    predictionCol="lr_prediction",
    metricName="rmse"
)
lr_evaluator_mae  = RegressionEvaluator(
    labelCol="line_total",
    predictionCol="lr_prediction",
    metricName="mae"
)
lr_evaluator_r2   = RegressionEvaluator(
    labelCol="line_total",
    predictionCol="lr_prediction",
    metricName="r2"
)

rmse = lr_evaluator_rmse.evaluate(lr_predictions)
mae  = lr_evaluator_mae.evaluate(lr_predictions)
r2   = lr_evaluator_r2.evaluate(lr_predictions)

# -- Get model coefficients ------------------------------------
lr_model_stage = lr_fitted.stages[-1]
coefficients   = lr_model_stage.coefficients.toArray()
intercept      = lr_model_stage.intercept

print(f"\n  MODEL 1 -- LINEAR REGRESSION RESULTS:")
print(f"  {'='*45}")
print(f"  RMSE (Root Mean Sq Error) : {rmse:>15,.2f}")
print(f"  MAE  (Mean Abs Error)     : {mae:>15,.2f}")
print(f"  R²   (Explained Variance) : {r2:>15.4f}")
print(f"  Intercept                 : {intercept:>15,.2f}")
print(f"  {'='*45}")
print(f"  Interpretation:")
print(f"  R² = {r2:.4f} means the model explains "
      f"{r2*100:.1f}% of revenue variance")

print(f"\n  Feature coefficients:")
for fname, coef in zip(lr_feature_cols, coefficients):
    direction = "+" if coef > 0 else "-"
    print(f"  {fname:<22} {coef:>12.2f}  {direction}")

print(f"\n  Sample predictions vs actual:")
lr_predictions.select(
    "line_total",
    F.round("lr_prediction", 2).alias("predicted"),
    F.round(
        F.abs(col("line_total") -
              col("lr_prediction")), 2
    ).alias("error")
).show(8, truncate=False)

# -- VISUALIZATION WITH PLOTLY --
import plotly.graph_objects as go
plot_dir = OUTPUT_DIRS["plots"]

sample_pd = lr_predictions.select("line_total", "lr_prediction").limit(1000).toPandas()

# Plot 1: Actual vs Predicted
fig1 = go.Figure()
fig1.add_trace(go.Scatter(
    x=sample_pd["line_total"], y=sample_pd["lr_prediction"],
    mode="markers", marker=dict(color="#1565C0", size=5, opacity=0.4),
    name="Predictions", hovertemplate="Actual: $%{x:,.0f}<br>Predicted: $%{y:,.0f}<extra></extra>"
))
min_val = builtins.min(sample_pd["line_total"].min(), sample_pd["lr_prediction"].min())
max_val = builtins.max(sample_pd["line_total"].max(), sample_pd["lr_prediction"].max())
fig1.add_trace(go.Scatter(
    x=[min_val, max_val], y=[min_val, max_val],
    mode="lines", line=dict(color="#E53935", width=2, dash="dash"), name="Perfect prediction"
))
fig1.update_layout(
    title=dict(text=f"Actual vs Predicted Revenue (R2 = {r2:.4f})", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Actual Revenue", gridcolor="rgba(0,0,0,0.06)"),
    yaxis=dict(title="Predicted Revenue", gridcolor="rgba(0,0,0,0.06)"),
    plot_bgcolor="white", height=450, margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig1.show()
fig1.write_image(os.path.join(plot_dir, "9_2_lr_actual_vs_predicted.png"), scale=2)
print("  Saved: 9_2_lr_actual_vs_predicted.png")

# Plot 2: Residuals
residuals = sample_pd["line_total"] - sample_pd["lr_prediction"]
fig2 = go.Figure()
fig2.add_trace(go.Scatter(
    x=sample_pd["lr_prediction"], y=residuals,
    mode="markers", marker=dict(color="#388E3C", size=5, opacity=0.4),
    hovertemplate="Predicted: $%{x:,.0f}<br>Residual: $%{y:,.0f}<extra></extra>"
))
fig2.add_hline(y=0, line_dash="dash", line_color="#E53935", line_width=1.5)
fig2.update_layout(
    title=dict(text="Residual Plot (should scatter around 0)", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Predicted Revenue", gridcolor="rgba(0,0,0,0.06)"),
    yaxis=dict(title="Residuals", gridcolor="rgba(0,0,0,0.06)"),
    plot_bgcolor="white", height=450, margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig2.show()
fig2.write_image(os.path.join(plot_dir, "9_2_lr_residuals.png"), scale=2)
print("  Saved: 9_2_lr_residuals.png")

# Plot 3: Feature Coefficients
coef_colors = ["#1E88E5" if c > 0 else "#E53935" for c in coefficients]
fig3 = go.Figure()
fig3.add_trace(go.Bar(
    y=lr_feature_cols, x=coefficients, orientation="h",
    marker=dict(color=coef_colors, line=dict(color="white", width=1)),
    text=[f"{v:.2f}" for v in coefficients], textposition="outside",
    textfont=dict(size=10, color="#2c3e50"),
    hovertemplate="<b>%{y}</b><br>Coefficient: %{x:.4f}<extra></extra>"
))
fig3.add_vline(x=0, line_color="black", line_width=0.8)
fig3.update_layout(
    title=dict(text="Feature Coefficients (blue=positive, red=negative)", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Coefficient Value", gridcolor="rgba(0,0,0,0.06)"),
    yaxis=dict(title=""),
    plot_bgcolor="white", height=450, margin=dict(l=20, r=80, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig3.show()
fig3.write_image(os.path.join(plot_dir, "9_2_lr_coefficients.png"), scale=2)
print("  Saved: 9_2_lr_coefficients.png")

print("=" * 60)
print(f"    Model 1 -- Linear Regression complete")
print(f"  RMSE: {rmse:,.2f} | MAE: {mae:,.2f} | R²: {r2:.4f}")
print("=" * 60)
```

#### 9.3 Model 2: Random Forest (Predict Review Score)

```python
# 
# Model 2 -- Random Forest Regression
# Purpose: Predict review_score (customer satisfaction)
#          Business value: Predict satisfaction BEFORE sale
#          Helps identify high-risk low-satisfaction orders
#
# Features: category, customer_segment, discount_pct,
#           quantity, unit_price, age_group, is_weekend
# Target  : review_score (1-5, treated as continuous)
#


print(f"{'MODEL 2 -- RANDOM FOREST REGRESSION':^60}")


# -- Step 1: Filter valid review scores only -------------------
print("\n  Step 1: Filtering valid review scores (1-5)...")
ml_rf = ml_base \
    .filter(
        (col("review_score") >= 1) &
        (col("review_score") <= 5)
    ) \
    .withColumn("is_weekend_int",
        col("is_weekend").cast(IntegerType()))

rf_count = ml_rf.count()
print(f"  Rows with valid review scores: {rf_count:,}")

# -- Step 2: Encode categorical features ----------------------
print("  Step 2: Encoding categorical features...")

rf_cat_indexer = StringIndexer(
    inputCol="category",
    outputCol="category_idx",
    handleInvalid="keep"
)
rf_seg_indexer = StringIndexer(
    inputCol="customer_segment",
    outputCol="segment_idx",
    handleInvalid="keep"
)
rf_age_indexer = StringIndexer(
    inputCol="age_group",
    outputCol="age_group_idx",
    handleInvalid="keep"
)
rf_status_indexer = StringIndexer(
    inputCol="order_status",
    outputCol="order_status_idx",
    handleInvalid="keep"
)

# -- Step 3: Assemble features ---------------------------------
rf_feature_cols = [
    "category_idx",
    "segment_idx",
    "age_group_idx",
    "order_status_idx",
    "quantity",
    "unit_price",
    "discount_pct",
    "month_number",
    "is_weekend_int"
]

rf_assembler = VectorAssembler(
    inputCols=rf_feature_cols,
    outputCol="features",
    handleInvalid="skip"
)

rf_model_def = RandomForestRegressor(
    featuresCol="features",
    labelCol="review_score",
    predictionCol="rf_prediction",
    numTrees=50,
    maxDepth=8,
    seed=42,
    featureSubsetStrategy="auto"
)

rf_pipeline = Pipeline(stages=[
    rf_cat_indexer,
    rf_seg_indexer,
    rf_age_indexer,
    rf_status_indexer,
    rf_assembler,
    rf_model_def
])

# -- Step 4: Train/test split ----------------------------------
print("  Step 3: Splitting train/test (80/20)...")
rf_train, rf_test = ml_rf.randomSplit(
    [0.8, 0.2], seed=42)
print(f"  Training rows : {rf_train.count():,}")
print(f"  Test rows     : {rf_test.count():,}")

# -- Step 5: Train model ---------------------------------------
print("\n  Step 4: Training Random Forest model...")
t_start    = time.time()
rf_fitted  = rf_pipeline.fit(rf_train)
t_elapsed  = time.time() - t_start
print(f"  Training time : {t_elapsed:.1f}s")

# -- Step 6: Predict and evaluate -----------------------------
print("\n  Step 5: Evaluating on test set...")
rf_predictions = rf_fitted.transform(rf_test)

rf_eval_rmse = RegressionEvaluator(
    labelCol="review_score",
    predictionCol="rf_prediction",
    metricName="rmse"
)
rf_eval_mae  = RegressionEvaluator(
    labelCol="review_score",
    predictionCol="rf_prediction",
    metricName="mae"
)
rf_eval_r2   = RegressionEvaluator(
    labelCol="review_score",
    predictionCol="rf_prediction",
    metricName="r2"
)

rf_rmse = rf_eval_rmse.evaluate(rf_predictions)
rf_mae  = rf_eval_mae.evaluate(rf_predictions)
rf_r2   = rf_eval_r2.evaluate(rf_predictions)

# -- Feature importance ----------------------------------------
rf_model_stage    = rf_fitted.stages[-1]
feature_importances = rf_model_stage \
    .featureImportances.toArray()

print(f"\n  MODEL 2 -- RANDOM FOREST RESULTS:")
print(f"  {'='*45}")
print(f"  RMSE  : {rf_rmse:>10.4f}")
print(f"  MAE   : {rf_mae:>10.4f}")
print(f"  R²    : {rf_r2:>10.4f}")
print(f"  Trees : 50")
print(f"  Depth : 8")
print(f"  {'='*45}")

print(f"\n  Feature importance ranking:")
importance_df = sorted(
    zip(rf_feature_cols, feature_importances),
    key=lambda x: x[1], reverse=True
)
for fname, imp in importance_df:
    bar = "#" * int(imp * 50)
    print(f"  {fname:<22} {imp:.4f}  {bar}")

print(f"\n  Sample predictions vs actual:")
rf_predictions.select(
    "review_score",
    F.round("rf_prediction", 2).alias("predicted"),
    F.round(
        F.abs(col("review_score") -
              col("rf_prediction")), 2
    ).alias("error")
).show(8, truncate=False)

# -- VISUALIZATION WITH PLOTLY --
import plotly.graph_objects as go
plot_dir = OUTPUT_DIRS["plots"]

# Plot 1: Feature Importance
imp_names = [x[0] for x in importance_df]
imp_vals = [x[1] for x in importance_df]
fig1 = go.Figure()
fig1.add_trace(go.Bar(
    y=imp_names[::-1], x=imp_vals[::-1], orientation="h",
    marker=dict(color=imp_vals[::-1], colorscale="Greens", line=dict(color="white", width=1)),
    text=[f"{v:.4f}" for v in imp_vals[::-1]], textposition="outside",
    textfont=dict(size=10, color="#2c3e50"),
    hovertemplate="<b>%{y}</b><br>Importance: %{x:.4f}<extra></extra>"
))
fig1.update_layout(
    title=dict(text="Feature Importance Ranking", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Feature Importance", gridcolor="rgba(0,0,0,0.06)"),
    yaxis=dict(title=""),
    plot_bgcolor="white", height=450, margin=dict(l=20, r=80, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig1.show()
fig1.write_image(os.path.join(plot_dir, "9_3_rf_feature_importance.png"), scale=2)
print("  Saved: 9_3_rf_feature_importance.png")

# Plot 2: Actual vs Predicted
rf_sample_pd = rf_predictions.select("review_score", "rf_prediction").limit(2000).toPandas()
fig2 = go.Figure()
fig2.add_trace(go.Scatter(
    x=rf_sample_pd["review_score"], y=rf_sample_pd["rf_prediction"],
    mode="markers", marker=dict(color="#388E3C", size=5, opacity=0.3),
    hovertemplate="Actual: %{x}<br>Predicted: %{y:.2f}<extra></extra>"
))
fig2.add_trace(go.Scatter(
    x=[1, 5], y=[1, 5], mode="lines",
    line=dict(color="#E53935", width=2, dash="dash"), name="Perfect prediction"
))
fig2.update_layout(
    title=dict(text=f"Actual vs Predicted Review Score (R2 = {rf_r2:.4f})", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Actual Review Score", gridcolor="rgba(0,0,0,0.06)", range=[0.5, 5.5]),
    yaxis=dict(title="Predicted Review Score", gridcolor="rgba(0,0,0,0.06)", range=[0.5, 5.5]),
    plot_bgcolor="white", height=450, margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig2.show()
fig2.write_image(os.path.join(plot_dir, "9_3_rf_actual_vs_predicted.png"), scale=2)
print("  Saved: 9_3_rf_actual_vs_predicted.png")

# Plot 3: Error Distribution
import numpy as np
errors = np.abs(rf_sample_pd["review_score"] - rf_sample_pd["rf_prediction"])
fig3 = go.Figure()
fig3.add_trace(go.Histogram(
    x=errors, nbinsx=30,
    marker=dict(color="#7B1FA2", line=dict(color="white", width=0.5)),
    opacity=0.85, hovertemplate="Error: %{x:.2f}<br>Count: %{y}<extra></extra>"
))
fig3.add_vline(x=errors.mean(), line_dash="dash", line_color="#E53935", line_width=2,
               annotation_text=f"Mean: {errors.mean():.3f}", annotation_position="top right",
               annotation_font=dict(size=10, color="#E53935"))
fig3.add_vline(x=rf_mae, line_dash="dash", line_color="#FB8C00", line_width=2,
               annotation_text=f"MAE: {rf_mae:.3f}", annotation_position="top left",
               annotation_font=dict(size=10, color="#FB8C00"))
fig3.update_layout(
    title=dict(text="Prediction Error Distribution", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Absolute Error", gridcolor="rgba(0,0,0,0.06)"),
    yaxis=dict(title="Count", gridcolor="rgba(0,0,0,0.06)"),
    plot_bgcolor="white", height=450, margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig3.show()
fig3.write_image(os.path.join(plot_dir, "9_3_rf_error_dist.png"), scale=2)
print("  Saved: 9_3_rf_error_dist.png")

print("=" * 60)
print(f"    Model 2 -- Random Forest complete")
print(f"  RMSE: {rf_rmse:.4f} | MAE: {rf_mae:.4f} | "
      f"R²: {rf_r2:.4f}")
print("=" * 60)
```

#### 9.4 Model 3: Logistic Regression (Predict Order Status)

```python
# 
# Model 3 -- Logistic Regression (Classification)
# Purpose: Predict if order will be Delivered (1) or
#          Cancelled (0)
#          Business value: Identify at-risk orders early
#          Flag orders likely to be cancelled for
#          proactive customer service intervention
#
# Features: discount_pct, quantity, unit_price,
#           payment_method, shipping_country,
#           category, customer_segment, month_number
# Target  : binary label (Delivered=1, Cancelled=0)
# 

print("=" * 60)
print(f"{'CELL 50: MODEL 3 -- LOGISTIC REGRESSION':^60}")
print("=" * 60)

# -- Step 1: Filter Delivered and Cancelled only ---------------
print("\n  Step 1: Filtering Delivered vs Cancelled orders...")
ml_log = ml_base \
    .filter(
        col("order_status").isin(
            ["Delivered", "Cancelled"])
    ) \
    .withColumn("label",
        when(col("order_status") == "Delivered",
             lit(1.0))
        .otherwise(lit(0.0))
    ) \
    .withColumn("is_weekend_int",
        col("is_weekend").cast(IntegerType()))

log_count    = ml_log.count()
delivered    = ml_log.filter(
    col("label") == 1.0).count()
cancelled    = ml_log.filter(
    col("label") == 0.0).count()

print(f"  Total rows      : {log_count:,}")
print(f"  Delivered (1)   : {delivered:,} "
      f"({delivered/log_count*100:.1f}%)")
print(f"  Cancelled (0)   : {cancelled:,} "
      f"({cancelled/log_count*100:.1f}%)")

# -- Step 2: Encode categorical features ----------------------
print("\n  Step 2: Encoding categorical features...")

log_cat_indexer = StringIndexer(
    inputCol="category",
    outputCol="category_idx",
    handleInvalid="keep"
)
log_pay_indexer = StringIndexer(
    inputCol="payment_method",
    outputCol="payment_idx",
    handleInvalid="keep"
)
log_ship_indexer = StringIndexer(
    inputCol="shipping_country",
    outputCol="shipping_idx",
    handleInvalid="keep"
)
log_seg_indexer = StringIndexer(
    inputCol="customer_segment",
    outputCol="segment_idx",
    handleInvalid="keep"
)

# -- Step 3: Assemble features ---------------------------------
log_feature_cols = [
    "quantity",
    "unit_price",
    "discount_pct",
    "category_idx",
    "payment_idx",
    "shipping_idx",
    "segment_idx",
    "month_number",
    "is_weekend_int"
]

log_assembler = VectorAssembler(
    inputCols=log_feature_cols,
    outputCol="features_raw",
    handleInvalid="skip"
)

log_scaler = StandardScaler(
    inputCol="features_raw",
    outputCol="features",
    withStd=True,
    withMean=True
)

log_model_def = LogisticRegression(
    featuresCol="features",
    labelCol="label",
    predictionCol="log_prediction",
    probabilityCol="probability",
    rawPredictionCol="rawPrediction",
    maxIter=100,
    regParam=0.01,
    elasticNetParam=0.0
)

log_pipeline = Pipeline(stages=[
    log_cat_indexer,
    log_pay_indexer,
    log_ship_indexer,
    log_seg_indexer,
    log_assembler,
    log_scaler,
    log_model_def
])

# -- Step 4: Train/test split ----------------------------------
print("  Step 3: Splitting train/test (80/20)...")
log_train, log_test = ml_log.randomSplit(
    [0.8, 0.2], seed=42)
print(f"  Training rows : {log_train.count():,}")
print(f"  Test rows     : {log_test.count():,}")

# -- Step 5: Train model ---------------------------------------
print("\n  Step 4: Training Logistic Regression model...")
t_start    = time.time()
log_fitted = log_pipeline.fit(log_train)
t_elapsed  = time.time() - t_start
print(f"  Training time : {t_elapsed:.1f}s")

# -- Step 6: Predict and evaluate -----------------------------
print("\n  Step 5: Evaluating on test set...")
log_predictions = log_fitted.transform(log_test)

# Accuracy
acc_eval = MulticlassClassificationEvaluator(
    labelCol="label",
    predictionCol="log_prediction",
    metricName="accuracy"
)
# Precision
prec_eval = MulticlassClassificationEvaluator(
    labelCol="label",
    predictionCol="log_prediction",
    metricName="weightedPrecision"
)
# Recall
rec_eval = MulticlassClassificationEvaluator(
    labelCol="label",
    predictionCol="log_prediction",
    metricName="weightedRecall"
)
# F1
f1_eval = MulticlassClassificationEvaluator(
    labelCol="label",
    predictionCol="log_prediction",
    metricName="f1"
)

accuracy  = acc_eval.evaluate(log_predictions)
precision = prec_eval.evaluate(log_predictions)
recall    = rec_eval.evaluate(log_predictions)
f1        = f1_eval.evaluate(log_predictions)

# -- Confusion matrix ------------------------------------------
conf_matrix = log_predictions \
    .groupBy("label", "log_prediction") \
    .count() \
    .toPandas()

print(f"\n  MODEL 3 -- LOGISTIC REGRESSION RESULTS:")
print(f"  {'='*45}")
print(f"  Accuracy  : {accuracy:>10.4f} ({accuracy*100:.2f}%)")
print(f"  Precision : {precision:>10.4f}")
print(f"  Recall    : {recall:>10.4f}")
print(f"  F1 Score  : {f1:>10.4f}")
print(f"  {'='*45}")
print(f"  Interpretation:")
print(f"  Accuracy {accuracy*100:.1f}% means the model")
print(f"  correctly classifies {accuracy*100:.1f}% of orders")

print(f"\n  Confusion matrix:")
print(f"  {'':>15} {'Pred Cancel':>12} {'Pred Deliver':>13}")
for actual in [0.0, 1.0]:
    row_label = "  Actual Cancel" \
        if actual == 0.0 else "  Actual Deliver"
    vals = []
    for pred in [0.0, 1.0]:
        match = conf_matrix[
            (conf_matrix["label"] == actual) &
            (conf_matrix["log_prediction"] == pred)
        ]
        val = int(match["count"].values[0]) \
            if len(match) > 0 else 0
        vals.append(val)
    print(f"  {row_label:<16} {vals[0]:>12,} {vals[1]:>13,}")

# -- VISUALIZATION WITH PLOTLY --
import plotly.graph_objects as go
import numpy as np
plot_dir = OUTPUT_DIRS["plots"]

# Plot 1: Confusion Matrix Heatmap
cm_data = np.zeros((2, 2))
for _, row_cm in conf_matrix.iterrows():
    i = int(row_cm["label"])
    j = int(row_cm["log_prediction"])
    cm_data[i][j] = row_cm["count"]

fig1 = go.Figure(data=go.Heatmap(
    z=cm_data, x=["Cancelled", "Delivered"], y=["Cancelled", "Delivered"],
    colorscale="Blues",
    text=[[f"{int(cm_data[i,j]):,}" for j in range(2)] for i in range(2)],
    texttemplate="%{text}", textfont=dict(size=16, color="#2c3e50"),
    hovertemplate="Actual: %{y}<br>Predicted: %{x}<br>Count: %{text}<extra></extra>",
    colorbar=dict(title="Count", thickness=15)
))
fig1.update_layout(
    title=dict(text="Confusion Matrix", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Predicted"), yaxis=dict(title="Actual", autorange="reversed"),
    plot_bgcolor="white", height=420, margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig1.show()
fig1.write_image(os.path.join(plot_dir, "9_4_log_confusion_matrix.png"), scale=2)
print("  Saved: 9_4_log_confusion_matrix.png")

# Plot 2: Classification Metrics
metrics = ["Accuracy", "Precision", "Recall", "F1"]
metric_vals = [accuracy, precision, recall, f1]
metric_colors = ["#1E88E5", "#43A047", "#FB8C00", "#8E24AA"]
fig2 = go.Figure()
fig2.add_trace(go.Bar(
    x=metrics, y=metric_vals,
    marker=dict(color=metric_colors, line=dict(color="white", width=1)),
    text=[f"{v:.4f}" for v in metric_vals], textposition="outside",
    textfont=dict(size=11, color="#2c3e50"),
    hovertemplate="<b>%{x}</b><br>Score: %{y:.4f}<extra></extra>"
))
fig2.add_hline(y=0.5, line_dash="dash", line_color="#E53935", opacity=0.5,
               annotation_text="Random baseline (0.5)", annotation_position="top left",
               annotation_font=dict(size=9, color="#E53935"))
fig2.update_layout(
    title=dict(text="Classification Metrics", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title=""), yaxis=dict(title="Score", gridcolor="rgba(0,0,0,0.06)", range=[0, 1.15]),
    plot_bgcolor="white", height=430, margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12), bargap=0.35
)
fig2.show()
fig2.write_image(os.path.join(plot_dir, "9_4_log_metrics.png"), scale=2)
print("  Saved: 9_4_log_metrics.png")

# Plot 3: Probability Distribution
prob_pd = log_predictions.select(
    "label",
    F.udf(lambda v: float(v[1]), FloatType())("probability").alias("prob_delivered")
).limit(3000).toPandas()

fig3 = go.Figure()
fig3.add_trace(go.Histogram(
    x=prob_pd[prob_pd["label"]==1]["prob_delivered"], nbinsx=30,
    marker=dict(color="#1E88E5", line=dict(color="white", width=0.5)),
    opacity=0.7, name="Delivered"
))
fig3.add_trace(go.Histogram(
    x=prob_pd[prob_pd["label"]==0]["prob_delivered"], nbinsx=30,
    marker=dict(color="#E53935", line=dict(color="white", width=0.5)),
    opacity=0.7, name="Cancelled"
))
fig3.add_vline(x=0.5, line_dash="dash", line_color="black", line_width=1.5,
               annotation_text="Threshold (0.5)", annotation_position="top right",
               annotation_font=dict(size=10))
fig3.update_layout(
    title=dict(text="Probability Distribution by Actual Label", font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Predicted Probability (Delivered)", gridcolor="rgba(0,0,0,0.06)"),
    yaxis=dict(title="Count", gridcolor="rgba(0,0,0,0.06)"),
    barmode="overlay", plot_bgcolor="white", height=450,
    margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1)
)
fig3.show()
fig3.write_image(os.path.join(plot_dir, "9_4_log_probability_dist.png"), scale=2)
print("  Saved: 9_4_log_probability_dist.png")

print("=" * 60)
print(f"    Model 3 -- Logistic Regression complete")
print(f"  Accuracy: {accuracy:.4f} | "
      f"Precision: {precision:.4f} | "
      f"Recall: {recall:.4f} | F1: {f1:.4f}")
print("=" * 60)
```

## Section 10 — Market Basket Analysis

#### 10.1 Prepare Transaction Data

```python
# 
# Prepare Transaction Data
# Purpose: Transform FACT_SALES into transaction format
#          required by FP-Growth algorithm
#          Each transaction = one order_id with list of
#          all product categories purchased together
#
# FP-Growth requires:
#   - Each row = one transaction (order)
#   - Items column = list of products in that transaction
#   - Minimum support threshold = how often items appear
#
# We use CATEGORY level (not product level) because:
#   - 4,781 unique products = too sparse for rules
#   - Categories produce meaningful business rules
#   - "Electronics + Accessories" is actionable insight
# 


# Re-import to ensure F refers to pyspark.sql.functions (not overwritten by pandas/numpy)
from pyspark.sql import functions as F
from pyspark.sql.functions import col

print(f"{'PREPARE TRANSACTION DATA':^60}")


# -- Step 1: Get distinct categories per order -----------------
print("\n  Step 1: Building category-level transactions...")

transactions_raw = spark.sql("""
    SELECT
        o.order_id,
        COLLECT_SET(p.category) AS items
    FROM FACT_SALES   f
    JOIN DIM_ORDER    o ON f.order_sk   = o.order_sk
    JOIN DIM_PRODUCT  p ON f.product_sk = p.product_sk
    WHERE p.category IS NOT NULL
    GROUP BY o.order_id
    HAVING SIZE(COLLECT_SET(p.category)) >= 2
""")

transactions_raw = transactions_raw.cache()
trans_count = transactions_raw.count()

print(f"\n  Total transactions     : {trans_count:,}")
print(f"  (orders with 2+ categories)")

# -- Step 2: Transaction size distribution --------------------
print(f"\n  Transaction size distribution:")
transactions_raw \
    .withColumn("basket_size",
        F.size(col("items"))) \
    .groupBy("basket_size") \
    .count() \
    .orderBy("basket_size") \
    .show(truncate=False)

# -- Step 3: Category frequency across all transactions -------
print(f"  Category frequency in transactions:")
spark.sql("""
    SELECT
        p.category,
        COUNT(DISTINCT o.order_id)  AS order_count,
        ROUND(COUNT(DISTINCT o.order_id) * 100.0 /
              (SELECT COUNT(DISTINCT order_id)
               FROM DIM_ORDER), 2)  AS support_pct
    FROM FACT_SALES   f
    JOIN DIM_ORDER    o ON f.order_sk   = o.order_sk
    JOIN DIM_PRODUCT  p ON f.product_sk = p.product_sk
    GROUP BY p.category
    ORDER BY order_count DESC
""").show(truncate=False)

# -- Step 4: Also prepare subcategory transactions -------------
print(f"  Building subcategory-level transactions...")
transactions_sub = spark.sql("""
    SELECT
        o.order_id,
        COLLECT_SET(p.subcategory) AS items
    FROM FACT_SALES   f
    JOIN DIM_ORDER    o ON f.order_sk   = o.order_sk
    JOIN DIM_PRODUCT  p ON f.product_sk = p.product_sk
    WHERE p.subcategory IS NOT NULL
    GROUP BY o.order_id
    HAVING SIZE(COLLECT_SET(p.subcategory)) >= 2
""")
transactions_sub = transactions_sub.cache()
sub_count = transactions_sub.count()

print(f"  Subcategory transactions: {sub_count:,}")
print(f"\n  Sample transactions (category level):")
transactions_raw.show(5, truncate=False)

print("=" * 60)
print(f"    Transaction data prepared")
print(f"    Category transactions  : {trans_count:,}")
print(f"    Subcategory transactions: {sub_count:,}")
print("=" * 60)
```

#### 10.2 Apply FP-Growth Algorithm

```python
# 
# Apply FP-Growth Algorithm
# Purpose: Find frequent itemsets and association rules
#          FP-Growth is the standard algorithm for MBA
#          Available natively in PySpark MLlib
#
# Key parameters:
#   minSupport    -- minimum fraction of transactions
#                   an itemset must appear in
#                   0.05 = itemset in at least 5% orders
#   minConfidence -- minimum probability that if you buy
#                   item A, you also buy item B
#                   0.3 = 30% of A buyers also buy B
#
# Output:
#   freqItemsets  -- all item combinations above minSupport
#   associationRules -- if-then rules above minConfidence
# 


print(f"{'FP-GROWTH ALGORITHM':^60}")

# -- Run 1: Category level FP-Growth --------------------------
print("\n  Running FP-Growth on CATEGORY level...")
print(f"  minSupport    : 0.05 (5% of transactions)")
print(f"  minConfidence : 0.3  (30% confidence)")

t_start = time.time()

fpgrowth_cat = FPGrowth(
    itemsCol="items",
    minSupport=0.05,
    minConfidence=0.3
)

fpgrowth_model_cat = fpgrowth_cat.fit(transactions_raw)
t_elapsed = time.time() - t_start

# -- Get frequent itemsets -------------------------------------
freq_itemsets_cat = fpgrowth_model_cat.freqItemsets
freq_count_cat    = freq_itemsets_cat.count()

# -- Get association rules -------------------------------------
assoc_rules_cat   = fpgrowth_model_cat.associationRules
rules_count_cat   = assoc_rules_cat.count()

print(f"\n  Training complete in {t_elapsed:.1f}s")
print(f"  Frequent itemsets found : {freq_count_cat:,}")
print(f"  Association rules found : {rules_count_cat:,}")

# -- Add lift calculation --------------------------------------
# Lift = confidence / expected confidence
# Lift > 1 = items bought together MORE than by chance
assoc_rules_cat = assoc_rules_cat \
    .withColumn("lift",
        F.round(col("confidence") / col("support"), 4)) \
    .withColumn("antecedent_str",
        F.concat_ws(" + ", col("antecedent"))) \
    .withColumn("consequent_str",
        F.concat_ws(" + ", col("consequent")))

print(f"\n  All association rules (category level):")
print(f"  {'Antecedent':<30} {'→':>2} "
      f"{'Consequent':<25} "
      f"{'Support':>8} {'Confidence':>11} {'Lift':>7}")
print(f"  {'-'*90}")

rules_pd = assoc_rules_cat \
    .orderBy(desc("lift")) \
    .toPandas()

for _, row in rules_pd.iterrows():
    print(f"  {row['antecedent_str']:<30} → "
          f"{row['consequent_str']:<25} "
          f"{row['support']:>8.4f} "
          f"{row['confidence']:>11.4f} "
          f"{row['lift']:>7.4f}")

# -- Run 2: Subcategory level FP-Growth -----------------------
print(f"\n  Running FP-Growth on SUBCATEGORY level...")
print(f"  minSupport    : 0.02 (2% -- lower for subcategory)")
print(f"  minConfidence : 0.25")

t_start = time.time()

fpgrowth_sub = FPGrowth(
    itemsCol="items",
    minSupport=0.02,
    minConfidence=0.25
)

fpgrowth_model_sub = fpgrowth_sub.fit(transactions_sub)
t_elapsed = time.time() - t_start

freq_itemsets_sub = fpgrowth_model_sub.freqItemsets
assoc_rules_sub   = fpgrowth_model_sub.associationRules

rules_sub_count = assoc_rules_sub.count()
freq_sub_count  = freq_itemsets_sub.count()

# Add lift and string columns
assoc_rules_sub = assoc_rules_sub \
    .withColumn("lift",
        F.round(col("confidence") / col("support"), 4)) \
    .withColumn("antecedent_str",
        F.concat_ws(" + ", col("antecedent"))) \
    .withColumn("consequent_str",
        F.concat_ws(" + ", col("consequent")))

print(f"\n  Subcategory FP-Growth complete in {t_elapsed:.1f}s")
print(f"  Frequent itemsets : {freq_sub_count:,}")
print(f"  Association rules : {rules_sub_count:,}")

print(f"\n  Top 15 subcategory rules by lift:")
assoc_rules_sub \
    .orderBy(desc("lift")) \
    .select(
        "antecedent_str",
        "consequent_str",
        F.round("support", 4).alias("support"),
        F.round("confidence", 4).alias("confidence"),
        F.round("lift", 4).alias("lift")
    ) \
    .show(15, truncate=False)

print("=" * 60)
print(f"    FP-Growth complete")
print(f"    Category rules    : {rules_count_cat:,}")
print(f"    Subcategory rules : {rules_sub_count:,}")
print("=" * 60)
```

#### 10.3 Analyse & Visualize Association Rules

```python
# 
# Analyse & Visualize Association Rules
# Purpose: Deep analysis of association rules
#          Visualize support, confidence, lift relationships
#          Business insight: Product recommendation engine
# 


# Re-import to restore names overwritten by pandas/numpy
from pyspark.sql import functions as F
from pyspark.sql.functions import col, count, desc

print(f"{'ASSOCIATION RULES ANALYSIS':^60}")


# -- Frequent itemsets analysis --------------------------------
print("\n  Frequent itemsets by size:")
freq_itemsets_cat \
    .withColumn("itemset_size",
        F.size(col("items"))) \
    .groupBy("itemset_size") \
    .agg(
        count("*").alias("count"),
        F.round(F.avg("freq"), 1).alias("avg_frequency"),
        F.max("freq").alias("max_frequency")
    ) \
    .orderBy("itemset_size") \
    .show(truncate=False)

print(f"\n  Top 10 most frequent individual categories:")
freq_itemsets_cat \
    .filter(F.size(col("items")) == 1) \
    .orderBy(desc("freq")) \
    .show(10, truncate=False)

print(f"\n  Top 10 most frequent 2-item combinations:")
freq_itemsets_cat \
    .filter(F.size(col("items")) == 2) \
    .orderBy(desc("freq")) \
    .show(10, truncate=False)

# -- Convert to Pandas for visualization ----------------------
rules_pd_cat = assoc_rules_cat \
    .orderBy(desc("lift")) \
    .toPandas()

rules_pd_sub = assoc_rules_sub \
    .orderBy(desc("lift")) \
    .limit(20) \
    .toPandas()

freq_pd = freq_itemsets_cat \
    .filter(F.size(col("items")) == 2) \
    .orderBy(desc("freq")) \
    .limit(15) \
    .toPandas()
freq_pd["items_str"] = freq_pd["items"].apply(
    lambda x: " + ".join(sorted(x)))

# -- VISUALIZATION (Plotly) ------------------------------------
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import os

# === Plot 1: Support vs Confidence Bubble Chart ===
fig1 = go.Figure()
if len(rules_pd_cat) > 0:
    fig1.add_trace(go.Scatter(
        x=rules_pd_cat["support"],
        y=rules_pd_cat["confidence"],
        mode="markers+text",
        marker=dict(
            size=rules_pd_cat["lift"] * 40,
            color=rules_pd_cat["lift"],
            colorscale="YlOrRd",
            colorbar=dict(title="Lift"),
            opacity=0.8,
            line=dict(color="grey", width=0.5)
        ),
        text=[f"{r['antecedent_str'][:15]} -> {r['consequent_str'][:12]}"
              for _, r in rules_pd_cat.iterrows()],
        textposition="top center",
        textfont=dict(size=8),
        hovertemplate=(
            "Antecedent: %{customdata[0]}<br>"
            "Consequent: %{customdata[1]}<br>"
            "Support: %{x:.4f}<br>"
            "Confidence: %{y:.4f}<br>"
            "Lift: %{customdata[2]:.4f}<extra></extra>"
        ),
        customdata=list(zip(
            rules_pd_cat["antecedent_str"],
            rules_pd_cat["consequent_str"],
            rules_pd_cat["lift"]
        )),
        showlegend=False
    ))
fig1.update_layout(
    title=dict(text="Support vs Confidence (bubble size = lift)",
               font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Support", gridcolor="rgba(0,0,0,0.06)"),
    yaxis=dict(title="Confidence", gridcolor="rgba(0,0,0,0.06)"),
    plot_bgcolor="white",
    height=500, width=700,
    margin=dict(l=20, r=20, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
plot_dir = OUTPUT_DIRS["plots"]
fig1.write_image(os.path.join(plot_dir, "10_3_support_vs_confidence.png"), scale=2)
fig1.show()

# === Plot 2: Top Rules by Lift - Horizontal Bar ===
fig2 = go.Figure()
if len(rules_pd_cat) > 0:
    import builtins
    top_rules = rules_pd_cat.head(builtins.min(10, len(rules_pd_cat)))
    rule_labels = [f"{r['antecedent_str'][:18]} -> {r['consequent_str'][:15]}"
                   for _, r in top_rules.iterrows()]
    fig2.add_trace(go.Bar(
        y=rule_labels[::-1],
        x=top_rules["lift"][::-1].values,
        orientation="h",
        marker=dict(
            color=top_rules["lift"][::-1].values,
            colorscale=[[0, "#FFCC80"], [1, "#E65100"]],
            line=dict(color="white", width=0.5)
        ),
        text=[f"{v:.4f}" for v in top_rules["lift"][::-1].values],
        textposition="outside",
        textfont=dict(size=9),
        showlegend=False
    ))
    fig2.add_vline(x=1.0, line_dash="dash", line_color="red",
                   line_width=1.5,
                   annotation_text="Lift=1 (no association)")
fig2.update_layout(
    title=dict(text="Top Rules by Lift Score",
               font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Lift", gridcolor="rgba(0,0,0,0.06)"),
    plot_bgcolor="white",
    height=450, width=700,
    margin=dict(l=200, r=60, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig2.write_image(os.path.join(plot_dir, "10_3_top_rules_lift.png"), scale=2)
fig2.show()

# === Plot 3: Top Rules by Confidence ===
fig3 = go.Figure()
if len(rules_pd_cat) > 0:
    top_conf = rules_pd_cat.sort_values("confidence", ascending=False) \
        .head(builtins.min(10, len(rules_pd_cat)))
    conf_labels = [f"{r['antecedent_str'][:18]} -> {r['consequent_str'][:15]}"
                   for _, r in top_conf.iterrows()]
    fig3.add_trace(go.Bar(
        y=conf_labels[::-1],
        x=top_conf["confidence"][::-1].values,
        orientation="h",
        marker=dict(
            color=top_conf["confidence"][::-1].values,
            colorscale=[[0, "#BBDEFB"], [1, "#1565C0"]],
            line=dict(color="white", width=0.5)
        ),
        text=[f"{v:.4f}" for v in top_conf["confidence"][::-1].values],
        textposition="outside",
        textfont=dict(size=9),
        showlegend=False
    ))
fig3.update_layout(
    title=dict(text="Top Rules by Confidence",
               font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Confidence", gridcolor="rgba(0,0,0,0.06)"),
    plot_bgcolor="white",
    height=450, width=700,
    margin=dict(l=200, r=60, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig3.write_image(os.path.join(plot_dir, "10_3_top_rules_confidence.png"), scale=2)
fig3.show()

# === Plot 4: Top 15 Frequent 2-Category Pairs ===
fig4 = go.Figure()
if len(freq_pd) > 0:
    fig4.add_trace(go.Bar(
        y=freq_pd["items_str"][::-1],
        x=freq_pd["freq"][::-1],
        orientation="h",
        marker=dict(
            color=freq_pd["freq"][::-1].values,
            colorscale=[[0, "#C8E6C9"], [1, "#2E7D32"]],
            line=dict(color="white", width=0.5)
        ),
        text=[f"{v:,}" for v in freq_pd["freq"][::-1]],
        textposition="outside",
        textfont=dict(size=9),
        showlegend=False
    ))
fig4.update_layout(
    title=dict(text="Top 15 Most Frequent 2-Category Combinations",
               font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Transaction Frequency", gridcolor="rgba(0,0,0,0.06)"),
    plot_bgcolor="white",
    height=500, width=700,
    margin=dict(l=220, r=60, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig4.write_image(os.path.join(plot_dir, "10_3_frequent_pairs.png"), scale=2)
fig4.show()

# === Plot 5: Top Subcategory Rules by Lift ===
fig5 = go.Figure()
if len(rules_pd_sub) > 0:
    top_sub = rules_pd_sub.head(builtins.min(12, len(rules_pd_sub)))
    sub_labels = [f"{r['antecedent_str'][:20]} -> {r['consequent_str'][:18]}"
                  for _, r in top_sub.iterrows()]
    fig5.add_trace(go.Bar(
        y=sub_labels[::-1],
        x=top_sub["lift"][::-1].values,
        orientation="h",
        marker=dict(
            color=top_sub["lift"][::-1].values,
            colorscale=[[0, "#E1BEE7"], [1, "#6A1B9A"]],
            line=dict(color="white", width=0.5)
        ),
        text=[f"{v:.3f}" for v in top_sub["lift"][::-1].values],
        textposition="outside",
        textfont=dict(size=8),
        showlegend=False
    ))
    fig5.add_vline(x=1.0, line_dash="dash", line_color="red",
                   line_width=1.5,
                   annotation_text="Lift=1")
fig5.update_layout(
    title=dict(text="Top Subcategory Rules by Lift",
               font=dict(size=16, color="#2c3e50")),
    xaxis=dict(title="Lift", gridcolor="rgba(0,0,0,0.06)"),
    plot_bgcolor="white",
    height=500, width=750,
    margin=dict(l=250, r=60, t=60, b=40),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig5.write_image(os.path.join(plot_dir, "10_3_subcategory_rules_lift.png"), scale=2)
fig5.show()

# === Plot 6: MBA Metrics Table ===
fig6 = go.Figure(data=[go.Table(
    header=dict(
        values=["<b>Metric</b>", "<b>Formula</b>", "<b>Meaning</b>"],
        fill_color="#1565C0",
        font=dict(color="white", size=12),
        align="center",
        height=32
    ),
    cells=dict(
        values=[
            ["Support", "Confidence", "Lift", "", ""],
            ["P(A and B)", "P(B|A)", "Conf / P(B)", "Lift = 1", "Lift < 1"],
            ["How often A+B appear together",
             "P(buy B given bought A)",
             "Lift>1 = strong association",
             "No association (random)",
             "Negative association"]
        ],
        fill_color=[["#EBF3FB", "white"] * 3],
        font=dict(size=11),
        align="center",
        height=28
    )
)])
fig6.update_layout(
    title=dict(text="MBA Metrics - Definitions",
               font=dict(size=16, color="#2c3e50")),
    height=300, width=700,
    margin=dict(l=20, r=20, t=60, b=20),
    font=dict(family="Segoe UI, sans-serif", size=12)
)
fig6.write_image(os.path.join(plot_dir, "10_3_mba_metrics_table.png"), scale=2)
fig6.show()

print(f"  Plots saved to: {plot_dir}")

print("\n" + "=" * 60)
print("   Association rules analysis complete")
print("=" * 60)
```

## Section 11 — Real-Time Streaming Pipeline

#### 11.1 Prepare Streaming Source Data

```python
# 
# Prepare Streaming Source Data
# Purpose: Create a simulated streaming data source
#          Write sample orders as JSON files into a
#          watched folder -- simulates live order arrivals
#
# How streaming simulation works:
#   1. We write batches of orders as JSON files
#   2. Spark Structured Streaming watches the folder
#   3. Every time a new file appears, Spark processes it
#   4. Results update in real-time
#
# Real-world equivalent:
#   Kafka topic / AWS Kinesis / Azure Event Hub
#   All produce data the same way -- Spark reads it
#   The folder-watching approach is functionally identical
#   for demonstration and coursework purposes
# 


print(f"{'PREPARE STREAMING SOURCE DATA':^60}")


import shutil
import json

# -- Clean and recreate stream input folder --------------------
STREAM_INPUT = OUTPUT_DIRS["stream_input"]
STREAM_OUTPUT = os.path.join(
    OUTPUT_DIRS["streaming"], "results")
STREAM_CHECKPOINT = os.path.join(
    OUTPUT_DIRS["streaming"], "checkpoint")

# Clean old streaming folders
for folder in [STREAM_INPUT, STREAM_OUTPUT,
               STREAM_CHECKPOINT]:
    if os.path.exists(folder):
        shutil.rmtree(folder)
    os.makedirs(folder, exist_ok=True)

print(f"\n  Stream folders prepared:")
print(f"  Input      : {STREAM_INPUT}")
print(f"  Output     : {STREAM_OUTPUT}")
print(f"  Checkpoint : {STREAM_CHECKPOINT}")

# -- Build enriched streaming dataset from star schema --------
# Use FACT_SALES joined with dimensions for rich stream data
print(f"\n  Building streaming source dataset...")

streaming_source = spark.sql("""
    SELECT
        f.order_item_id,
        o.order_id,
        o.order_date,
        o.order_status,
        o.payment_method,
        o.shipping_country,
        c.customer_segment,
        c.country         AS customer_country,
        c.age_group,
        p.category,
        p.subcategory,
        ROUND(f.line_total, 2)       AS line_total,
        f.quantity,
        ROUND(f.unit_price_at_sale, 2) AS unit_price_at_sale,
        ROUND(f.discount_pct, 4)     AS discount_pct,
        f.review_score,
        d.year,
        d.month_number,
        d.month_name,
        d.quarter,
        d.is_weekend
    FROM FACT_SALES   f
    JOIN DIM_ORDER    o ON f.order_sk    = o.order_sk
    JOIN DIM_CUSTOMER c ON f.customer_sk = c.customer_sk
    JOIN DIM_PRODUCT  p ON f.product_sk  = p.product_sk
    JOIN DIM_DATE     d ON f.date_sk     = d.date_sk
""")

total_stream_rows = streaming_source.count()
print(f"  Total rows available for streaming: "
      f"{total_stream_rows:,}")

# -- Write 5 batches of JSON files -----------------------------
# Each batch = one simulated "wave" of incoming orders
BATCH_SIZE = 500
N_BATCHES  = 5

print(f"\n  Writing {N_BATCHES} streaming batches "
      f"({BATCH_SIZE} rows each)...")
print(f"  {'Batch':<8} {'Rows':>6} {'File':>8} {'Status':>8}")
print(f"  {'-'*40}")

# Collect sample rows for batch writing
sample_rows = streaming_source \
    .limit(BATCH_SIZE * N_BATCHES) \
    .toPandas()

for batch_num in range(N_BATCHES):
    start_idx = batch_num * BATCH_SIZE
    end_idx   = start_idx + BATCH_SIZE
    batch_df  = sample_rows.iloc[start_idx:end_idx]

    # Convert boolean columns
    batch_df = batch_df.copy()
    for bool_col in ["is_weekend"]:
        if bool_col in batch_df.columns:
            batch_df[bool_col] = batch_df[
                bool_col].astype(bool)

    # Convert date columns to string
    for date_col in ["order_date"]:
        if date_col in batch_df.columns:
            batch_df[date_col] = batch_df[
                date_col].astype(str)

    # Write as JSON (one record per line -- JSONL format)
    file_path = os.path.join(
        STREAM_INPUT,
        f"orders_batch_{batch_num:02d}.json"
    )
    batch_df.to_json(
        file_path,
        orient="records",
        lines=True
    )
    print(f"  Batch {batch_num:<3} "
          f"{len(batch_df):>6,} rows  "
          f"batch_{batch_num:02d}.json ")

# -- Verify files written --------------------------------------
files_written = [f for f in os.listdir(STREAM_INPUT)
                 if f.endswith(".json")]
total_written = N_BATCHES * BATCH_SIZE

print(f"\n  Files in stream input folder:")
for f in sorted(files_written):
    fpath = os.path.join(STREAM_INPUT, f)
    fsize = os.path.getsize(fpath) / 1024
    print(f"  {f:<35} ({fsize:.1f} KB)")

print(f"\n  Total rows written : {total_written:,}")
print(f"  Total files        : {len(files_written)}")
print(f"  Format             : JSONL (one record/line)")
print(f"  Simulates          : Live order feed")

print("=" * 60)
print("  Streaming source data prepared")
print("  5 batches ready for stream processing")
print("=" * 60)
```

#### 11.2 Define Streaming Schema & Read Stream

```python
# 
# Define Streaming Schema & Read Stream
# Purpose: Define exact schema for streaming data
#          Create readStream to watch input folder
#          Spark will detect new files automatically
#
# Key concepts:
#   readStream  -- continuous read (not one-time)
#   schema      -- must be explicitly defined for streams
#   maxFilesPerTrigger -- process 1 file at a time
#                        simulates real-time arrival
# 


print(f"{'DEFINE STREAMING SCHEMA & READ STREAM':^60}")


# -- Define streaming schema -----------------------------------
# Must exactly match JSON field names and types
stream_schema = StructType([
    StructField("order_item_id",     StringType(),  True),
    StructField("order_id",          StringType(),  True),
    StructField("order_date",        StringType(),  True),
    StructField("order_status",      StringType(),  True),
    StructField("payment_method",    StringType(),  True),
    StructField("shipping_country",  StringType(),  True),
    StructField("customer_segment",  StringType(),  True),
    StructField("customer_country",  StringType(),  True),
    StructField("age_group",         StringType(),  True),
    StructField("category",          StringType(),  True),
    StructField("subcategory",       StringType(),  True),
    StructField("line_total",        DoubleType(),  True),
    StructField("quantity",          IntegerType(), True),
    StructField("unit_price_at_sale",DoubleType(),  True),
    StructField("discount_pct",      DoubleType(),  True),
    StructField("review_score",      IntegerType(), True),
    StructField("year",              IntegerType(), True),
    StructField("month_number",      IntegerType(), True),
    StructField("month_name",        StringType(),  True),
    StructField("quarter",           StringType(),  True),
    StructField("is_weekend",        BooleanType(), True),
])

print(f"\n  Schema defined: {len(stream_schema)} fields")

# -- Create streaming DataFrame --------------------------------
stream_df = spark.readStream \
    .schema(stream_schema) \
    .option("maxFilesPerTrigger", 1) \
    .json(STREAM_INPUT)

print(f"\n  Streaming DataFrame created:")
print(f"  Source folder  : {STREAM_INPUT}")
print(f"  Format         : JSON")
print(f"  Files per batch: 1 (maxFilesPerTrigger)")
print(f"  isStreaming    : {stream_df.isStreaming}")

print(f"\n  Stream schema:")
for field in stream_schema.fields:
    print(f"  {field.name:<25} "
          f"{str(field.dataType):<15}")

# -- Apply real-time transformations --------------------------
print(f"\n  Applying real-time transformations...")

stream_transformed = stream_df \
    .withColumn("order_date_cast",
        to_date(col("order_date"))) \
    .withColumn("revenue_band",
        when(col("line_total") < 10000,
             lit("Low (<10K)"))
        .when(col("line_total") < 50000,
             lit("Medium (10K-50K)"))
        .when(col("line_total") < 100000,
             lit("High (50K-100K)"))
        .otherwise(lit("Premium (100K+)"))
    ) \
    .withColumn("is_discounted",
        when(col("discount_pct") > 0,
             lit("Discounted"))
        .otherwise(lit("Full Price"))
    ) \
    .filter(col("line_total") > 0) \
    .filter(col("category").isNotNull())

print(f"  revenue_band column derived")
print(f"  is_discounted column derived")
print(f"  NULL filters applied")

print("\n" + "=" * 60)
print("  Streaming DataFrame ready")
print(f"  isStreaming = {stream_df.isStreaming}")
print("=" * 60)
```

#### 11.3 Real-Time Aggregations & Stream Output

```python
# 
# Real-Time Streaming with Live Dashboard
# Purpose: Start streaming queries + launch Streamlit
#          dashboard that auto-refreshes every 5 seconds
#          showing live aggregations as batches arrive
#
# What happens:
#   1. Streaming query writes processed data to Parquet
#   2. Streamlit dashboard reads Parquet every 5 seconds
#   3. Dashboard URL appears below - open in browser
#   4. Run NEXT CELL to stop streaming + save results
# 


print(f"{'STREAMING + LIVE DASHBOARD':^60}")


import subprocess, json as _json, time as _time, signal

# -- Write streaming status file --------------------------------
status_file = os.path.join(
    OUTPUT_DIRS["streaming"], "stream_status.json")

status_data = {
    "status": "running",
    "start_time": str(_time.time()),
    "batches": N_BATCHES,
    "batch_size": BATCH_SIZE,
}
with open(status_file, "w") as _f:
    _json.dump(status_data, _f)

print(f"  Status file written: {status_file}")

# -- Stop any existing streaming queries -----------------------
for _q in spark.streams.active:
    _q.stop()
    print(f"  Stopped existing query: {_q.name}")

# -- Clean old output for fresh run ----------------------------
import shutil as _shutil
for _folder in [STREAM_OUTPUT, STREAM_CHECKPOINT]:
    if os.path.exists(_folder):
        _shutil.rmtree(_folder)
    os.makedirs(_folder, exist_ok=True)

# -- Start streaming query: Save to Parquet --------------------
print(f"\n  Starting streaming query (Parquet sink)...")

stream_query = stream_transformed \
    .writeStream \
    .outputMode("append") \
    .format("parquet") \
    .option("path", STREAM_OUTPUT) \
    .option("checkpointLocation", STREAM_CHECKPOINT) \
    .queryName("realtime_parquet_sink") \
    .trigger(processingTime="5 seconds") \
    .start()

print(f"  Stream query started: realtime_parquet_sink")
print(f"  Output: {STREAM_OUTPUT}")
print(f"  Trigger: every 5 seconds")
print(f"  isStreaming: {stream_query.isActive}")

# -- Wait for first batch to process ---------------------------
print(f"\n  Waiting for first batch to process...")
_time.sleep(8)

print(f"  Stream progress:")
if stream_query.lastProgress:
    prog = stream_query.lastProgress
    print(f"  Batch ID     : {prog.get('batchId', 'N/A')}")
    print(f"  Input rows   : {prog.get('numInputRows', 'N/A')}")
    print(f"  Rows/sec     : "
          f"{prog.get('processedRowsPerSecond', 0):.1f}")

# -- Launch Streamlit Dashboard --------------------------------
print(f"\n  Launching Streamlit real-time dashboard...")

streamlit_proc = subprocess.Popen(
    [
        os.path.join(os.path.dirname(os.getcwd()), "venv", "bin", "streamlit"),
        "run", os.path.join(os.path.dirname(os.getcwd()), "streamlit_realtime.py"),
        "--server.port", "8502",
        "--server.headless", "true",
        "--browser.gatherUsageStats", "false",
    ],
    stdout=subprocess.DEVNULL,
    stderr=subprocess.DEVNULL,
    preexec_fn=os.setpgrp,
)

# Save process info for cleanup
_streamlit_pid = streamlit_proc.pid
pid_file = os.path.join(
    OUTPUT_DIRS["streaming"], "streamlit_pid.txt")
with open(pid_file, "w") as _f:
    _f.write(str(_streamlit_pid))

_time.sleep(3)

print(f"\n" + "=" * 60)
print(f"  LIVE DASHBOARD READY")
print(f"=" * 60)
print(f"\n  Open in browser:")
print(f"  --> http://localhost:8502")
print(f"\n  Dashboard auto-refreshes every 5 seconds")
print(f"  showing live data as Spark processes batches.")
print(f"\n  Streaming is RUNNING in background.")
print(f"  --> Run the NEXT CELL to stop streaming,")
print(f"      shut down dashboard, and save results.")
print(f"\n" + "=" * 60)
```

#### 11.4 Streaming Results Analysis & Visualization

```python
# 
# Stop Streaming, Close Dashboard & Save Results
# Purpose: Stop all streaming queries
#          Shut down Streamlit dashboard
#          Read back processed data
#          Save final results + visualization
# 


print(f"{'STOP STREAMING & SAVE RESULTS':^60}")


import signal, json as _json, time as _time

# -- Stop all streaming queries --------------------------------
print(f"\n  Stopping streaming queries...")
for q in spark.streams.active:
    q.stop()
    print(f"  Stopped: {q.name}")

if not spark.streams.active:
    print(f"  All streaming queries stopped.")

# -- Shut down Streamlit dashboard -----------------------------
print(f"\n  Shutting down Streamlit dashboard...")
pid_file = os.path.join(
    OUTPUT_DIRS["streaming"], "streamlit_pid.txt")

try:
    with open(pid_file) as _f:
        _pid = int(_f.read().strip())
    os.killpg(os.getpgid(_pid), signal.SIGTERM)
    print(f"  Streamlit process (PID {_pid}) terminated.")
except Exception as _e:
    print(f"  Streamlit cleanup: {str(_e)[:50]}")
    # Fallback: kill by port
    import subprocess as _sp
    _sp.run(["pkill", "-f", "streamlit_realtime.py"],
            capture_output=True)
    print(f"  Streamlit process killed by name.")

# -- Update status file ----------------------------------------
status_file = os.path.join(
    OUTPUT_DIRS["streaming"], "stream_status.json")
try:
    with open(status_file) as _f:
        status_data = _json.load(_f)
except Exception:
    status_data = {}

status_data["status"] = "stopped"
status_data["end_time"] = str(_time.time())

with open(status_file, "w") as _f:
    _json.dump(status_data, _f)

print(f"  Status updated: stopped")

# -- Read back processed streaming data ------------------------
print(f"\n  Reading processed streaming data...")

try:
    streamed_df = spark.read.parquet(STREAM_OUTPUT)
    streamed_count = streamed_df.count()
    print(f"  Rows processed by stream: {streamed_count:,}")
except Exception as _e:
    print(f"  Stream output note: {str(_e)[:60]}")
    streamed_count = 0

# -- Recompute aggregations on final streamed data -------------
if streamed_count > 0:
    print(f"\n  Final aggregations from streamed data:")

    print(f"\n  Revenue by category (from stream):")
    stream_cat = streamed_df \
        .groupBy("category") \
        .agg(
            count("order_item_id").alias("items_processed"),
            F.round(sum("line_total"), 2).alias(
                "streaming_revenue"),
            F.round(avg("discount_pct") * 100, 2).alias(
                "avg_discount")
        ) \
        .orderBy(desc("streaming_revenue"))
    stream_cat.show(truncate=False)

    print(f"\n  Revenue by country (from stream):")
    stream_country = streamed_df \
        .groupBy("shipping_country") \
        .agg(
            count("order_id").alias("orders"),
            F.round(sum("line_total"), 2).alias("revenue")
        ) \
        .orderBy(desc("revenue"))
    stream_country.show(10, truncate=False)

    print(f"\n  Revenue band distribution:")
    streamed_df \
        .groupBy("revenue_band") \
        .count() \
        .orderBy(desc("count")) \
        .show(truncate=False)

    # Convert to Pandas
    stream_cat_pd = stream_cat.toPandas()
    stream_country_pd = stream_country.limit(10).toPandas()
else:
    # Fallback: use batch aggregation
    print(f"  Using batch aggregation for visualization...")
    sample_stream = streaming_source \
        .limit(BATCH_SIZE * N_BATCHES)

    stream_cat_pd = sample_stream \
        .groupBy("category") \
        .agg(
            count("order_item_id").alias("items_processed"),
            F.round(sum("line_total"), 2).alias(
                "streaming_revenue"),
            F.round(avg("discount_pct") * 100, 2).alias(
                "avg_discount")
        ) \
        .orderBy(desc("streaming_revenue")) \
        .toPandas()

    stream_country_pd = sample_stream \
        .groupBy("shipping_country") \
        .agg(
            count("order_id").alias("orders"),
            F.round(sum("line_total"), 2).alias("revenue")
        ) \
        .orderBy(desc("revenue")) \
        .limit(10) \
        .toPandas()

    streamed_count = BATCH_SIZE * N_BATCHES

# -- VISUALIZATION (Plotly) ------------------------------------
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        "Streaming Revenue by Category",
        "Streaming Revenue by Country (Top 10)",
        "Streaming Pipeline Architecture",
        "Streaming Pipeline KPIs"
    ),
    specs=[
        [{"type": "bar"}, {"type": "bar"}],
        [{"type": "table"}, {"type": "table"}]
    ],
    vertical_spacing=0.12,
    horizontal_spacing=0.08,
)

# Plot 1: Revenue by Category
if len(stream_cat_pd) > 0:
    cat_sorted = stream_cat_pd.sort_values(
        "streaming_revenue", ascending=True)
    fig.add_trace(go.Bar(
        y=cat_sorted["category"],
        x=cat_sorted["streaming_revenue"],
        orientation="h",
        marker=dict(
            color=cat_sorted["streaming_revenue"].values,
            colorscale=[[0, "#BBDEFB"], [1, "#1565C0"]]),
        text=[f"${v/1e3:,.0f}K"
              for v in cat_sorted["streaming_revenue"]],
        textposition="outside",
        textfont=dict(size=9),
        showlegend=False,
    ), row=1, col=1)

# Plot 2: Revenue by Country
if len(stream_country_pd) > 0:
    cty_sorted = stream_country_pd.sort_values(
        "revenue", ascending=True)
    fig.add_trace(go.Bar(
        y=cty_sorted["shipping_country"],
        x=cty_sorted["revenue"],
        orientation="h",
        marker=dict(
            color=cty_sorted["revenue"].values,
            colorscale=[[0, "#C8E6C9"], [1, "#2E7D32"]]),
        text=[f"${v/1e3:,.0f}K"
              for v in cty_sorted["revenue"]],
        textposition="outside",
        textfont=dict(size=9),
        showlegend=False,
    ), row=1, col=2)

# Plot 3: Pipeline Architecture Table
fig.add_trace(go.Table(
    header=dict(
        values=["<b>Component</b>", "<b>Detail</b>",
                "<b>Status</b>"],
        fill_color="#1565C0",
        font=dict(color="white", size=11),
        align="center", height=28,
    ),
    cells=dict(
        values=[
            ["Source Folder", "Spark Streaming",
             "Transformations", "Parquet Sink"],
            [f"{N_BATCHES} JSON batches",
             "readStream (5s trigger)",
             "revenue_band, is_discounted",
             f"{streamed_count:,} rows saved"],
            ["Completed", "Completed",
             "Completed", "Completed"],
        ],
        fill_color=[["#EBF3FB", "white"] * 2],
        font=dict(size=10), align="center", height=26,
    ),
), row=2, col=1)

# Plot 4: KPI Table
stream_total_rev = (stream_cat_pd["streaming_revenue"].sum()
                    if len(stream_cat_pd) > 0 else 0)
fig.add_trace(go.Table(
    header=dict(
        values=["<b>KPI</b>", "<b>Value</b>"],
        fill_color="#2E7D32",
        font=dict(color="white", size=11),
        align="center", height=28,
    ),
    cells=dict(
        values=[
            ["Rows Processed", "Batches", "Categories",
             "Total Revenue", "Trigger Interval", "Output"],
            [f"{streamed_count:,}", f"{N_BATCHES}",
             f"{len(stream_cat_pd)}",
             f"${stream_total_rev/1e3:,.0f}K",
             "5 seconds", "Parquet + Dashboard"],
        ],
        fill_color=[["#E8F5E9", "white"] * 3],
        font=dict(size=10), align="center", height=26,
    ),
), row=2, col=2)

fig.update_layout(
    title=dict(
        text="Real-Time Streaming Pipeline - Final Results",
        font=dict(size=18, color="#2c3e50")),
    plot_bgcolor="white",
    height=750, width=1100,
    margin=dict(l=20, r=20, t=80, b=30),
    font=dict(family="Segoe UI, sans-serif", size=12),
)

plot_dir = OUTPUT_DIRS["plots"]
plot_path = os.path.join(plot_dir,
    "11_4_streaming_results.png")
fig.write_image(plot_path, scale=2)
fig.show()
print(f"\n  Plot saved: {plot_path}")

print("\n" + "=" * 60)
print("  Streaming pipeline stopped")
print("  Streamlit dashboard shut down")
print("  All results saved")
print("=" * 60)
```

## Section 12 — Parquet Output Storage

#### 12.1 Save All Cleaned Datasets to Parquet

```python
# 
#  Save All Cleaned Datasets to Parquet
# Purpose: Persist all 7 cleaned datasets to Parquet
#          Parquet is the industry standard for big data
#          storage -- columnar, compressed, schema-preserving
#
# Why Parquet over CSV:
#     5-10x smaller file size (columnar compression)
#     Preserves data types (no re-casting on reload)
#     Columnar reads -- only reads needed columns
#     Supports predicate pushdown (filter before read)
#     Compatible with Spark, Power BI, Pandas, Hive
# 


print(f"{'SAVE CLEANED DATASETS TO PARQUET':^60}")


# -- Rebuild cleaned DataFrames if not in memory ---------------
#    (needed when resuming via Quick Resume cell)

import os
import time as time_module

DATA_PATH = os.path.join(BASE_PATH, "data")

_need_rebuild = False
for _vname in ["customers_clean", "orders_clean", "order_items_clean",
               "products_clean", "reviews_clean", "suppliers_clean", "date_clean"]:
    if _vname not in globals():
        _need_rebuild = True
        break

if _need_rebuild:
    print("  Rebuilding cleaned DataFrames from raw CSVs...\n")

    read_opts = {
        "header": "true", "escape": '"', "encoding": "UTF-8",
        "ignoreLeadingWhiteSpace": "true",
        "ignoreTrailingWhiteSpace": "true"
    }

    # -- customers_clean --
    _cust_raw = spark.read.options(**read_opts) \
        .csv(os.path.join(DATA_PATH, "customers.csv"),
             inferSchema=True)
    customers_clean = _cust_raw \
        .withColumn("email",
            when(col("email").isNull() | (trim(col("email")) == ""),
                 lit("unknown@unknown.com"))
            .otherwise(lower(regexp_replace(col("email"), " ", "")))) \
        .withColumn("age_group",
            when(col("age_group").isNull() | (trim(col("age_group")) == ""),
                 lit("Unknown"))
            .otherwise(col("age_group"))) \
        .withColumn("customer_name", lower(trim(col("customer_name")))) \
        .withColumn("city", lower(trim(col("city")))) \
        .withColumn("state_province", lower(trim(col("state_province")))) \
        .withColumn("country", lower(trim(col("country")))) \
        .withColumn("customer_segment", lower(trim(col("customer_segment")))) \
        .withColumn("registration_date",
            to_date(col("registration_date"), "M/d/yyyy"))
    customers_clean.cache()
    print(f"    customers_clean  : {customers_clean.count():>10,} rows")

    # -- products_clean (needed before order_items) --
    _prod_raw = spark.read.options(**read_opts) \
        .csv(os.path.join(DATA_PATH, "products.csv"),
             inferSchema=True)
    products_clean = _prod_raw \
        .withColumn("category",
            when(col("category").isNull() | (trim(col("category")) == ""),
                 col("subcategory"))
            .otherwise(col("category"))) \
        .withColumn("unit_price",
            when(col("unit_price").isNull() | (col("unit_price") == 0),
                 lit(-1.0))
            .otherwise(col("unit_price"))) \
        .withColumn("is_active",
            col("is_active").cast(BooleanType()))
    products_clean.cache()
    print(f"    products_clean   : {products_clean.count():>10,} rows")

    # -- orders_clean --
    _ord_raw = spark.read.options(**read_opts) \
        .csv(os.path.join(DATA_PATH, "orders.csv"),
             inferSchema=True)
    _ord_deduped = _ord_raw.dropDuplicates(["order_id"])
    _valid_cust_ids = customers_clean.select("customer_id")
    _ord_no_orphans = _ord_deduped.join(
        _valid_cust_ids, "customer_id", "inner")
    orders_clean = _ord_no_orphans \
        .withColumn("discount_pct",
            when(col("discount_pct").isNull(), lit(0.0))
            .otherwise(col("discount_pct"))) \
        .withColumn("order_date",
            to_date(col("order_date"), "M/d/yyyy"))
    orders_clean.cache()
    print(f"    orders_clean     : {orders_clean.count():>10,} rows")

    # -- order_items_clean --
    _oi_raw = spark.read.options(**read_opts) \
        .csv(os.path.join(DATA_PATH, "order_items.csv"),
             inferSchema=True)
    _valid_prod_ids = _prod_raw.select("product_id")
    _oi_no_orphans = _oi_raw.join(
        _valid_prod_ids, "product_id", "inner")
    order_items_clean = _oi_no_orphans \
        .withColumn("line_total",
            F.round(col("quantity") * col("unit_price_at_sale"), 2))
    order_items_clean.cache()
    print(f"    order_items_clean: {order_items_clean.count():>10,} rows")

    # -- reviews_clean --
    _rev_raw = spark.read.options(**read_opts) \
        .csv(os.path.join(DATA_PATH, "reviews.csv"),
             inferSchema=True)
    reviews_clean = _rev_raw \
        .withColumn("review_score",
            when(col("review_score").isNull(), lit(0))
            .when(col("review_score") == 6, lit(3))
            .otherwise(col("review_score").cast(IntegerType()))) \
        .withColumn("review_date",
            to_date(col("review_date"), "M/d/yyyy"))
    reviews_clean.cache()
    print(f"    reviews_clean    : {reviews_clean.count():>10,} rows")

    # -- suppliers_clean --
    _sup_raw = spark.read.options(**read_opts) \
        .csv(os.path.join(DATA_PATH, "suppliers.csv"),
             inferSchema=True)
    suppliers_clean = _sup_raw \
        .withColumn("contract_start",
            to_date(col("contract_start"), "M/d/yyyy"))
    suppliers_clean.cache()
    print(f"    suppliers_clean  : {suppliers_clean.count():>10,} rows")

    # -- date_clean --
    _date_raw = spark.read.options(**read_opts) \
        .csv(os.path.join(DATA_PATH, "date_dim_helper.csv"),
             inferSchema=True)
    date_clean = _date_raw \
        .withColumn("full_date",
            to_date(col("full_date"), "M/d/yyyy")) \
        .withColumn("day_of_week",
            when(col("day_of_week") == "Monday", "Mon")
            .when(col("day_of_week") == "Tuesday", "Tue")
            .when(col("day_of_week") == "Wednesday", "Wed")
            .when(col("day_of_week") == "Thursday", "Thu")
            .when(col("day_of_week") == "Friday", "Fri")
            .when(col("day_of_week") == "Saturday", "Sat")
            .when(col("day_of_week") == "Sunday", "Sun")
            .otherwise(col("day_of_week"))) \
        .withColumn("month_name",
            when(col("month_name") == "January", "Jan")
            .when(col("month_name") == "February", "Feb")
            .when(col("month_name") == "March", "Mar")
            .when(col("month_name") == "April", "Apr")
            .when(col("month_name") == "May", "May")
            .when(col("month_name") == "June", "Jun")
            .when(col("month_name") == "July", "Jul")
            .when(col("month_name") == "August", "Aug")
            .when(col("month_name") == "September", "Sep")
            .when(col("month_name") == "October", "Oct")
            .when(col("month_name") == "November", "Nov")
            .when(col("month_name") == "December", "Dec")
            .otherwise(col("month_name"))) \
        .withColumn("bs_year",
            when(col("month_number") <= 4, col("year") + 56)
            .otherwise(col("year") + 57)) \
        .withColumn("bs_full_date",
            concat(
                col("bs_year").cast(StringType()), lit("-"),
                lpad(col("month_number").cast(StringType()), 2, "0"), lit("-"),
                lpad(col("day_number").cast(StringType()), 2, "0"))) \
        .withColumn("is_weekend",
            col("is_weekend").cast(BooleanType())) \
        .withColumn("is_np_holiday",
            col("is_np_holiday").cast(BooleanType()))
    date_clean.cache()
    print(f"    date_clean       : {date_clean.count():>10,} rows")

    print("\n  All 7 cleaned DataFrames rebuilt successfully\n")
else:
    print("  All cleaned DataFrames already in memory\n")


# -- Save to Parquet -------------------------------------------

PARQUET_CLEANED = os.path.join(
    OUTPUT_DIRS["parquet"], "cleaned")
os.makedirs(PARQUET_CLEANED, exist_ok=True)

# All 7 cleaned datasets
cleaned_datasets = {
    "customers"  : customers_clean,
    "orders"     : orders_clean,
    "order_items": order_items_clean,
    "products"   : products_clean,
    "reviews"    : reviews_clean,
    "suppliers"  : suppliers_clean,
    "date"       : date_clean,
}

print(f"\n  Saving to: {PARQUET_CLEANED}")
print(f"\n  {'Dataset':<15} {'Rows':>10} "
      f"{'Columns':>8} {'Time':>8} {'Status':>8}")
print(f"  {'-'*55}")

total_rows  = 0
total_start = time_module.time()

for name, df in cleaned_datasets.items():
    t    = time_module.time()
    path = os.path.join(PARQUET_CLEANED, name)
    try:
        df_cached = df.cache()
        rows = df_cached.count()
        cols = len(df_cached.columns)
        df_cached.write \
            .mode("overwrite") \
            .parquet(path)
        df_cached.unpersist()
        elapsed    = time_module.time() - t
        total_rows += rows
        print(f"  {name:<15} {rows:>10,} "
              f"{cols:>8} {elapsed:>7.1f}s {' ':>8}")
    except Exception as e:
        elapsed = time_module.time() - t
        print(f"  {name:<15} {'ERROR':>10} "
              f"{'':>8} {elapsed:>7.1f}s {'x':>8}")
        print(f"    Cause: {str(e)[:200]}")
total_elapsed = time_module.time() - total_start
print(f"  {'-'*55}")
print(f"  {'TOTAL':<15} {total_rows:>10,} "
      f"{'':>8} {total_elapsed:>7.1f}s")

print(f"\n  Saved paths:")
for name in cleaned_datasets.keys():
    path = os.path.join(PARQUET_CLEANED, name)
    print(f"    {path}")

print("=" * 60)
print("    All 7 cleaned datasets saved to Parquet")
print("=" * 60)
```

#### 12.2 Save Star Schema to Parquet

```python
# 
# Save Star Schema Dimensions & Fact to Parquet
# Purpose: Persist final star schema tables
#          This is the authoritative warehouse output
#          These files are ready for Power BI, SSAS,
#          or any analytical tool consumption
# 


print(f"{'SAVE STAR SCHEMA TO PARQUET':^60}")


PARQUET_DIMS = os.path.join(
    OUTPUT_DIRS["parquet"], "dimensions")
PARQUET_FACT = os.path.join(
    OUTPUT_DIRS["parquet"], "fact")

os.makedirs(PARQUET_DIMS, exist_ok=True)
os.makedirs(PARQUET_FACT, exist_ok=True)

# -- Save 6 dimension tables -----------------------------------
dimensions = {
    "DIM_CUSTOMER": DIM_CUSTOMER,
    "DIM_PRODUCT" : DIM_PRODUCT,
    "DIM_SUPPLIER": DIM_SUPPLIER,
    "DIM_ORDER"   : DIM_ORDER,
    "DIM_REVIEW"  : DIM_REVIEW,
    "DIM_DATE"    : DIM_DATE,
}

print(f"\n  Saving dimension tables to:")
print(f"  {PARQUET_DIMS}")
print(f"\n  {'Table':<15} {'Rows':>10} "
      f"{'Columns':>8} {'Time':>8} {'Status':>8}")
print(f"  {'-'*55}")

dim_total_rows  = 0
dim_total_start = time_module.time()

for name, df in dimensions.items():
    t    = time_module.time()
    path = os.path.join(PARQUET_DIMS, name)
    rows = df.count()
    cols = len(df.columns)

    df.write \
        .mode("overwrite") \
        .parquet(path)

    elapsed = time_module.time() - t
    dim_total_rows += rows
    print(f"  {name:<15} {rows:>10,} "
          f"{cols:>8} {elapsed:>7.1f}s {' ':>8}")

dim_elapsed = time_module.time() - dim_total_start

# -- Save FACT_SALES -------------------------------------------
print(f"\n  Saving fact table to:")
print(f"  {PARQUET_FACT}")
print(f"\n  {'Table':<15} {'Rows':>10} "
      f"{'Columns':>8} {'Time':>8} {'Status':>8}")
print(f"  {'-'*55}")

t         = time_module.time()
fact_path = os.path.join(PARQUET_FACT, "FACT_SALES")
fact_rows = FACT_SALES.count()
fact_cols = len(FACT_SALES.columns)

FACT_SALES.write \
    .mode("overwrite") \
    .parquet(fact_path)

fact_elapsed = time_module.time() - t
print(f"  {'FACT_SALES':<15} {fact_rows:>10,} "
      f"{fact_cols:>8} {fact_elapsed:>7.1f}s {' ':>8}")

# -- Summary ---------------------------------------------------
print(f"\n  STAR SCHEMA SAVE SUMMARY:")
print(f"  {'='*45}")
print(f"  Dimension tables saved : {len(dimensions)}")
print(f"  Dimension rows total   : {dim_total_rows:,}")
print(f"  Fact table rows        : {fact_rows:,}")
print(f"  Total rows saved       : "
      f"{dim_total_rows + fact_rows:,}")
print(f"  Dimension save time    : {dim_elapsed:.1f}s")
print(f"  Fact save time         : {fact_elapsed:.1f}s")

print("=" * 60)
print("    All 6 dimensions saved to Parquet")
print("    FACT_SALES saved to Parquet")
print("=" * 60)
```

#### 12.3 Save ML & Analytics Results to Parquet

```python
# 
# Save ML & Analytics Results to Parquet
# Purpose: Persist all ML outputs and analytics results
#          Customer clusters, product clusters,
#          model predictions, MBA results
# 

print("=" * 60)
print(f"{'CELL 63: SAVE ML & ANALYTICS RESULTS':^60}")
print("=" * 60)

PARQUET_ML = os.path.join(
    OUTPUT_DIRS["parquet"], "ml_results")
os.makedirs(PARQUET_ML, exist_ok=True)

print(f"\n  Saving to: {PARQUET_ML}")
print(f"\n  {'Dataset':<30} {'Rows':>10} "
      f"{'Time':>8} {'Status':>8}")
print(f"  {'-'*55}")

ml_results = {}

# -- Customer clusters -----------------------------------------
try:
    t = time_module.time()
    rows = customer_clusters_labeled.count()
    path = os.path.join(PARQUET_ML,
                        "customer_clusters")
    customer_clusters_labeled.write \
        .mode("overwrite") \
        .parquet(path)
    elapsed = time_module.time() - t
    ml_results["customer_clusters"] = rows
    print(f"  {'customer_clusters':<30} "
          f"{rows:>10,} {elapsed:>7.1f}s {' ':>8}")
except Exception as e:
    print(f"  customer_clusters: {str(e)[:40]}")

# -- Product clusters ------------------------------------------
try:
    t = time_module.time()
    rows = prod_clusters_labeled.count()
    path = os.path.join(PARQUET_ML,
                        "product_clusters")
    prod_clusters_labeled.write \
        .mode("overwrite") \
        .parquet(path)
    elapsed = time_module.time() - t
    ml_results["product_clusters"] = rows
    print(f"  {'product_clusters':<30} "
          f"{rows:>10,} {elapsed:>7.1f}s {' ':>8}")
except Exception as e:
    print(f"  product_clusters: {str(e)[:40]}")

# -- Linear Regression predictions ----------------------------
try:
    t    = time_module.time()
    lr_save = lr_predictions.select(
        "sales_sk",
        "line_total",
        F.round("lr_prediction", 2).alias(
            "predicted_revenue"),
        F.round(
            F.abs(col("line_total") -
                  col("lr_prediction")), 2
        ).alias("prediction_error")
    )
    rows = lr_save.count()
    path = os.path.join(PARQUET_ML,
                        "lr_predictions")
    lr_save.write \
        .mode("overwrite") \
        .parquet(path)
    elapsed = time_module.time() - t
    ml_results["lr_predictions"] = rows
    print(f"  {'lr_revenue_predictions':<30} "
          f"{rows:>10,} {elapsed:>7.1f}s {' ':>8}")
except Exception as e:
    print(f"  lr_predictions: {str(e)[:40]}")

# -- Random Forest predictions ---------------------------------
try:
    t    = time_module.time()
    rf_save = rf_predictions.select(
        "review_score",
        F.round("rf_prediction", 2).alias(
            "predicted_review_score"),
        F.round(
            F.abs(col("review_score") -
                  col("rf_prediction")), 2
        ).alias("prediction_error")
    )
    rows = rf_save.count()
    path = os.path.join(PARQUET_ML,
                        "rf_predictions")
    rf_save.write \
        .mode("overwrite") \
        .parquet(path)
    elapsed = time_module.time() - t
    ml_results["rf_predictions"] = rows
    print(f"  {'rf_review_predictions':<30} "
          f"{rows:>10,} {elapsed:>7.1f}s {' ':>8}")
except Exception as e:
    print(f"  rf_predictions: {str(e)[:40]}")

# -- Logistic Regression predictions --------------------------
try:
    t    = time_module.time()
    log_save = log_predictions.select(
        "label",
        "log_prediction",
        "order_status",
        F.udf(lambda v: float(v[1]),
              FloatType()
        )("probability").alias("prob_delivered")
    )
    rows = log_save.count()
    path = os.path.join(PARQUET_ML,
                        "log_predictions")
    log_save.write \
        .mode("overwrite") \
        .parquet(path)
    elapsed = time_module.time() - t
    ml_results["log_predictions"] = rows
    print(f"  {'log_order_predictions':<30} "
          f"{rows:>10,} {elapsed:>7.1f}s {' ':>8}")
except Exception as e:
    print(f"  log_predictions: {str(e)[:40]}")

# -- RFM features ----------------------------------------------
try:
    t    = time_module.time()
    rows = rfm_df.count()
    path = os.path.join(PARQUET_ML, "rfm_features")
    rfm_df.select(
        "customer_sk",
        "customer_id",
        "customer_segment",
        "country",
        "recency_days",
        "frequency",
        "monetary",
        "avg_order_value",
        "avg_review_score"
    ).write \
        .mode("overwrite") \
        .parquet(path)
    elapsed = time_module.time() - t
    ml_results["rfm_features"] = rows
    print(f"  {'rfm_features':<30} "
          f"{rows:>10,} {elapsed:>7.1f}s {' ':>8}")
except Exception as e:
    print(f"  rfm_features: {str(e)[:40]}")

print(f"\n  ML results saved to: {PARQUET_ML}")
print("=" * 60)
print("    All ML results saved to Parquet")
print("=" * 60)
```

#### 12.4 Parquet File Size Comparison

```python
import builtins
# 
# Parquet File Size Comparison
# Purpose: Compare CSV vs Parquet file sizes
#          Demonstrate compression benefit of Parquet
#          Quantify storage efficiency gains
# 


print(f"{'CSV vs PARQUET SIZE COMPARISON':^60}")


def get_folder_size_mb(path):
    """Recursively get folder size in MB"""
    total = 0
    if not os.path.exists(path):
        return 0
    for dirpath, dirnames, filenames \
            in os.walk(path):
        for f in filenames:
            fp    = os.path.join(dirpath, f)
            total += os.path.getsize(fp)
    return total / (1024 * 1024)

def get_file_size_mb(path):
    """Get single file size in MB"""
    if os.path.exists(path):
        return os.path.getsize(path) / (1024 * 1024)
    return 0

print(f"\n  CSV vs PARQUET SIZE COMPARISON:")
print(f"  {'='*65}")
print(f"  {'Dataset':<16} {'CSV Size':>10} "
      f"{'Parquet Size':>14} {'Reduction':>11} "
      f"{'Ratio':>8}")
print(f"  {'-'*65}")

comparison_data = []

csv_parquet_map = {
    "customers"  : ("customers.csv",        "cleaned/customers"),
    "orders"     : ("orders.csv",           "cleaned/orders"),
    "order_items": ("order_items.csv",      "cleaned/order_items"),
    "products"   : ("products.csv",         "cleaned/products"),
    "reviews"    : ("reviews.csv",          "cleaned/reviews"),
    "suppliers"  : ("suppliers.csv",        "cleaned/suppliers"),
    "date"       : ("date_dim_helper.csv",  "cleaned/date"),
}

total_csv     = 0
total_parquet = 0

for name, (csv_file, parq_sub) in csv_parquet_map.items():
    csv_path  = os.path.join(
        BASE_PATH, "data", csv_file)
    parq_path = os.path.join(
        OUTPUT_DIRS["parquet"], parq_sub)

    csv_size  = get_file_size_mb(csv_path)
    parq_size = get_folder_size_mb(parq_path)

    if csv_size > 0 and parq_size > 0:
        reduction = ((csv_size - parq_size)
                     / csv_size * 100)
        ratio     = csv_size / parq_size \
            if parq_size > 0 else 0
    else:
        reduction = 0
        ratio     = 0

    total_csv     += csv_size
    total_parquet += parq_size

    comparison_data.append({
        "dataset"     : name,
        "csv_mb"      : builtins.round(csv_size, 2),
        "parquet_mb"  : builtins.round(parq_size, 2),
        "reduction_pct": builtins.round(reduction, 1),
        "ratio"       : builtins.round(ratio, 2)
    })

    print(f"  {name:<16} {csv_size:>9.1f}MB "
          f"{parq_size:>13.1f}MB "
          f"{reduction:>10.1f}% "
          f"{ratio:>7.1f}x")

total_reduction = ((total_csv - total_parquet)
                   / total_csv * 100) \
    if total_csv > 0 else 0
total_ratio = total_csv / total_parquet \
    if total_parquet > 0 else 0

print(f"  {'='*65}")
print(f"  {'TOTAL':<16} {total_csv:>9.1f}MB "
      f"{total_parquet:>13.1f}MB "
      f"{total_reduction:>10.1f}% "
      f"{total_ratio:>7.1f}x")
print(f"  {'='*65}")

# -- Star schema parquet sizes ---------------------------------
print(f"\n  STAR SCHEMA PARQUET SIZES:")
print(f"  {'='*45}")
print(f"  {'Table':<18} {'Size':>10} {'Rows':>10}")
print(f"  {'-'*45}")

schema_sizes = [
    ("DIM_CUSTOMER", "dimensions/DIM_CUSTOMER",
     DIM_CUSTOMER.count()),
    ("DIM_PRODUCT",  "dimensions/DIM_PRODUCT",
     DIM_PRODUCT.count()),
    ("DIM_SUPPLIER", "dimensions/DIM_SUPPLIER",
     DIM_SUPPLIER.count()),
    ("DIM_ORDER",    "dimensions/DIM_ORDER",
     DIM_ORDER.count()),
    ("DIM_REVIEW",   "dimensions/DIM_REVIEW",
     DIM_REVIEW.count()),
    ("DIM_DATE",     "dimensions/DIM_DATE",
     DIM_DATE.count()),
    ("FACT_SALES",   "fact/FACT_SALES",
     FACT_SALES.count()),
]

total_schema_size = 0
for table, sub_path, rows in schema_sizes:
    full_path = os.path.join(
        OUTPUT_DIRS["parquet"], sub_path)
    size = get_folder_size_mb(full_path)
    total_schema_size += size
    print(f"  {table:<18} {size:>9.2f}MB "
          f"{rows:>10,}")

print(f"  {'-'*45}")
print(f"  {'TOTAL':<18} "
      f"{total_schema_size:>9.2f}MB")
print(f"  {'='*45}")

# -- Save comparison CSV ---------------------------------------
comparison_df = pd.DataFrame(comparison_data)
comp_path = os.path.join(
    OUTPUT_DIRS["reports"],
    "parquet_size_comparison.csv"
)
comparison_df.to_csv(comp_path, index=False)
print(f"\n    Comparison saved: {comp_path}")

print("=" * 60)
print("    File size comparison complete")
print("=" * 60)
```

#### 12.5 Verify All Parquet Files

```python
# 
# Verify All Parquet Files
# Purpose: Read back ALL Parquet files and verify
#          Row counts match original DataFrames
#          Schema preserved correctly
#          Data integrity confirmed
# 


print(f"{'PARQUET VERIFICATION':^60}")


all_pass = True

# -- Verify cleaned datasets -----------------------------------
print(f"\n  CLEANED DATASETS VERIFICATION:")
print(f"  {'Dataset':<18} {'Parquet Rows':>14} {'Status':>8}")
print(f"  {'-'*45}")

cleaned_names = [
    "customers", "orders", "order_items",
    "products", "reviews", "suppliers", "date",
]

for name in cleaned_names:
    path = os.path.join(
        OUTPUT_DIRS["parquet"], "cleaned", name)
    try:
        df     = spark.read.parquet(path)
        actual = df.count()
        print(f"  {name:<18} {actual:>14,} {'  PASS':>8}")
    except Exception as e:
        print(f"  {name:<18} {'ERROR':>14} {'x FAIL':>8}")
        print(f"    Cause: {str(e)[:120]}")
        all_pass = False

# -- Verify star schema dimensions ----------------------------
print(f"\n  STAR SCHEMA VERIFICATION:")
print(f"  {'Table':<18} {'Parquet Rows':>14} {'Cols':>6} {'Status':>8}")
print(f"  {'-'*50}")

star_tables = {
    "DIM_CUSTOMER": "dimensions/DIM_CUSTOMER",
    "DIM_PRODUCT" : "dimensions/DIM_PRODUCT",
    "DIM_SUPPLIER": "dimensions/DIM_SUPPLIER",
    "DIM_ORDER"   : "dimensions/DIM_ORDER",
    "DIM_REVIEW"  : "dimensions/DIM_REVIEW",
    "DIM_DATE"    : "dimensions/DIM_DATE",
    "FACT_SALES"  : "fact/FACT_SALES",
}

for table, sub_path in star_tables.items():
    path = os.path.join(OUTPUT_DIRS["parquet"], sub_path)
    try:
        df     = spark.read.parquet(path)
        actual = df.count()
        cols   = len(df.columns)
        print(f"  {table:<18} {actual:>14,} {cols:>6} {'  PASS':>8}")
    except Exception as e:
        print(f"  {table:<18} {'ERROR':>14} {'':<6} {'x FAIL':>8}")
        print(f"    Cause: {str(e)[:120]}")
        all_pass = False

# -- Schema preservation check ---------------------------------
print(f"\n  SCHEMA PRESERVATION CHECK (FACT_SALES):")
print(f"  {'-'*45}")

fact_path = os.path.join(
    OUTPUT_DIRS["parquet"], "fact", "FACT_SALES")
try:
    fact_reload = spark.read.parquet(fact_path)
    orig_schema = {f.name: str(f.dataType)
                   for f in FACT_SALES.schema.fields}
    load_schema = {f.name: str(f.dataType)
                   for f in fact_reload.schema.fields}

    for col_name, orig_type in orig_schema.items():
        loaded_type = load_schema.get(col_name, "MISSING")
        match       = orig_type == loaded_type
        status      = " " if match else "x"
        if not match:
            all_pass = False
        print(f"  {col_name:<25} "
              f"{orig_type:<15} "
              f"{loaded_type:<15} {status}")
except Exception as e:
    print(f"  Could not verify schema: {str(e)[:120]}")
    all_pass = False

# -- Final output structure ------------------------------------
print(f"\n  COMPLETE PARQUET OUTPUT STRUCTURE:")
print(f"  {'-'*55}")

output_structure = {
    "parquet/cleaned/"    : "7 cleaned source datasets",
    "parquet/dimensions/" : "6 dimension tables",
    "parquet/fact/"       : "FACT_SALES table",
    "parquet/ml_results/" : "Clusters + predictions",
    "analytics/"          : "MBA frequent itemsets + rules",
    "clusters/"           : "Customer + product CSVs",
    "models/"             : "Model performance summary",
    "streaming/"          : "Stream results + summary",
    "plots/"              : "All visualization PNGs",
    "reports/"            : "Size comparison CSV",
}

for path, description in output_structure.items():
    full_path = os.path.join(OUTPUT_PATH, path)
    exists    = os.path.exists(full_path)
    status    = " " if exists else "x"
    print(f"  {status} output/{path:<30} {description}")

print(f"\n  {'='*55}")
overall = "  ALL PASS" if all_pass else "x SOME FAILED"
print(f"  Overall Status : {overall}")
print(f"  {'='*55}")

print("\n" + "=" * 60)
print("    All Parquet files verified")
print("    Row counts confirmed from saved files")
print("    Schema types preserved")
print("    Section 12 complete")
print("    Proceed to Section 13: Parquet Verification")
print("=" * 60)
```

#### 13.2 Star Schema & Analytics Summary

```python
# 
# Star Schema & Analytics Summary
# Purpose: Document star schema build results
#          Summarise all EDA business insights
#          Document key metrics discovered
# 


print(f"{'STAR SCHEMA & ANALYTICS SUMMARY':^65}")


# -- Section 5: Star Schema ------------------------------------
print(f"\n  SECTION 5: STAR SCHEMA BUILD")
print(f"  {'-'*55}")
print(f"  Schema type          : Kimball Star Schema")
print(f"  Fact tables          : 1 (FACT_SALES)")
print(f"  Dimension tables     : 6")
print(f"  Total FK relations   : 6 verified")

schema_summary = [
    ("DIM_CUSTOMER", "customer_sk", 120000,  10),
    ("DIM_PRODUCT",  "product_sk",  5000,     7),
    ("DIM_SUPPLIER", "supplier_sk", 200,      6),
    ("DIM_ORDER",    "order_sk",    398800,   7),
    ("DIM_REVIEW",   "review_sk",   280000,   5),
    ("DIM_DATE",     "date_sk",     2557,    13),
    ("FACT_SALES",   "sales_sk",    916167,  14),
]
print(f"\n  {'Table':<15} {'PK':<15} "
      f"{'Rows':>10} {'Cols':>6}")
print(f"  {'-'*50}")
for table, pk, rows, cols in schema_summary:
    print(f"  {table:<15} {pk:<15} "
          f"{rows:>10,} {cols:>6}")
print(f"  Status               :   COMPLETE")

# -- Section 6: EDA Key Insights ------------------------------
print(f"\n  SECTION 6: EDA & BUSINESS INTELLIGENCE")
print(f"  {'-'*55}")

# Compute live KPIs
kpi_live = spark.sql("""
    SELECT
        ROUND(SUM(f.line_total), 2)           AS total_revenue,
        COUNT(DISTINCT f.order_sk)            AS total_orders,
        COUNT(DISTINCT f.customer_sk)         AS unique_customers,
        COUNT(DISTINCT f.product_sk)          AS unique_products,
        COUNT(DISTINCT f.supplier_sk)         AS unique_suppliers,
        ROUND(AVG(f.line_total), 2)           AS avg_order_value,
        ROUND(AVG(f.discount_pct)*100, 2)     AS avg_discount_pct,
        ROUND(AVG(CASE WHEN f.review_score>0
              THEN f.review_score END), 3)    AS avg_review_score
    FROM FACT_SALES f
""").collect()[0]

print(f"  Total Revenue        : "
      f"${kpi_live['total_revenue']:>15,.2f}")
print(f"  Total Orders         : "
      f"{kpi_live['total_orders']:>15,}")
print(f"  Unique Customers     : "
      f"{kpi_live['unique_customers']:>15,}")
print(f"  Unique Products      : "
      f"{kpi_live['unique_products']:>15,}")
print(f"  Unique Suppliers     : "
      f"{kpi_live['unique_suppliers']:>15,}")
print(f"  Avg Order Value      : "
      f"${kpi_live['avg_order_value']:>15,.2f}")
print(f"  Avg Discount Rate    : "
      f"{kpi_live['avg_discount_pct']:>14.2f}%")
print(f"  Avg Review Score     : "
      f"{kpi_live['avg_review_score']:>15.3f}/5.0")

# Top category
top_cat = spark.sql("""
    SELECT p.category,
           ROUND(SUM(f.line_total), 2) AS rev
    FROM FACT_SALES f
    JOIN DIM_PRODUCT p ON f.product_sk = p.product_sk
    GROUP BY p.category
    ORDER BY rev DESC LIMIT 1
""").collect()[0]

# Top country
top_country = spark.sql("""
    SELECT c.country,
           ROUND(SUM(f.line_total), 2) AS rev
    FROM FACT_SALES   f
    JOIN DIM_CUSTOMER c ON f.customer_sk = c.customer_sk
    GROUP BY c.country
    ORDER BY rev DESC LIMIT 1
""").collect()[0]

# Best year
best_year = spark.sql("""
    SELECT d.year,
           ROUND(SUM(f.line_total), 2) AS rev
    FROM FACT_SALES f
    JOIN DIM_DATE   d ON f.date_sk = d.date_sk
    GROUP BY d.year
    ORDER BY rev DESC LIMIT 1
""").collect()[0]

print(f"\n  KEY FINDINGS:")
print(f"  Top category         : "
      f"{top_cat['category']} "
      f"(${top_cat['rev']:,.0f})")
print(f"  Top country          : "
      f"{top_country['country']} "
      f"(${top_country['rev']:,.0f})")
print(f"  Best revenue year    : "
      f"{best_year['year']} "
      f"(${best_year['rev']:,.0f})")
print(f"  Visualizations       : 7 charts generated")
print(f"  Status               :   COMPLETE")

print("\n" + "=" * 65)
print("    Star schema and analytics summary complete")
print("=" * 65)
```
